## 🟦 Ejercicio 1: Autovalores y Autovectores de una Matriz $2\times 2$

Sea:
$$A = \begin{pmatrix} 5 & -1 \\ 3 & 1 \end{pmatrix}$$

**Tareas:**
1. Calcula los autovalores y autovectores usando `np.linalg.eig()`.
2. Verifica cada autovalor con la ecuación $A\mathbf{v} = \lambda\mathbf{v}$.
3. Calcula la traza y el determinante para verificar las propiedades $\text{tr}(A)=\sum\lambda_i$ y $\det(A)=\prod\lambda_i$.

````python
import numpy as np

A = np.array([[5.0, -1.0],
              [3.0,  1.0]])

print('Ejercicio 1: A = [[5,-1],[3,1]]')
print('='*50)

# 1. Calcular autovalores y autovectores
eigenvalues, eigenvectors = np.linalg.eig(A)
print(f'Autovalores: λ = {eigenvalues}')
print(f'Autovectores (columnas de P):')
print(eigenvectors)

# 2. Verificar A*v = λ*v para cada par
print('\nVerificación Av = λv:')
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v   = eigenvectors[:, i]
    Av  = A @ v
    lv  = lam * v
    ok  = '✅' if np.allclose(Av, lv) else '❌'
    print(f'  λ = {lam:.4f}: Av = {np.round(Av,4)}, λv = {np.round(lv,4)} {ok}')

# 3. Verificar propiedades de traza y determinante
print('\nPropiedades de traza y determinante:')
tr = np.trace(A)
det = np.linalg.det(A)
sum_lam = np.sum(eigenvalues)
prod_lam = np.prod(eigenvalues)
print(f'  tr(A) = {tr:.4f} = Σλᵢ = {sum_lam.real:.4f}  {"✅" if np.isclose(tr, sum_lam.real) else "❌"}')
print(f'  det(A) = {det:.4f} = Πλᵢ = {prod_lam.real:.4f}  {"✅" if np.isclose(det, prod_lam.real) else "❌"}')
````

## 🟩 Ejercicio 2: Verificar que los Autovectores Forman una Base

Los autovectores de una matriz diagonalizable deben ser **linealmente independientes**,
lo que permite construir la base $\{\mathbf{v}_1, \mathbf{v}_2\}$.

**Tareas:**
1. Verifica que la matriz $P$ (autovectores como columnas) es invertible.
2. Calcula el determinante de $P$ (si $\neq 0$, los vectores son LI).
3. Calcula el número de condición de $P$.

````python
import numpy as np

A = np.array([[5.0, -1.0], [3.0, 1.0]])
eigenvalues, eigenvectors = np.linalg.eig(A)

P = eigenvectors  # columnas = autovectores

print('Ejercicio 2: ¿Son los autovectores linealmente independientes?')
print('='*55)

# 1. Determinante de P
det_P = np.linalg.det(P)
print(f'det(P) = {det_P:.6f}')
print(f'¿P invertible (det ≠ 0)? {not np.isclose(det_P, 0)} ✅')

# 2. Rango de P
rank_P = np.linalg.matrix_rank(P)
print(f'rango(P) = {rank_P} (= n = 2) ✅')

# 3. Número de condición
cond_P = np.linalg.cond(P)
print(f'Número de condición: {cond_P:.4f}')
print('(Un número de condición cercano a 1 indica vectores casi ortogonales)')

# 4. Verificar que P*P^{-1} = I
P_inv = np.linalg.inv(P)
I_check = P @ P_inv
print(f'\nP · P⁻¹ =\n{np.round(I_check, 10)}')
print(f'¿P · P⁻¹ = I? {np.allclose(I_check, np.eye(2))} ✅')

# 5. Verificar diagonalización completa
D = np.diag(eigenvalues)
A_check = P @ D @ P_inv
print(f'\nVerificación A = PDP⁻¹:')
print(f'¿A = PDP⁻¹? {np.allclose(A, A_check.real)} ✅')
````

## 🟨 Ejercicio 3: Calcular $A^{15}$ usando Diagonalización

Usando la fórmula $A^k = PD^kP^{-1}$, calcula $A^{15}$ para la misma matriz del Ejercicio 1.

**¿Por qué es útil?**
- Multiplicar $A$ directamente 15 veces requiere 14 multiplicaciones de matrices.
- Con diagonalización: solo necesitamos elevar los **escalares** $\lambda_1^{15}$ y $\lambda_2^{15}$.

**Verifica** que ambos métodos dan el mismo resultado.

````python
import numpy as np

A = np.array([[5.0, -1.0], [3.0, 1.0]])
eigenvalues, eigenvectors = np.linalg.eig(A)
P = eigenvectors
P_inv = np.linalg.inv(P)
k = 15

print(f'Ejercicio 3: Calcular A^{k}')
print('='*50)

# Método 1: Diagonalización (eficiente)
D_k = np.diag(eigenvalues**k)  # solo potencias de escalares
A_power_diag = P @ D_k @ P_inv

print(f'Método 1 — Diagonalización:')
print(f'  λ₁^{k} = {eigenvalues[0]:.2f}^{k} = {eigenvalues[0]**k:.4e}')
print(f'  λ₂^{k} = {eigenvalues[1]:.2f}^{k} = {eigenvalues[1]**k:.4e}')
print(f'\n  A^{k} =\n{np.round(A_power_diag.real)}')

# Método 2: Multiplicación directa (referencia)
A_power_direct = np.eye(2)
for _ in range(k):
    A_power_direct = A_power_direct @ A

print(f'\nMétodo 2 — Multiplicación directa:')
print(f'  A^{k} =\n{np.round(A_power_direct)}')

# Verificar igualdad
print(f'\n¿Ambos métodos coinciden? {np.allclose(A_power_diag.real, A_power_direct, rtol=1e-6)} ✅')
print(f'\nNota: los valores son muy grandes porque λ₁ = {eigenvalues[0]:.1f} > 1')
print(f'      ({eigenvalues[0]:.1f}^{k} ≈ {eigenvalues[0]**k:.3e})')
````

## 🟥 Ejercicio 4: Autovalores de una Matriz Simétrica $3\times 3$

Sea:
$$S = \begin{pmatrix} 4 & 2 & 0 \\ 2 & 3 & 1 \\ 0 & 1 & 2 \end{pmatrix}$$

**Tareas:**
1. Verifica que $S$ es simétrica.
2. Calcula sus autovalores y verifica que todos son **reales**.
3. Verifica que sus autovectores son **ortogonales** entre sí.
4. Construye la descomposición espectral $S = QDQ^T$.

````python
import numpy as np

S = np.array([[4.0, 2.0, 0.0],
              [2.0, 3.0, 1.0],
              [0.0, 1.0, 2.0]])

print('Ejercicio 4: Matriz simétrica 3×3')
print('='*50)

# 1. Verificar simetría
print(f'1. ¿S = Sᵀ? {np.allclose(S, S.T)} ✅')

# 2. Autovalores
eigenvalues, eigenvectors = np.linalg.eig(S)
print(f'\n2. Autovalores: {np.round(eigenvalues, 6)}')
print(f'   ¿Todos reales? {np.allclose(eigenvalues.imag, 0)} ✅')
print(f'   (Para matrices simétricas, el Teorema Espectral garantiza autovalores reales)')

# 3. Ortogonalidad de autovectores
print(f'\n3. Productos punto entre autovectores:')
for i in range(3):
    for j in range(i+1, 3):
        dot = np.dot(eigenvectors[:, i], eigenvectors[:, j])
        print(f'   v_{i+1} · v_{j+1} = {dot:.2e}  {"≈ 0 ✅" if abs(dot) < 1e-10 else "❌"}')

# 4. Descomposición espectral S = Q D Qᵀ
Q = eigenvectors  # columnas ortogonales
D = np.diag(eigenvalues.real)
S_check = Q @ D @ Q.T
print(f'\n4. Verificación S = QDQᵀ: {np.allclose(S, S_check.real, atol=1e-10)} ✅')
print(f'   ¿Q ortogonal (QᵀQ = I)? {np.allclose(Q.T @ Q, np.eye(3), atol=1e-10)} ✅')
print(f'\n   Descomposición espectral:')
print(f'   S = λ₁v₁v₁ᵀ + λ₂v₂v₂ᵀ + λ₃v₃v₃ᵀ')
for i, lam in enumerate(eigenvalues.real):
    vi = eigenvectors[:, i:i+1]
    print(f'   Término {i+1}: {lam:.4f} × v{i+1}v{i+1}ᵀ')
````

## 🟪 Ejercicio 5: Cadena de Markov — Convergencia al Estado Estacionario

Una empresa de renta de autos tiene 3 localizaciones: A, B, C.  
Cada día, los clientes devuelven los autos según la siguiente matriz de transición:

$$T = \begin{pmatrix} 0.8 & 0.1 & 0.2 \\ 0.1 & 0.7 & 0.2 \\ 0.1 & 0.2 & 0.6 \end{pmatrix}$$

**Tareas:**
1. Verifica que $T$ es estocástica (columnas suman 1).
2. Encuentra el autovector correspondiente a $\lambda = 1$ (distribución estacionaria).
3. Simula la evolución desde distintos estados iniciales y grafica la convergencia.

````python
import numpy as np
import matplotlib.pyplot as plt

T = np.array([[0.8, 0.1, 0.2],
              [0.1, 0.7, 0.2],
              [0.1, 0.2, 0.6]])

locaciones = ['Sucursal A', 'Sucursal B', 'Sucursal C']

print('Ejercicio 5: Cadena de Markov — Empresa de Renta de Autos')
print('='*55)

# 1. Verificar estocástica
col_sums = T.sum(axis=0)
print(f'1. Sumas de columnas: {np.round(col_sums, 4)}')
print(f'   ¿Estocástica? {np.allclose(col_sums, 1.0)} ✅')

# 2. Autovector para λ = 1
eigenvalues, eigenvectors = np.linalg.eig(T)
print(f'\n2. Autovalores: {np.round(eigenvalues.real, 6)}')

idx = np.argmin(np.abs(eigenvalues - 1.0))
steady = eigenvectors[:, idx].real
steady = steady / steady.sum()

print(f'\n   Distribución estacionaria (autovector de λ=1):')
for loc, prob in zip(locaciones, steady):
    print(f'   P({loc}) = {prob:.4f} ({prob*100:.1f}%)')

# Verificar: T * x* = x*
check = T @ steady
print(f'\n   Verificación T·x* = x*: {np.allclose(check, steady)} ✅')

# 3. Simulación desde distintos estados iniciales
estados_iniciales = [
    np.array([1.0, 0.0, 0.0]),  # todos en A
    np.array([0.0, 1.0, 0.0]),  # todos en B
    np.array([1/3, 1/3, 1/3]), # distribuido uniformemente
]

fig, ax = plt.subplots(figsize=(10, 6))
colors_loc = ['#e74c3c', '#3498db', '#27ae60']
n_steps = 25

for x0 in estados_iniciales:
    history = [x0.copy()]
    x = x0.copy()
    for _ in range(n_steps):
        x = T @ x
        history.append(x.copy())
    history = np.array(history)

    for j, (loc, color) in enumerate(zip(locaciones, colors_loc)):
        ls = '-' if j == 0 else ('--' if j == 1 else ':')
        label = f'{loc} (x₀={tuple(np.round(x0,2))})' if j == 0 else None
        ax.plot(history[:, j], color=color, lw=1.8, ls=ls,
                alpha=0.7, label=label if x0[0]==1.0 else None)

for j, (loc, color) in enumerate(zip(locaciones, colors_loc)):
    ax.axhline(steady[j], color=color, lw=2.5, ls='-', alpha=0.9,
               label=f'Estacionario {loc}: {steady[j]:.3f}')

ax.set_xlabel('Pasos de tiempo', fontsize=11)
ax.set_ylabel('Proporción de autos', fontsize=11)
ax.set_title('Convergencia al estado estacionario\n(todas las condiciones iniciales convergen)', fontsize=11)
ax.legend(fontsize=9, loc='right')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('ejercicio5_markov.png', dpi=100, bbox_inches='tight')
plt.show()
print('\nSin importar el estado inicial, la distribución converge al mismo estado estacionario.')
````
