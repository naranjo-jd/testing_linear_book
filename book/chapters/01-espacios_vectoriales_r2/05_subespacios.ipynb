#  Subespacios de $\mathbb{R}^2$

En esta sección estudiaremos los **subespacios vectoriales** de $\mathbb{R}^2$. Un subespacio es, intuitivamente, un "espacio vectorial más pequeño" que vive dentro de otro espacio vectorial más grande.

Piensa en los subespacios como subconjuntos "bien portados" de $\mathbb{R}^2$: conjuntos que respetan las operaciones de suma y multiplicación por escalar que definen a un espacio vectorial.

---

## 🔹 1. Definición de Subespacio

:::{admonition} Definición: Subespacio Vectorial
:class: tip

Un conjunto no vacío $S \subseteq \mathbb{R}^2$ es un **subespacio vectorial** de $\mathbb{R}^2$ si cumple las siguientes tres condiciones:

1. **Contiene el vector cero:** $\mathbf{0} \in S$

2. **Cerrado bajo la suma:** Si $\mathbf{u}, \mathbf{v} \in S$, entonces $\mathbf{u} + \mathbf{v} \in S$

3. **Cerrado bajo multiplicación por escalar:** Si $\mathbf{v} \in S$ y $\alpha \in \mathbb{R}$, entonces $\alpha \mathbf{v} \in S$

:::

### Explicación de cada condición

**Condición 1: Contiene el vector cero**

Todo subespacio debe contener el origen $\mathbf{0} = (0, 0)$. Esta es la condición más fácil de verificar y, si un conjunto no la cumple, **automáticamente no es un subespacio**.

**Condición 2: Cerrado bajo la suma**

Si tomamos dos vectores cualesquiera del conjunto y los sumamos, el resultado debe estar también en el conjunto. Esto significa que "no podemos salirnos" del subespacio al sumar vectores.

**Condición 3: Cerrado bajo multiplicación por escalar**

Si tomamos cualquier vector del conjunto y lo multiplicamos por cualquier número real, el resultado debe permanecer en el conjunto. Esto incluye multiplicar por negativos (lo que invierte la dirección) y por valores entre 0 y 1 (lo que acorta el vector).

---

## 🔹 2. Clasificación de Subespacios de $\mathbb{R}^2$

:::{admonition} Teorema: Clasificación de Subespacios de $\mathbb{R}^2$
:class: important

Los **únicos** subespacios de $\mathbb{R}^2$ son:

1. **El subespacio trivial:** $\{\mathbf{0}\} = \{(0, 0)\}$ (solo el origen)

2. **Rectas que pasan por el origen:** Líneas de la forma $y = mx$ o el eje $y$

3. **Todo $\mathbb{R}^2$:** El espacio completo

:::

Este es un resultado fundamental: en $\mathbb{R}^2$ **no hay otros tipos de subespacios**. Por ejemplo:
- Una recta que NO pasa por el origen **no es** un subespacio
- Un círculo **no es** un subespacio
- El primer cuadrante **no es** un subespacio

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(14, 4.5))

# Configuración común
for ax in axes:
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_aspect('equal')
    ax.axhline(y=0, color='lightgray', linewidth=0.5)
    ax.axvline(x=0, color='lightgray', linewidth=0.5)
    ax.grid(True, alpha=0.3)

# 1. Subespacio trivial {0}
axes[0].scatter([0], [0], color='red', s=150, zorder=5, label='Origen')
axes[0].set_title(r'1. Subespacio trivial $\{\mathbf{0}\}$', fontsize=12, fontweight='bold')
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$y$')
axes[0].annotate(r'$\{(0,0)\}$', xy=(0, 0), xytext=(0.5, 0.8),
                 fontsize=11, arrowprops=dict(arrowstyle='->', color='red'))

# 2. Rectas por el origen
t = np.linspace(-3, 3, 100)
# Recta y = 2x
axes[1].plot(t, 2*t, 'b-', linewidth=2.5, label=r'$y = 2x$')
# Recta y = -0.5x
axes[1].plot(t, -0.5*t, 'g-', linewidth=2.5, label=r'$y = -\frac{1}{2}x$')
# Eje x (y = 0)
axes[1].plot(t, 0*t, 'purple', linewidth=2, linestyle='--', label=r'$y = 0$ (eje $x$)')
axes[1].scatter([0], [0], color='red', s=100, zorder=5)
axes[1].set_title('2. Rectas por el origen', fontsize=12, fontweight='bold')
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$y$')
axes[1].legend(loc='upper left', fontsize=9)

# 3. Todo R²
# Rellenar todo el plano con un color suave
axes[2].fill([-3, 3, 3, -3], [-3, -3, 3, 3], alpha=0.3, color='orange')
axes[2].scatter([0], [0], color='red', s=100, zorder=5)
axes[2].set_title(r'3. Todo $\mathbb{R}^2$', fontsize=12, fontweight='bold')
axes[2].set_xlabel('$x$')
axes[2].set_ylabel('$y$')
axes[2].text(0, 0, r'$\mathbb{R}^2$', fontsize=20, ha='center', va='center', color='darkorange', fontweight='bold')

plt.tight_layout()
plt.show()
```

## 🔹 3. Ejemplos y No-Ejemplos

Veamos cómo verificar si un conjunto es o no un subespacio aplicando las tres condiciones.

### [OK] Ejemplo 1: La recta $y = 2x$ ES un subespacio

Sea $S = \{(x, y) \in \mathbb{R}^2 : y = 2x\}$

**Verificación de las tres condiciones:**

**1. ¿Contiene el vector cero?**

Sustituyendo $x = 0$: $y = 2(0) = 0$. 

Sí, $(0, 0) \in S$ ✓

**2. ¿Cerrado bajo la suma?**

Sean $\mathbf{u} = (u_1, 2u_1)$ y $\mathbf{v} = (v_1, 2v_1)$ dos vectores en $S$.

$$\mathbf{u} + \mathbf{v} = (u_1 + v_1, 2u_1 + 2v_1) = (u_1 + v_1, 2(u_1 + v_1))$$

Este vector satisface $y = 2x$, así que $\mathbf{u} + \mathbf{v} \in S$ ✓

**3. ¿Cerrado bajo multiplicación por escalar?**

Sea $\mathbf{v} = (v_1, 2v_1) \in S$ y $\alpha \in \mathbb{R}$.

$$\alpha \mathbf{v} = (\alpha v_1, 2\alpha v_1) = (\alpha v_1, 2(\alpha v_1))$$

Este vector satisface $y = 2x$, así que $\alpha \mathbf{v} \in S$ ✓

**Conclusión:** $S$ es un subespacio de $\mathbb{R}^2$.

```python
fig, ax = plt.subplots(figsize=(8, 6))

t = np.linspace(-3, 3, 100)
ax.plot(t, 2*t, 'b-', linewidth=2.5, label=r'Subespacio $S$: $y = 2x$')

# Mostrar dos vectores y su suma
u = np.array([1, 2])
v = np.array([0.5, 1])
suma = u + v

ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.02, label=r'$\mathbf{u} = (1, 2)$')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='orange', width=0.02, label=r'$\mathbf{v} = (0.5, 1)$')
ax.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.02, label=r'$\mathbf{u}+\mathbf{v} = (1.5, 3)$')

ax.scatter([0], [0], color='black', s=80, zorder=5)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_aspect('equal')
ax.axhline(y=0, color='lightgray', linewidth=0.5)
ax.axvline(x=0, color='lightgray', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title(r'Ejemplo 1: $y = 2x$ es un subespacio', fontsize=12, fontweight='bold')
ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

### [X] Ejemplo 2: La recta $y = x + 1$ NO es un subespacio

Sea $T = \{(x, y) \in \mathbb{R}^2 : y = x + 1\}$

**Verificación de la condición 1:**

¿Contiene el vector cero? Sustituyendo $x = 0$: $y = 0 + 1 = 1 \neq 0$.

El punto $(0, 0)$ NO está en $T$ porque $0 \neq 0 + 1$.

**Conclusión:** $T$ **no es** un subespacio de $\mathbb{R}^2$ (falla la condición 1).

:::{admonition} Observación importante
:class: warning

**Toda recta que NO pasa por el origen NO puede ser un subespacio.** Esta es la forma más rápida de descartar muchos conjuntos.

:::

```python
fig, ax = plt.subplots(figsize=(8, 6))

t = np.linspace(-3, 3, 100)
ax.plot(t, t + 1, 'r-', linewidth=2.5, label=r'$y = x + 1$ (NO es subespacio)')

# Marcar el origen con una X
ax.scatter([0], [0], color='black', s=100, marker='o', zorder=5, label='Origen (0,0)')
ax.scatter([0], [1], color='red', s=100, marker='s', zorder=5, label='Punto donde $x=0$: $(0, 1)$')

# Línea punteada mostrando que no pasa por el origen
ax.annotate('', xy=(0, 0), xytext=(0, 1),
            arrowprops=dict(arrowstyle='<->', color='purple', lw=2))
ax.text(0.15, 0.5, 'Falta el\norigen', fontsize=10, color='purple')

ax.set_xlim(-3, 3)
ax.set_ylim(-2, 4)
ax.set_aspect('equal')
ax.axhline(y=0, color='lightgray', linewidth=0.5)
ax.axvline(x=0, color='lightgray', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title(r'Ejemplo 2: $y = x + 1$ NO es subespacio', fontsize=12, fontweight='bold')
ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

### [X] No-Ejemplo: El primer cuadrante NO es un subespacio

Sea $Q_1 = \{(x, y) \in \mathbb{R}^2 : x \geq 0 \text{ y } y \geq 0\}$ (el primer cuadrante)

**Verificación:**

**1. ¿Contiene el vector cero?**

Sí, $(0, 0) \in Q_1$ porque $0 \geq 0$ y $0 \geq 0$ ✓

**2. ¿Cerrado bajo la suma?**

Si $(x_1, y_1)$ y $(x_2, y_2)$ están en $Q_1$, entonces $x_1, y_1, x_2, y_2 \geq 0$.

Por lo tanto, $(x_1 + x_2, y_1 + y_2)$ también tiene componentes no negativas. ✓

**3. ¿Cerrado bajo multiplicación por escalar?**

Tomemos $\mathbf{v} = (1, 1) \in Q_1$ y $\alpha = -1$.

$$\alpha \mathbf{v} = -1 \cdot (1, 1) = (-1, -1)$$

Pero $(-1, -1) \notin Q_1$ porque tiene componentes negativas. ✗

**Conclusión:** El primer cuadrante **no es** un subespacio (falla la condición 3).

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Rellenar el primer cuadrante
ax.fill([0, 3, 3, 0], [0, 0, 3, 3], alpha=0.3, color='blue', label='Primer cuadrante $Q_1$')

# Vector v en Q1
v = np.array([1, 1])
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.02, label=r'$\mathbf{v} = (1, 1)$')

# Vector -v fuera de Q1
ax.quiver(0, 0, -v[0], -v[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.02, label=r'$-\mathbf{v} = (-1, -1) \notin Q_1$')

ax.scatter([0], [0], color='black', s=80, zorder=5)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_aspect('equal')
ax.axhline(y=0, color='gray', linewidth=1)
ax.axvline(x=0, color='gray', linewidth=1)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('El primer cuadrante NO es subespacio', fontsize=12, fontweight='bold')
ax.legend(loc='upper left')

# Añadir texto explicativo
ax.text(-1.5, -1.5, r'$-\mathbf{v}$ está fuera', fontsize=10, color='red', ha='center')

plt.tight_layout()
plt.show()
```

## 🔹 4. Subespacio Generado (Span)

:::{admonition} Definición: Subespacio Generado
:class: tip

Dado un conjunto de vectores $\{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k\}$ en $\mathbb{R}^2$, el **subespacio generado** por estos vectores (también llamado **span**) es:

$$\text{span}\{\mathbf{v}_1, \ldots, \mathbf{v}_k\} = \{\alpha_1 \mathbf{v}_1 + \alpha_2 \mathbf{v}_2 + \cdots + \alpha_k \mathbf{v}_k : \alpha_i \in \mathbb{R}\}$$

Es decir, el conjunto de **todas las combinaciones lineales** de esos vectores.

:::

### Casos importantes en $\mathbb{R}^2$:

| Vectores | Subespacio Generado |
|----------|---------------------|
| $\text{span}\{\mathbf{0}\}$ | $\{\mathbf{0}\}$ (subespacio trivial) |
| $\text{span}\{\mathbf{v}\}$ con $\mathbf{v} \neq \mathbf{0}$ | Recta que pasa por el origen en dirección $\mathbf{v}$ |
| $\text{span}\{\mathbf{v}_1, \mathbf{v}_2\}$ con $\mathbf{v}_1, \mathbf{v}_2$ **linealmente independientes** | Todo $\mathbb{R}^2$ |
| $\text{span}\{\mathbf{v}_1, \mathbf{v}_2\}$ con $\mathbf{v}_1, \mathbf{v}_2$ **linealmente dependientes** | Una recta por el origen |

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Caso 1: Un solo vector -> recta
ax1 = axes[0]
v = np.array([1, 2])
t = np.linspace(-2, 2, 100)
ax1.plot(t * v[0], t * v[1], 'b-', linewidth=2.5, label=r'span$\{\mathbf{v}\}$')
ax1.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.02, label=r'$\mathbf{v} = (1, 2)$')
ax1.scatter([0], [0], color='black', s=80, zorder=5)
ax1.set_xlim(-3, 3)
ax1.set_ylim(-3, 3)
ax1.set_aspect('equal')
ax1.axhline(y=0, color='lightgray', linewidth=0.5)
ax1.axvline(x=0, color='lightgray', linewidth=0.5)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title(r'span$\{\mathbf{v}\}$ = Recta por el origen', fontsize=12, fontweight='bold')
ax1.legend(loc='upper left')

# Caso 2: Dos vectores independientes -> todo R²
ax2 = axes[1]
v1 = np.array([1, 0])
v2 = np.array([0, 1])
ax2.fill([-3, 3, 3, -3], [-3, -3, 3, 3], alpha=0.2, color='orange')
ax2.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.02, label=r'$\mathbf{v}_1 = (1, 0)$')
ax2.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.02, label=r'$\mathbf{v}_2 = (0, 1)$')
ax2.scatter([0], [0], color='black', s=80, zorder=5)
ax2.set_xlim(-3, 3)
ax2.set_ylim(-3, 3)
ax2.set_aspect('equal')
ax2.axhline(y=0, color='lightgray', linewidth=0.5)
ax2.axvline(x=0, color='lightgray', linewidth=0.5)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_title(r'span$\{\mathbf{v}_1, \mathbf{v}_2\}$ = $\mathbb{R}^2$ (si son l.i.)', fontsize=12, fontweight='bold')
ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

## 🔹 5. Conexión con Sistemas Homogéneos

:::{admonition} Teorema: Soluciones de Sistemas Homogéneos
:class: important

El conjunto de soluciones de un **sistema homogéneo** $A\mathbf{x} = \mathbf{0}$ siempre es un **subespacio** del espacio de vectores columna.

:::

Un sistema homogéneo es aquel donde el lado derecho es el vector cero. Las soluciones siempre forman un subespacio porque:

1. $\mathbf{x} = \mathbf{0}$ siempre es solución (el vector cero está en el conjunto)
2. Si $\mathbf{x}_1$ y $\mathbf{x}_2$ son soluciones, entonces $\mathbf{x}_1 + \mathbf{x}_2$ también lo es
3. Si $\mathbf{x}$ es solución y $\alpha \in \mathbb{R}$, entonces $\alpha\mathbf{x}$ también es solución

### Ejemplo: Sistema $x + 2y = 0$

Consideremos el sistema homogéneo en $\mathbb{R}^2$:

$$x + 2y = 0$$

Despejando $x$: $x = -2y$

Las soluciones tienen la forma:
$$\mathbf{x} = \begin{pmatrix} -2y \\ y \end{pmatrix} = y \begin{pmatrix} -2 \\ 1 \end{pmatrix}$$

Esto es una **recta que pasa por el origen** en dirección $(-2, 1)$.

```python
import numpy as np

# Definir la matriz del sistema x + 2y = 0
A = np.array([[1, 2]])
b = np.array([0])

# Encontrar la solución paramétrica
# Si y = t, entonces x = -2t
print("Sistema: x + 2y = 0")
print("\nSolución paramétrica:")
print("  x = -2t")
print("  y = t")
print("\nVector director del subespacio: (-2, 1)")

# Verificar algunas soluciones
print("\nVerificación de soluciones:")
for t in [-1, 0, 1, 2]:
    x, y = -2*t, t
    resultado = A @ np.array([x, y])
    print(f"  t = {t}: ({x}, {y}) -> A·x = {resultado[0]} ✓" if resultado[0] == 0 else f"  t = {t}: ({x}, {y}) -> A·x = {resultado[0]} ✗")
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Recta de soluciones: x = -2y, o equivalentemente y = -x/2
t = np.linspace(-3, 3, 100)
direccion = np.array([-2, 1])
ax.plot(t * direccion[0], t * direccion[1], 'b-', linewidth=2.5, label=r'Subespacio de soluciones de $x + 2y = 0$')

# Mostrar algunos puntos solución
for t_val in [-1, 0, 0.5, 1]:
    punto = t_val * direccion
    ax.scatter(punto[0], punto[1], color='red', s=60, zorder=5)
    ax.annotate(f'({punto[0]:.0f}, {punto[1]:.1f})', xy=punto, xytext=(punto[0]+0.3, punto[1]+0.3),
                fontsize=9, color='darkred')

ax.quiver(0, 0, direccion[0], direccion[1], angles='xy', scale_units='xy', scale=1, 
          color='green', width=0.02, label=r'Vector director $(-2, 1)$')

ax.scatter([0], [0], color='black', s=100, zorder=5)
ax.set_xlim(-4, 4)
ax.set_ylim(-3, 3)
ax.set_aspect('equal')
ax.axhline(y=0, color='lightgray', linewidth=0.5)
ax.axvline(x=0, color='lightgray', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title(r'Soluciones del sistema $x + 2y = 0$', fontsize=12, fontweight='bold')
ax.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

Determina si los siguientes conjuntos son subespacios de $\mathbb{R}^2$. Justifica verificando las tres condiciones.

**a)** $S_1 = \{(x, y) : y = 3x\}$

**b)** $S_2 = \{(x, y) : x = 2\}$

**c)** $S_3 = \{(x, y) : x + y = 0\}$

**d)** $S_4 = \{(x, y) : xy = 0\}$

:::{admonition} Soluciones
:class: dropdown tip

**a) $S_1 = \{(x, y) : y = 3x\}$ → SÍ es subespacio**
- Contiene $(0,0)$: $0 = 3(0)$ ✓
- Cerrado bajo suma: Si $y_1 = 3x_1$ y $y_2 = 3x_2$, entonces $y_1 + y_2 = 3(x_1 + x_2)$ ✓
- Cerrado bajo escalares: Si $y = 3x$, entonces $\alpha y = 3(\alpha x)$ ✓

**b) $S_2 = \{(x, y) : x = 2\}$ → NO es subespacio**
- NO contiene $(0,0)$: El punto $(0,0)$ no satisface $x = 2$ ✗

**c) $S_3 = \{(x, y) : x + y = 0\}$ → SÍ es subespacio**
- Contiene $(0,0)$: $0 + 0 = 0$ ✓
- Cerrado bajo suma: Si $x_1 + y_1 = 0$ y $x_2 + y_2 = 0$, entonces $(x_1+x_2) + (y_1+y_2) = 0$ ✓
- Cerrado bajo escalares: Si $x + y = 0$, entonces $\alpha x + \alpha y = \alpha(x+y) = 0$ ✓

**d) $S_4 = \{(x, y) : xy = 0\}$ → NO es subespacio**
- Contiene $(0,0)$: $0 \cdot 0 = 0$ ✓
- NO cerrado bajo suma: $(1, 0)$ y $(0, 1)$ están en $S_4$ (ya que $1 \cdot 0 = 0$), pero $(1,0) + (0,1) = (1,1)$ y $1 \cdot 1 = 1 \neq 0$ ✗
