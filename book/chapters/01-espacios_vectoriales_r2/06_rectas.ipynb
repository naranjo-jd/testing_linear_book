# Rectas en $\mathbb{R}^2$

Las **rectas** son objetos geométricos fundamentales que aparecen constantemente en matemáticas, física e ingeniería. En esta sección estudiaremos las rectas desde la perspectiva del álgebra lineal, entendiendo cómo se relacionan con los vectores y los espacios vectoriales.

Desde el punto de vista del álgebra lineal, una recta en $\mathbb{R}^2$ puede verse como:
- Un **subespacio vectorial** si pasa por el origen
- Un **subespacio afín** (traslación de un subespacio) si no pasa por el origen

Aprenderemos diferentes formas de representar rectas y cómo convertir entre ellas.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Configuración de estilo para las figuras
plt.rcParams['figure.figsize'] = [8, 6]
plt.rcParams['font.size'] = 12
```

## 1. Forma Paramétrica

:::{admonition} Definición: Ecuación Paramétrica de una Recta
:class: tip

Una recta en $\mathbb{R}^2$ puede expresarse en **forma paramétrica** como:

$$
\mathbf{r}(t) = \mathbf{r}_0 + t\mathbf{v}
$$

donde:
- $\mathbf{r}_0 = (x_0, y_0)$ es un **punto** que pertenece a la recta
- $\mathbf{v} = (v_x, v_y)$ es el **vector dirección** de la recta
- $t \in \mathbb{R}$ es el **parámetro**
:::

La forma paramétrica tiene una interpretación geométrica muy clara:
- Partimos del punto $\mathbf{r}_0$
- Nos movemos en la dirección de $\mathbf{v}$
- El parámetro $t$ indica "cuánto" nos movemos

En forma de componentes:
$$
\begin{cases}
x(t) = x_0 + t \cdot v_x \\
y(t) = y_0 + t \cdot v_y
\end{cases}
$$

```python
# Visualización de la forma paramétrica
fig, ax = plt.subplots(figsize=(10, 8))

# Punto inicial y vector dirección
r0 = np.array([1, 1])
v = np.array([2, 1])

# Generar puntos de la recta para diferentes valores de t
t_values = np.linspace(-2, 3, 100)
x_line = r0[0] + t_values * v[0]
y_line = r0[1] + t_values * v[1]

# Dibujar la recta
ax.plot(x_line, y_line, 'b-', linewidth=2, label='Recta $\\mathbf{r}(t) = \\mathbf{r}_0 + t\\mathbf{v}$')

# Marcar el punto r0
ax.plot(r0[0], r0[1], 'ro', markersize=10, label=f'$\\mathbf{{r}}_0 = ({r0[0]}, {r0[1]})$', zorder=5)

# Dibujar el vector dirección desde r0
ax.annotate('', xy=(r0[0] + v[0], r0[1] + v[1]), xytext=(r0[0], r0[1]),
            arrowprops=dict(arrowstyle='->', color='green', lw=2))
ax.text(r0[0] + v[0]/2 + 0.2, r0[1] + v[1]/2 + 0.2, f'$\\mathbf{{v}} = ({v[0]}, {v[1]})$', 
        fontsize=12, color='green')

# Marcar algunos puntos para diferentes valores de t
t_markers = [-1, 0, 1, 2]
for t in t_markers:
    pt = r0 + t * v
    ax.plot(pt[0], pt[1], 'ko', markersize=6)
    ax.text(pt[0] + 0.15, pt[1] + 0.15, f't={t}', fontsize=10)

ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$y$', fontsize=14)
ax.set_xlim(-4, 8)
ax.set_ylim(-2, 5)
ax.set_aspect('equal')
ax.legend(loc='upper left')
ax.set_title('Forma Paramétrica de una Recta', fontsize=14)
plt.tight_layout()
plt.show()
```

## 2.  Exploración: Forma Paramétrica

Utiliza los deslizadores para modificar el **punto inicial** $\mathbf{r}_0 = (x_0, y_0)$ y el **vector dirección** $\mathbf{v} = (v_x, v_y)$. Observa cómo cambia la recta.

```python
def visualizar_recta_parametrica(x0, y0, vx, vy):
    """Visualiza una recta en forma paramétrica."""
    fig, ax = plt.subplots(figsize=(10, 8))
    
    r0 = np.array([x0, y0])
    v = np.array([vx, vy])
    
    # Verificar que el vector dirección no sea nulo
    if np.linalg.norm(v) < 0.01:
        ax.text(0, 0, 'El vector dirección no puede ser nulo', 
                fontsize=14, ha='center', color='red')
        ax.set_xlim(-5, 5)
        ax.set_ylim(-5, 5)
        plt.show()
        return
    
    # Generar puntos de la recta
    t_values = np.linspace(-5, 5, 100)
    x_line = r0[0] + t_values * v[0]
    y_line = r0[1] + t_values * v[1]
    
    # Dibujar la recta
    ax.plot(x_line, y_line, 'b-', linewidth=2, label='Recta')
    
    # Marcar el punto r0
    ax.plot(r0[0], r0[1], 'ro', markersize=12, label=f'$\\mathbf{{r}}_0 = ({x0:.1f}, {y0:.1f})$', zorder=5)
    
    # Dibujar el vector dirección desde r0
    scale = 1.0
    ax.annotate('', xy=(r0[0] + scale*v[0], r0[1] + scale*v[1]), xytext=(r0[0], r0[1]),
                arrowprops=dict(arrowstyle='->', color='green', lw=2.5))
    ax.text(r0[0] + scale*v[0]/2 + 0.3, r0[1] + scale*v[1]/2 + 0.3, 
            f'$\\mathbf{{v}} = ({vx:.1f}, {vy:.1f})$', fontsize=12, color='green')
    
    # Marcar el origen
    ax.plot(0, 0, 'k+', markersize=10, mew=2)
    
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('$x$', fontsize=14)
    ax.set_ylabel('$y$', fontsize=14)
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.set_aspect('equal')
    ax.legend(loc='upper left')
    ax.set_title('Exploración: Forma Paramétrica', fontsize=14)
    plt.tight_layout()
    plt.show()

# Crear sliders
interact(visualizar_recta_parametrica,
         x0=FloatSlider(min=-4, max=4, step=0.5, value=1, description='$x_0$:'),
         y0=FloatSlider(min=-4, max=4, step=0.5, value=1, description='$y_0$:'),
         vx=FloatSlider(min=-3, max=3, step=0.5, value=2, description='$v_x$:'),
         vy=FloatSlider(min=-3, max=3, step=0.5, value=1, description='$v_y$:'));
```

:::{admonition} Preguntas para reflexionar
:class: note

1. **¿Cómo cambia la recta al modificar el punto $\mathbf{r}_0$?**
   - La recta se traslada, manteniendo su dirección (inclinación)

2. **¿Cómo cambia la recta al modificar el vector dirección $\mathbf{v}$?**
   - La inclinación de la recta cambia
   - La recta siempre pasa por $\mathbf{r}_0$

3. **¿Qué pasa si multiplicas $\mathbf{v}$ por una constante negativa?**
   - La recta es la misma (solo cambia el sentido de recorrido)
:::

## 3. Forma Cartesiana

:::{admonition} Definición: Ecuación Cartesiana de una Recta
:class: tip

La **forma cartesiana** (o implícita) de una recta es:

$$
Ax + By + C = 0
$$

donde $A$, $B$ y $C$ son constantes reales con $A$ y $B$ no ambos cero.
:::

### Interpretación del vector normal

El vector $\mathbf{n} = (A, B)$ es **perpendicular** (normal) a la recta. Esto significa que:
- $\mathbf{n}$ es ortogonal al vector dirección de la recta
- Si $\mathbf{v}$ es el vector dirección, entonces $\mathbf{n} \cdot \mathbf{v} = 0$

### Casos especiales

| Condición | Tipo de recta | Ecuación simplificada |
|-----------|---------------|----------------------|
| $B \neq 0$ | Recta no vertical | $y = -\frac{A}{B}x - \frac{C}{B}$ (pendiente $m = -\frac{A}{B}$) |
| $B = 0$ | Recta vertical | $x = -\frac{C}{A}$ |
| $A = 0$ | Recta horizontal | $y = -\frac{C}{B}$ |

```python
# Visualización de la forma cartesiana con vector normal
fig, ax = plt.subplots(figsize=(10, 8))

# Coeficientes de la recta: 2x + 3y - 6 = 0
A, B, C = 2, 3, -6

# Generar puntos de la recta
x_range = np.linspace(-1, 5, 100)
y_line = (-A * x_range - C) / B

# Dibujar la recta
ax.plot(x_range, y_line, 'b-', linewidth=2, label=f'${A}x + {B}y + ({C}) = 0$')

# Vector normal (A, B)
n = np.array([A, B])
# Punto sobre la recta para dibujar el vector normal
x_pt = 1.5
y_pt = (-A * x_pt - C) / B
pt = np.array([x_pt, y_pt])

# Normalizar para visualización
n_normalized = n / np.linalg.norm(n) * 1.5

# Dibujar el vector normal
ax.annotate('', xy=(pt[0] + n_normalized[0], pt[1] + n_normalized[1]), xytext=(pt[0], pt[1]),
            arrowprops=dict(arrowstyle='->', color='red', lw=2.5))
ax.text(pt[0] + n_normalized[0]/2 + 0.2, pt[1] + n_normalized[1]/2 + 0.3, 
        f'$\\mathbf{{n}} = ({A}, {B})$', fontsize=12, color='red')

# Vector dirección (perpendicular al normal)
v = np.array([-B, A])  # (-B, A) es perpendicular a (A, B)
v_normalized = v / np.linalg.norm(v) * 1.5

ax.annotate('', xy=(pt[0] + v_normalized[0], pt[1] + v_normalized[1]), xytext=(pt[0], pt[1]),
            arrowprops=dict(arrowstyle='->', color='green', lw=2.5))
ax.text(pt[0] + v_normalized[0]/2 - 0.5, pt[1] + v_normalized[1]/2 - 0.3, 
        f'$\\mathbf{{v}} = ({-B}, {A})$', fontsize=12, color='green')

# Marcar el punto
ax.plot(pt[0], pt[1], 'ko', markersize=8)

# Dibujar ángulo recto entre n y v
from matplotlib.patches import Rectangle
import matplotlib.transforms as transforms
angle = np.arctan2(v_normalized[1], v_normalized[0])
rect_size = 0.2
rect = Rectangle((pt[0], pt[1]), rect_size, rect_size, 
                  angle=np.degrees(angle), fill=False, edgecolor='black', linewidth=1)
ax.add_patch(rect)

ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$y$', fontsize=14)
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 4)
ax.set_aspect('equal')
ax.legend(loc='upper right')
ax.set_title('Forma Cartesiana: Vector Normal y Vector Dirección', fontsize=14)
plt.tight_layout()
plt.show()
```

## 4. Conversión entre Formas

### 4.1 De Paramétrica a Cartesiana

Dada la forma paramétrica:
$$
\begin{cases}
x = x_0 + t \cdot v_x \\
y = y_0 + t \cdot v_y
\end{cases}
$$

**Método:** Eliminamos el parámetro $t$.

1. De la primera ecuación: $t = \frac{x - x_0}{v_x}$ (si $v_x \neq 0$)
2. Sustituimos en la segunda: $y = y_0 + \frac{v_y}{v_x}(x - x_0)$
3. Reorganizamos: $v_y(x - x_0) - v_x(y - y_0) = 0$
4. Expandimos: $v_y \cdot x - v_x \cdot y + (v_x \cdot y_0 - v_y \cdot x_0) = 0$

:::{admonition} Fórmula de Conversión
:class: important

Si $\mathbf{r}_0 = (x_0, y_0)$ y $\mathbf{v} = (v_x, v_y)$, la forma cartesiana es:

$$
v_y \cdot x - v_x \cdot y + (v_x \cdot y_0 - v_y \cdot x_0) = 0
$$

Es decir: $A = v_y$, $B = -v_x$, $C = v_x \cdot y_0 - v_y \cdot x_0$
:::

:::{admonition} Ejemplo: Paramétrica → Cartesiana
:class: example

Convertir la recta $\mathbf{r}(t) = (2, -1) + t(3, 4)$ a forma cartesiana.

**Solución:**
- $\mathbf{r}_0 = (2, -1)$, $\mathbf{v} = (3, 4)$
- $A = v_y = 4$
- $B = -v_x = -3$
- $C = v_x \cdot y_0 - v_y \cdot x_0 = 3(-1) - 4(2) = -3 - 8 = -11$

**Ecuación cartesiana:** $4x - 3y - 11 = 0$
:::

```python
# Verificación computacional
def parametrica_a_cartesiana(r0, v):
    """Convierte de forma paramétrica a cartesiana."""
    x0, y0 = r0
    vx, vy = v
    A = vy
    B = -vx
    C = vx * y0 - vy * x0
    return A, B, C

# Ejemplo
r0 = (2, -1)
v = (3, 4)
A, B, C = parametrica_a_cartesiana(r0, v)
print(f"Forma paramétrica: r(t) = {r0} + t·{v}")
print(f"Forma cartesiana: {A}x + ({B})y + ({C}) = 0")
print(f"Simplificado: {A}x - {-B}y - {-C} = 0")
```

### 4.2 De Cartesiana a Paramétrica

Dada la forma cartesiana $Ax + By + C = 0$:

**Método:**
1. Encontrar un **punto** sobre la recta (dar un valor a $x$ o $y$ y resolver)
2. El **vector dirección** es perpendicular al normal, por lo que $\mathbf{v} = (-B, A)$ o $(B, -A)$

:::{admonition} Fórmula de Conversión
:class: important

Para la recta $Ax + By + C = 0$:
- **Punto:** Si $B \neq 0$, usar $(0, -C/B)$. Si $B = 0$, usar $(-C/A, 0)$
- **Vector dirección:** $\mathbf{v} = (-B, A)$
:::

:::{admonition} Ejemplo: Cartesiana → Paramétrica
:class: example

Convertir la recta $3x - 2y + 6 = 0$ a forma paramétrica.

**Solución:**
- $A = 3$, $B = -2$, $C = 6$
- **Punto:** tomamos $x = 0$: $-2y + 6 = 0 \Rightarrow y = 3$. Entonces $\mathbf{r}_0 = (0, 3)$
- **Vector dirección:** $\mathbf{v} = (-B, A) = (2, 3)$

**Ecuación paramétrica:** $\mathbf{r}(t) = (0, 3) + t(2, 3)$
:::

```python
# Verificación computacional
def cartesiana_a_parametrica(A, B, C):
    """Convierte de forma cartesiana a paramétrica."""
    # Encontrar un punto sobre la recta
    if B != 0:
        r0 = (0, -C / B)
    else:
        r0 = (-C / A, 0)
    # Vector dirección (perpendicular al normal)
    v = (-B, A)
    return r0, v

# Ejemplo
A, B, C = 3, -2, 6
r0, v = cartesiana_a_parametrica(A, B, C)
print(f"Forma cartesiana: {A}x + ({B})y + {C} = 0")
print(f"Forma paramétrica: r(t) = {r0} + t·{v}")
```

## 5. Otras Representaciones

### 5.1 Forma Punto-Pendiente

Si conocemos un punto $(x_0, y_0)$ y la **pendiente** $m$:

$$
y - y_0 = m(x - x_0)
$$

La pendiente se relaciona con el vector dirección: si $\mathbf{v} = (v_x, v_y)$, entonces $m = \frac{v_y}{v_x}$ (siempre que $v_x \neq 0$).

### 5.2 Forma de Dos Puntos

Si conocemos dos puntos $P_1 = (x_1, y_1)$ y $P_2 = (x_2, y_2)$:

$$
\frac{y - y_1}{y_2 - y_1} = \frac{x - x_1}{x_2 - x_1}
$$

Esto es equivalente a la forma paramétrica con:
- $\mathbf{r}_0 = (x_1, y_1)$
- $\mathbf{v} = (x_2 - x_1, y_2 - y_1) = P_2 - P_1$

### 5.3 ¿Cuándo usar cada forma?

| Forma | Útil cuando... |
|-------|---------------|
| Paramétrica | Se necesita generar puntos, estudiar movimiento, parametrizar curvas |
| Cartesiana | Se verifica si puntos pertenecen a la recta, se trabaja con desigualdades |
| Punto-pendiente | Se conoce pendiente y un punto, útil en cálculo |
| Dos puntos | Se conocen dos puntos específicos |

## 6. Distancia de Punto a Recta

:::{admonition} Fórmula de Distancia Punto-Recta
:class: tip

La distancia de un punto $P_0 = (x_0, y_0)$ a una recta $Ax + By + C = 0$ es:

$$
d = \frac{|Ax_0 + By_0 + C|}{\sqrt{A^2 + B^2}}
$$
:::

### Derivación usando proyección

La fórmula se deriva usando proyección ortogonal:

1. Sea $Q$ un punto cualquiera sobre la recta
2. El vector $\overrightarrow{QP_0}$ conecta la recta con $P_0$
3. La distancia es la longitud de la proyección de $\overrightarrow{QP_0}$ sobre el vector normal $\mathbf{n} = (A, B)$

$$
d = |\text{proy}_{\mathbf{n}} \overrightarrow{QP_0}| = \frac{|\overrightarrow{QP_0} \cdot \mathbf{n}|}{\|\mathbf{n}\|}
$$

```python
# Visualización de distancia punto a recta
fig, ax = plt.subplots(figsize=(10, 8))

# Recta: 3x + 4y - 12 = 0
A, B, C = 3, 4, -12

# Punto externo
P0 = np.array([5, 4])

# Dibujar la recta
x_range = np.linspace(-1, 6, 100)
y_line = (-A * x_range - C) / B
ax.plot(x_range, y_line, 'b-', linewidth=2, label=f'${A}x + {B}y - 12 = 0$')

# Calcular distancia
d = abs(A * P0[0] + B * P0[1] + C) / np.sqrt(A**2 + B**2)

# Punto más cercano sobre la recta (pie de la perpendicular)
t = -(A * P0[0] + B * P0[1] + C) / (A**2 + B**2)
Q = P0 + t * np.array([A, B])

# Dibujar el punto externo
ax.plot(P0[0], P0[1], 'ro', markersize=10, label=f'$P_0 = ({P0[0]}, {P0[1]})$', zorder=5)

# Dibujar la distancia (línea punteada)
ax.plot([P0[0], Q[0]], [P0[1], Q[1]], 'r--', linewidth=2, label=f'$d = {d:.2f}$')

# Dibujar el pie de la perpendicular
ax.plot(Q[0], Q[1], 'go', markersize=8, zorder=5)

# Vector normal
n = np.array([A, B])
n_norm = n / np.linalg.norm(n) * 1.2
ax.annotate('', xy=(Q[0] + n_norm[0], Q[1] + n_norm[1]), xytext=(Q[0], Q[1]),
            arrowprops=dict(arrowstyle='->', color='purple', lw=2))
ax.text(Q[0] + n_norm[0] + 0.1, Q[1] + n_norm[1] + 0.1, '$\\mathbf{n}$', fontsize=12, color='purple')

# Marcar ángulo recto
angle = np.arctan2(-B, A)
rect_size = 0.2
rect = Rectangle((Q[0], Q[1]), rect_size, rect_size, 
                  angle=np.degrees(angle), fill=False, edgecolor='black', linewidth=1)
ax.add_patch(rect)

ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$y$', fontsize=14)
ax.set_xlim(-1, 7)
ax.set_ylim(-1, 5)
ax.set_aspect('equal')
ax.legend(loc='upper right')
ax.set_title('Distancia de un Punto a una Recta', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
def distancia_punto_recta(P0, A, B, C):
    """Calcula la distancia de un punto a una recta Ax + By + C = 0."""
    x0, y0 = P0
    numerador = abs(A * x0 + B * y0 + C)
    denominador = np.sqrt(A**2 + B**2)
    return numerador / denominador

# Ejemplo: distancia de P0 = (5, 4) a la recta 3x + 4y - 12 = 0
P0 = (5, 4)
A, B, C = 3, 4, -12

d = distancia_punto_recta(P0, A, B, C)
print(f"Punto: P₀ = {P0}")
print(f"Recta: {A}x + {B}y + ({C}) = 0")
print(f"\nCálculo:")
print(f"d = |{A}·{P0[0]} + {B}·{P0[1]} + ({C})| / √({A}² + {B}²)")
print(f"d = |{A*P0[0]} + {B*P0[1]} + ({C})| / √({A**2} + {B**2})")
print(f"d = |{A*P0[0] + B*P0[1] + C}| / √{A**2 + B**2}")
print(f"d = {abs(A*P0[0] + B*P0[1] + C)} / {np.sqrt(A**2 + B**2)}")
print(f"d = {d}")
```

## 7. Rectas Paralelas y Perpendiculares

### 7.1 Rectas Paralelas

:::{admonition} Condición de Paralelismo
:class: tip

Dos rectas son **paralelas** si y solo si sus vectores dirección son **proporcionales** (o equivalentemente, sus vectores normales son proporcionales).

Para rectas $A_1x + B_1y + C_1 = 0$ y $A_2x + B_2y + C_2 = 0$:

$$
\text{Paralelas} \iff \frac{A_1}{A_2} = \frac{B_1}{B_2} \quad \text{(o equivalentemente } A_1 B_2 - A_2 B_1 = 0\text{)}
$$
:::

### 7.2 Rectas Perpendiculares

:::{admonition} Condición de Perpendicularidad
:class: tip

Dos rectas son **perpendiculares** si y solo si sus vectores dirección son **ortogonales** (producto punto igual a cero).

Para rectas $A_1x + B_1y + C_1 = 0$ y $A_2x + B_2y + C_2 = 0$:

$$
\text{Perpendiculares} \iff A_1 A_2 + B_1 B_2 = 0
$$

En términos de pendientes: $m_1 \cdot m_2 = -1$
:::

```python
# Visualización de rectas paralelas y perpendiculares
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --- Panel izquierdo: Rectas paralelas ---
ax1 = axes[0]

# Rectas: 2x + 3y - 6 = 0 y 2x + 3y - 12 = 0 (paralelas)
x_range = np.linspace(-1, 7, 100)
y1 = (-2 * x_range + 6) / 3
y2 = (-2 * x_range + 12) / 3

ax1.plot(x_range, y1, 'b-', linewidth=2, label='$2x + 3y - 6 = 0$')
ax1.plot(x_range, y2, 'r-', linewidth=2, label='$2x + 3y - 12 = 0$')

# Vectores dirección (iguales)
pt1 = np.array([1, (6-2)/3])
pt2 = np.array([1, (12-2)/3])
v = np.array([-3, 2])  # Vector dirección
v_norm = v / np.linalg.norm(v) * 1.5

ax1.annotate('', xy=(pt1[0] + v_norm[0], pt1[1] + v_norm[1]), xytext=(pt1[0], pt1[1]),
            arrowprops=dict(arrowstyle='->', color='blue', lw=2))
ax1.annotate('', xy=(pt2[0] + v_norm[0], pt2[1] + v_norm[1]), xytext=(pt2[0], pt2[1]),
            arrowprops=dict(arrowstyle='->', color='red', lw=2))
ax1.text(pt1[0] + v_norm[0] + 0.1, pt1[1] + v_norm[1], '$\\mathbf{v}$', fontsize=12, color='blue')
ax1.text(pt2[0] + v_norm[0] + 0.1, pt2[1] + v_norm[1], '$\\mathbf{v}$', fontsize=12, color='red')

ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('$x$', fontsize=14)
ax1.set_ylabel('$y$', fontsize=14)
ax1.set_xlim(-1, 7)
ax1.set_ylim(-1, 5)
ax1.set_aspect('equal')
ax1.legend(loc='upper right')
ax1.set_title('Rectas Paralelas\n(mismo vector dirección)', fontsize=14)

# --- Panel derecho: Rectas perpendiculares ---
ax2 = axes[1]

# Rectas: 2x + 3y - 6 = 0 y 3x - 2y - 4 = 0 (perpendiculares)
y1 = (-2 * x_range + 6) / 3
y3 = (3 * x_range - 4) / 2

ax2.plot(x_range, y1, 'b-', linewidth=2, label='$2x + 3y - 6 = 0$')
ax2.plot(x_range, y3, 'g-', linewidth=2, label='$3x - 2y - 4 = 0$')

# Punto de intersección
# Resolver: 2x + 3y = 6 y 3x - 2y = 4
A_sys = np.array([[2, 3], [3, -2]])
b_sys = np.array([6, 4])
interseccion = np.linalg.solve(A_sys, b_sys)
ax2.plot(interseccion[0], interseccion[1], 'ko', markersize=8, zorder=5)

# Vectores dirección
v1 = np.array([-3, 2])
v2 = np.array([2, 3])
v1_norm = v1 / np.linalg.norm(v1) * 1.2
v2_norm = v2 / np.linalg.norm(v2) * 1.2

ax2.annotate('', xy=(interseccion[0] + v1_norm[0], interseccion[1] + v1_norm[1]), 
            xytext=(interseccion[0], interseccion[1]),
            arrowprops=dict(arrowstyle='->', color='blue', lw=2))
ax2.annotate('', xy=(interseccion[0] + v2_norm[0], interseccion[1] + v2_norm[1]), 
            xytext=(interseccion[0], interseccion[1]),
            arrowprops=dict(arrowstyle='->', color='green', lw=2))

# Marcar ángulo recto
angle = np.arctan2(v1_norm[1], v1_norm[0])
rect_size = 0.25
rect = Rectangle((interseccion[0], interseccion[1]), rect_size, rect_size, 
                  angle=np.degrees(angle), fill=False, edgecolor='black', linewidth=1.5)
ax2.add_patch(rect)

ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.axvline(x=0, color='k', linewidth=0.5)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('$x$', fontsize=14)
ax2.set_ylabel('$y$', fontsize=14)
ax2.set_xlim(-1, 5)
ax2.set_ylim(-2, 4)
ax2.set_aspect('equal')
ax2.legend(loc='upper right')
ax2.set_title('Rectas Perpendiculares\n(vectores dirección ortogonales)', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
def verificar_relacion(A1, B1, C1, A2, B2, C2):
    """Verifica si dos rectas son paralelas, perpendiculares o secantes."""
    print(f"Recta 1: {A1}x + {B1}y + {C1} = 0")
    print(f"Recta 2: {A2}x + {B2}y + {C2} = 0")
    print()
    
    # Vectores dirección
    v1 = np.array([-B1, A1])
    v2 = np.array([-B2, A2])
    print(f"Vector dirección 1: v₁ = {tuple(v1)}")
    print(f"Vector dirección 2: v₂ = {tuple(v2)}")
    
    # Producto cruz (determinante 2D) para paralelismo
    cruz = A1 * B2 - A2 * B1
    # Producto punto para perpendicularidad
    punto = A1 * A2 + B1 * B2
    
    print(f"\nProducto cruz (A₁B₂ - A₂B₁): {cruz}")
    print(f"Producto punto (A₁A₂ + B₁B₂): {punto}")
    
    if abs(cruz) < 1e-10:
        print("\n→ Las rectas son PARALELAS")
    elif abs(punto) < 1e-10:
        print("\n→ Las rectas son PERPENDICULARES")
    else:
        print("\n→ Las rectas son SECANTES (ni paralelas ni perpendiculares)")

# Ejemplo 1: Paralelas
print("=" * 50)
print("EJEMPLO 1: Verificación de paralelismo")
print("=" * 50)
verificar_relacion(2, 3, -6, 2, 3, -12)

print()
print("=" * 50)
print("EJEMPLO 2: Verificación de perpendicularidad")
print("=" * 50)
verificar_relacion(2, 3, -6, 3, -2, -4)
```

## 8. Rectas como Subespacios Afines

Desde la perspectiva del álgebra lineal, las rectas en $\mathbb{R}^2$ tienen una estructura especial:

### Rectas que pasan por el origen: Subespacios vectoriales

Una recta que pasa por el origen $(0, 0)$ es un **subespacio vectorial** de $\mathbb{R}^2$. Esto significa que:
- Contiene al vector cero
- Es cerrada bajo suma de vectores
- Es cerrada bajo multiplicación por escalares

Estas rectas tienen ecuación $Ax + By = 0$ (sin término constante).

**Ejemplo:** La recta $y = 2x$ (o equivalentemente $2x - y = 0$) es un subespacio.

### Rectas que NO pasan por el origen: Subespacios afines

Una recta que **no** pasa por el origen es un **subespacio afín** (o variedad lineal). Puede verse como la **traslación** de un subespacio:

$$
L = \mathbf{r}_0 + W
$$

donde:
- $\mathbf{r}_0$ es un punto de la recta
- $W = \text{span}\{\mathbf{v}\}$ es el subespacio generado por el vector dirección

:::{admonition} Nota importante
:class: warning

Un subespacio afín que no pasa por el origen **no es un subespacio vectorial**, ya que no contiene al vector cero y no es cerrado bajo las operaciones.
:::

```python
# Visualización de subespacios y subespacios afines
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --- Panel izquierdo: Subespacio (pasa por origen) ---
ax1 = axes[0]

# Recta por el origen: y = 2x
x_range = np.linspace(-3, 3, 100)
y_sub = 2 * x_range

ax1.plot(x_range, y_sub, 'b-', linewidth=3, label='$W = \\{(x, 2x) : x \\in \\mathbb{R}\\}$')

# Marcar el origen
ax1.plot(0, 0, 'ro', markersize=12, label='Origen $(0,0)$', zorder=5)

# Vector generador
v = np.array([1, 2])
ax1.annotate('', xy=(v[0], v[1]), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', color='green', lw=2.5))
ax1.text(v[0] + 0.15, v[1] + 0.15, '$\\mathbf{v} = (1, 2)$', fontsize=12, color='green')

# Mostrar algunos múltiplos
for k in [-2, -1, 2]:
    pt = k * v
    ax1.plot(pt[0], pt[1], 'bo', markersize=6)
    ax1.text(pt[0] + 0.1, pt[1] + 0.1, f'${k}\\mathbf{{v}}$', fontsize=10)

ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('$x$', fontsize=14)
ax1.set_ylabel('$y$', fontsize=14)
ax1.set_xlim(-3, 3)
ax1.set_ylim(-3, 5)
ax1.set_aspect('equal')
ax1.legend(loc='upper left')
ax1.set_title('Subespacio Vectorial\n(recta por el origen)', fontsize=14)

# --- Panel derecho: Subespacio afín (no pasa por origen) ---
ax2 = axes[1]

# Subespacio: y = 2x
ax2.plot(x_range, y_sub, 'b--', linewidth=1.5, alpha=0.5, label='$W$ (subespacio)')

# Subespacio afín: traslación por r0 = (1, 1)
r0 = np.array([1, 1])
y_afin = y_sub + (r0[1] - 2*r0[0])  # y = 2x + (1 - 2) = 2x - 1
ax2.plot(x_range, y_afin, 'r-', linewidth=3, label='$L = \\mathbf{r}_0 + W$')

# Marcar r0
ax2.plot(r0[0], r0[1], 'go', markersize=12, label=f'$\\mathbf{{r}}_0 = ({r0[0]}, {r0[1]})$', zorder=5)

# Flecha de traslación
ax2.annotate('', xy=(r0[0], r0[1]), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', color='purple', lw=2, 
                           connectionstyle='arc3,rad=0.2'))
ax2.text(0.3, 0.8, 'traslación', fontsize=10, color='purple', rotation=30)

# Vector dirección en el subespacio afín
ax2.annotate('', xy=(r0[0] + v[0], r0[1] + v[1]), xytext=(r0[0], r0[1]),
            arrowprops=dict(arrowstyle='->', color='green', lw=2.5))
ax2.text(r0[0] + v[0] + 0.1, r0[1] + v[1] + 0.1, '$\\mathbf{v}$', fontsize=12, color='green')

# Marcar el origen
ax2.plot(0, 0, 'ko', markersize=8)
ax2.text(0.1, -0.3, 'Origen', fontsize=10)

ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.axvline(x=0, color='k', linewidth=0.5)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('$x$', fontsize=14)
ax2.set_ylabel('$y$', fontsize=14)
ax2.set_xlim(-3, 3)
ax2.set_ylim(-3, 5)
ax2.set_aspect('equal')
ax2.legend(loc='upper left')
ax2.set_title('Subespacio Afín\n(traslación de un subespacio)', fontsize=14)

plt.tight_layout()
plt.show()
```

:::{admonition} Solución

**1. Ecuación paramétrica:**
- $\mathbf{r}_0 = P_1 = (2, 3)$
- $\mathbf{v} = P_2 - P_1 = (5-2, 7-3) = (3, 4)$
- $\mathbf{r}(t) = (2, 3) + t(3, 4)$
