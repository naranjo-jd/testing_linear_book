## ❓ Pregunta 1: Verificación de linealidad

Sea $T: \mathbb{R}^2 \to \mathbb{R}^2$ definida por $T(x, y) = (2x + y,\; x - 3y)$.

**(a)** ¿Es $T$ una transformación lineal? Justifica tu respuesta.

**(b)** ¿Cuál es la matriz $A$ tal que $T(\mathbf{x}) = A\mathbf{x}$?

**(c)** Calcula $T(3, -1)$.

### ✅ Respuesta 1

**(a) Sí, $T$ es lineal.** Verificación:

Sean $(x_1,y_1), (x_2,y_2) \in \mathbb{R}^2$ y $c \in \mathbb{R}$.

**Aditividad:**
$$T\bigl((x_1+x_2,\,y_1+y_2)\bigr) = (2(x_1+x_2)+(y_1+y_2),\; (x_1+x_2)-3(y_1+y_2))$$
$$= (2x_1+y_1,\,x_1-3y_1) + (2x_2+y_2,\,x_2-3y_2) = T(x_1,y_1) + T(x_2,y_2). \checkmark$$

**Homogeneidad:**
$$T(cx, cy) = (2cx+cy,\; cx-3cy) = c(2x+y,\; x-3y) = cT(x,y). \checkmark$$

**(b)** Las columnas de $A$ son $T(1,0) = (2,1)$ y $T(0,1) = (1,-3)$:
$$A = \begin{pmatrix} 2 & 1 \\ 1 & -3 \end{pmatrix}$$

**(c)** $T(3,-1) = A\begin{pmatrix}3\\-1\end{pmatrix} = \begin{pmatrix}2\cdot3+1\cdot(-1)\\ 1\cdot3+(-3)\cdot(-1)\end{pmatrix} = \begin{pmatrix}5\\ 6\end{pmatrix}$

```python
import numpy as np

# Verificacion computacional de la Pregunta 1
A = np.array([[2, 1], [1, -3]], dtype=float)
x = np.array([3.0, -1.0])

print("Matriz A:")
print(A)
print(f"\nT(3,-1) = A @ x = {A @ x}")

# Verificar aditividad
u = np.array([1.0, 2.0])
v = np.array([-1.0, 3.0])
print(f"\nAditividad: T(u+v) = {A @ (u+v)}, T(u)+T(v) = {A@u + A@v}")
print(f"Se cumple: {np.allclose(A@(u+v), A@u + A@v)}")
```

## ❓ Pregunta 2: Cálculo del núcleo

Sea $T: \mathbb{R}^3 \to \mathbb{R}^2$ con matriz:
$$A = \begin{pmatrix} 1 & 2 & 3 \\ 2 & 4 & 6 \end{pmatrix}$$

**(a)** Calcula una base para $\ker(T)$.

**(b)** ¿Cuál es la dimensión del núcleo?

**(c)** ¿Es $T$ inyectiva? ¿Por qué?

### ✅ Respuesta 2

**(a)** Observamos que la fila 2 es el doble de la fila 1, por lo que la segunda ecuación es redundante. Resolvemos $x_1 + 2x_2 + 3x_3 = 0$:
$$x_1 = -2x_2 - 3x_3$$

Con $x_2 = s$ y $x_3 = t$ libres:
$$\mathbf{x} = s\begin{pmatrix}-2\\1\\0\end{pmatrix} + t\begin{pmatrix}-3\\0\\1\end{pmatrix}$$

Base del núcleo: $\left\{\begin{pmatrix}-2\\1\\0\end{pmatrix},\; \begin{pmatrix}-3\\0\\1\end{pmatrix}\right\}$

**(b)** $\dim(\ker T) = 2$ (nulidad $= 2$)

**(c)** **No es inyectiva**, porque $\ker(T) \neq \{\mathbf{0}\}$. Cualquier transformación con nulidad $> 0$ no puede ser inyectiva.

```python
import numpy as np
import sympy as sp

A = sp.Matrix([[1, 2, 3],
               [2, 4, 6]])
print("Rango de A:", A.rank())
print("Nulidad (dim nucleo):", A.shape[1] - A.rank())
print("\nBase del nucleo:")
for v in A.nullspace():
    sp.pprint(v.T)

# Verificar que los vectores del nucleo satisfacen Ax=0
A_np = np.array([[1,2,3],[2,4,6]], dtype=float)
v1 = np.array([-2.0, 1, 0])
v2 = np.array([-3.0, 0, 1])
print(f"\nA @ v1 = {A_np @ v1} (debe ser [0,0])")
print(f"A @ v2 = {A_np @ v2} (debe ser [0,0])")
```

## ❓ Pregunta 3: Determinación de la imagen

Para la misma transformación de la Pregunta 2, con $A = \begin{pmatrix}1 & 2 & 3 \\ 2 & 4 & 6\end{pmatrix}$:

**(a)** ¿Cuál es la dimensión de $\operatorname{Im}(T)$?

**(b)** Da una base explícita para $\operatorname{Im}(T)$.

**(c)** ¿Es $T$ suprayectiva (¿la imagen es todo $\mathbb{R}^2$?).

### ✅ Respuesta 3

**(a)** Por el teorema rango-nulidad: $\dim(\operatorname{Im} T) = \operatorname{rango}(A)$.

Dado que la fila 2 es múltiplo de la fila 1, el rango de $A$ es $1$. Entonces $\dim(\operatorname{Im} T) = 1$.

**(b)** La imagen es el espacio columna de $A$. La primera columna $\begin{pmatrix}1\\2\end{pmatrix}$ es una base (las otras son múltiplos de ella).

$$\operatorname{Im}(T) = \operatorname{span}\left\{\begin{pmatrix}1\\2\end{pmatrix}\right\}$$

Geométricamente, la imagen es la **recta** $\{(s, 2s) : s \in \mathbb{R}\}$ en $\mathbb{R}^2$.

**(c)** **No es suprayectiva**, porque $\dim(\operatorname{Im} T) = 1 < 2 = \dim(\mathbb{R}^2)$. Solo alcanza una recta en $\mathbb{R}^2$, no todo el plano.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

A = sp.Matrix([[1, 2, 3], [2, 4, 6]])
print("Espacio columna (imagen):")
for v in A.columnspace():
    sp.pprint(v.T)

# Visualizar la imagen
A_np = np.array([[1,2,3],[2,4,6]], dtype=float)
np.random.seed(0)
X = np.random.randn(3, 300)
Y = A_np @ X

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(Y[0], Y[1], s=5, alpha=0.5, color='tomato')
t = np.linspace(-5, 5, 100)
ax.plot(t, 2*t, 'k-', lw=2, label='Imagen: recta $y=2x$')
ax.set_aspect('equal')
ax.set_xlim(-5, 5); ax.set_ylim(-10, 10)
ax.axhline(0, color='gray', lw=0.5)
ax.axvline(0, color='gray', lw=0.5)
ax.legend(); ax.grid(True, alpha=0.3)
ax.set_title('Imagen de $T$ — solo la recta $y=2x$', fontsize=11)
plt.tight_layout()
plt.show()
```

## ❓ Pregunta 4: Teorema rango-nulidad

Sea $T: \mathbb{R}^4 \to \mathbb{R}^3$ una transformación lineal cuya matriz tiene rango $2$.

**(a)** ¿Cuánto vale $\dim(\ker T)$?

**(b)** ¿Puede $T$ ser inyectiva? ¿Puede $T$ ser suprayectiva?

**(c)** Si cambiamos a $T': \mathbb{R}^3 \to \mathbb{R}^3$ con rango $3$, ¿qué ocurre?

### ✅ Respuesta 4

**(a)** Por el teorema rango-nulidad con $n = \dim(\mathbb{R}^4) = 4$:
$$\dim(\ker T) + \operatorname{rango}(T) = 4 \implies \dim(\ker T) = 4 - 2 = 2$$

**(b)**
- **¿Inyectiva?** No. $\ker(T)$ tiene dimensión $2 \neq 0$, por lo que hay vectores no nulos que $T$ lleva al origen.
- **¿Suprayectiva?** No. $\operatorname{rango}(T) = 2 < 3 = \dim(\mathbb{R}^3)$, así que la imagen no cubre todo $\mathbb{R}^3$.

**(c)** Con $T': \mathbb{R}^3 \to \mathbb{R}^3$ y rango $= 3$:
- Nulidad $= 3 - 3 = 0$ → $\ker(T') = \{\mathbf{0}\}$ → **inyectiva** ✅
- Rango $= 3 = \dim(\mathbb{R}^3)$ → **suprayectiva** ✅
- Por tanto $T'$ es **biyectiva e invertible** ✅

## ❓ Pregunta 5: Composición de transformaciones

Sean $T_1: \mathbb{R}^2 \to \mathbb{R}^2$ la rotación de $90°$ y $T_2: \mathbb{R}^2 \to \mathbb{R}^2$ la reflexión sobre el eje $x$.

Sus matrices son:
$$R_{90} = \begin{pmatrix}0 & -1 \\ 1 & 0\end{pmatrix}, \qquad F_x = \begin{pmatrix}1 & 0 \\ 0 & -1\end{pmatrix}$$

**(a)** Calcula la matriz de $T_2 \circ T_1$ (primero rotar, luego reflejar).

**(b)** Calcula la matriz de $T_1 \circ T_2$ (primero reflejar, luego rotar).

**(c)** ¿Son iguales? ¿Qué nos dice esto sobre la composición de transformaciones lineales?

### ✅ Respuesta 5

**(a)** $T_2 \circ T_1$ significa aplicar primero $T_1$ (rotar) y luego $T_2$ (reflejar). La matriz es $F_x \cdot R_{90}$:

$$F_x R_{90} = \begin{pmatrix}1 & 0\\ 0 & -1\end{pmatrix}\begin{pmatrix}0 & -1\\ 1 & 0\end{pmatrix} = \begin{pmatrix}0 & -1\\ -1 & 0\end{pmatrix}$$

**(b)** $T_1 \circ T_2$ significa aplicar primero $T_2$ (reflejar) y luego $T_1$ (rotar). La matriz es $R_{90} \cdot F_x$:

$$R_{90} F_x = \begin{pmatrix}0 & -1\\ 1 & 0\end{pmatrix}\begin{pmatrix}1 & 0\\ 0 & -1\end{pmatrix} = \begin{pmatrix}0 & 1\\ 1 & 0\end{pmatrix}$$

**(c)** **No son iguales.** La composición de transformaciones lineales **no es conmutativa** en general. El orden importa: rotar-y-reflejar es diferente a reflejar-y-rotar.

Geométricamente: la primera composición produce una reflexión sobre la recta $y = -x$, mientras que la segunda produce una reflexión sobre $y = x$.

```python
import numpy as np
import matplotlib.pyplot as plt

R90 = np.array([[0, -1], [1, 0]], dtype=float)  # Rotacion 90 grados
Fx  = np.array([[1, 0], [0, -1]], dtype=float)  # Reflexion sobre eje x

T2_T1 = Fx @ R90   # Primero rotar, luego reflejar
T1_T2 = R90 @ Fx   # Primero reflejar, luego rotar

print("F_x @ R_90 (rotar luego reflejar):")
print(T2_T1)
print("\nR_90 @ F_x (reflejar luego rotar):")
print(T1_T2)
print(f"\n¿Son iguales? {np.allclose(T2_T1, T1_T2)}")

# Visualizar efecto de cada composicion sobre el cuadrado unitario
cuadrado = np.array([[0,1,1,0,0],[0,0,1,1,0]], dtype=float)

fig, axes = plt.subplots(1, 3, figsize=(13, 4))
matrices = [np.eye(2), T2_T1, T1_T2]
titulos  = ['Original', '$F_x \\circ R_{90}$\n(rotar→reflejar)', '$R_{90} \\circ F_x$\n(reflejar→rotar)']
colores  = ['steelblue', 'tomato', 'green']

for ax, M, titulo, col in zip(axes, matrices, titulos, colores):
    c = M @ cuadrado
    ax.fill(c[0], c[1], alpha=0.4, color=col)
    ax.plot(*np.hstack([cuadrado, cuadrado[:,:1]]), 'b--', lw=1.2, alpha=0.4)
    ax.plot(c[0], c[1], color=col, lw=2)
    ax.set_xlim(-1.5, 1.5); ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axhline(0, color='gray', lw=0.5)
    ax.axvline(0, color='gray', lw=0.5)
    ax.set_title(titulo, fontsize=11)
    ax.grid(True, alpha=0.3)

plt.suptitle('La composición de transformaciones NO es conmutativa', fontsize=12)
plt.tight_layout()
plt.show()
```
