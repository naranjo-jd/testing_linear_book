```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import random

# Base de datos de 10 preguntas con distractores "creíbles"
preguntas_db = [
    {
        "id": 1,
        "tema": "Subespacios en R^2",
        "enunciado": "El conjunto $W = \{(x, y) \in \mathbb{R}^2 \mid x - 2y = 0\}$ es un subespacio vectorial de $\mathbb{R}^2$.",
        "real_truth": "Verdadero",
        "just_correcta": "Es una línea recta que pasa por el origen, cumpliendo con la existencia del vector nulo y la cerradura bajo suma y producto escalar.",
        "distractores_verdad": [
            "Es un subespacio porque el determinante de su matriz de coeficientes $[1, -2]$ es distinto de cero.",
            "Todo conjunto definido por una ecuación lineal homogénea en $\mathbb{R}^2$ hereda la topología del espacio completo.",
            "Cumple con la cerradura porque la pendiente de la recta es un número racional positivo."
        ],
        "distractores_falso": [
            "No es subespacio porque la variable $y$ tiene un coeficiente negativo, rompiendo la cerradura escalar.",
            "Falla la cerradura bajo la suma ya que al sumar dos vectores de la recta se duplica la constante de la ecuación.",
            "No contiene al vector nulo porque $x$ e $y$ deben ser estrictamente mayores a cero para definir una dimensión.",
            "La ecuación representa un hiperplano y no un subespacio dentro de la geometría euclidiana plana."
        ]
    },
    {
        "id": 2,
        "tema": "Independencia Lineal",
        "enunciado": "Los vectores $v_1 = (1, 1)$ y $v_2 = (2, 2)$ son linealmente independientes en $\mathbb{R}^2$.",
        "real_truth": "Falso",
        "just_correcta": "Son linealmente dependientes porque $v_2$ es un múltiplo escalar de $v_1$ ($v_2 = 2v_1$), lo que indica que están en la misma dirección.",
        "distractores_falso": [
            "Son dependientes porque la suma de sus componentes ($1+1$ y $2+2$) resulta en números pares consecutivos.",
            "La independencia lineal en $\mathbb{R}^2$ solo ocurre si al menos uno de los vectores es unitario.",
            "Al ser vectores en el primer cuadrante, comparten el mismo signo algebraico, lo que anula su independencia."
        ],
        "distractores_verdad": [
            "Son independientes porque apuntan hacia el infinito positivo, cubriendo toda la extensión del primer cuadrante.",
            "Cualquier par de vectores con componentes distintas de cero genera una base en $\mathbb{R}^2$.",
            "La matriz formada por estos vectores tiene un rango completo igual a 2.",
            "Forman un conjunto independiente porque no son ortogonales entre sí."
        ]
    },
    {
        "id": 3,
        "tema": "Espacio Generado",
        "enunciado": "El conjunto $S = \{(1, 0), (0, 1), (1, 1)\}$ genera a todo $\mathbb{R}^2$.",
        "real_truth": "Verdadero",
        "just_correcta": "Contiene a los vectores de la base canónica, los cuales son suficientes para expresar cualquier vector $(x, y)$ como combinación lineal.",
        "distractores_verdad": [
            "Genera a $\mathbb{R}^2$ porque la suma de sus tres vectores produce el vector $(2, 2)$, que es el eje de simetría del plano.",
            "Al tener más vectores que la dimensión del espacio (3 > 2), el espacio generado se expande por sobrecompletitud.",
            "El conjunto genera el plano porque sus vectores no son paralelos al vector nulo."
        ],
        "distractores_falso": [
            "No genera el espacio porque el tercer vector es una combinación lineal de los dos primeros, lo que 'anula' la generación.",
            "Solo los conjuntos con exactamente 2 vectores pueden generar un espacio de dimensión 2.",
            "El espacio generado es una recta porque el determinante de la matriz $3 \times 2$ no está definido.",
            "Falla en generar $\mathbb{R}^2$ porque no incluye vectores con componentes negativas."
        ]
    },
    {
        "id": 4,
        "tema": "Combinación Lineal",
        "enunciado": "El vector $(5, 3)$ puede expresarse como combinación lineal de $(1, 0)$ y $(0, 1)$.",
        "real_truth": "Verdadero",
        "just_correcta": "Existen escalares $c_1=5$ y $c_2=3$ tales que $5(1,0) + 3(0,1) = (5,3)$, satisfaciendo la definición de combinación lineal.",
        "distractores_verdad": [
            "Es posible porque $(5, 3)$ se encuentra dentro del área barrida por los vectores de la base estándar.",
            "Todo vector cuya norma sea mayor a la de los vectores generadores es automáticamente una combinación de ellos.",
            "Al ser vectores ortogonales, la proyección de $(5,3)$ sobre ellos siempre es un número entero."
        ],
        "distractores_falso": [
            "No es posible porque los vectores generadores son unitarios y $(5,3)$ excede la magnitud permitida.",
            "Solo se puede si el producto punto entre el vector resultante y los generadores es igual a cero.",
            "La combinación lineal requiere que los escalares sean potencias de 2 para mantener la consistencia en $\mathbb{R}^2$.",
            "El vector $(5,3)$ pertenece a un subespacio diferente debido a su pendiente irracional."
        ]
    },
    {
        "id": 5,
        "tema": "Dimensión",
        "enunciado": "La dimensión de un subespacio que consiste únicamente en el vector nulo $\{(0, 0)\}$ es 1.",
        "real_truth": "Falso",
        "just_correcta": "La dimensión del subespacio trivial es 0, ya que su base es el conjunto vacío y no tiene grados de libertad.",
        "distractores_falso": [
            "Es 0 porque el vector nulo no ocupa un lugar geométrico en el plano cartesiano.",
            "La dimensión es nula porque el vector $(0,0)$ no tiene una norma calculable en el campo de los reales.",
            "Un punto siempre tiene dimensión negativa en el álgebra lineal abstracta."
        ],
        "distractores_verdad": [
            "Es 1 porque el conjunto contiene exactamente un elemento (el vector nulo).",
            "La dimensión es 1 porque el vector nulo es capaz de generar cualquier otro vector multiplicado por cero.",
            "Al estar en $\mathbb{R}^2$, cualquier subconjunto hereda al menos una de las dimensiones del espacio original.",
            "Es 1 porque el origen representa el eje fundamental de cualquier rotación en el plano."
        ]
    },
    {
        "id": 6,
        "tema": "Base",
        "enunciado": "Cualquier conjunto de dos vectores linealmente independientes en $\mathbb{R}^2$ es una base para $\mathbb{R}^2$.",
        "real_truth": "Verdadero",
        "just_correcta": "En un espacio de dimensión $n$, cualquier conjunto de $n$ vectores linealmente independientes es automáticamente un conjunto generador y, por tanto, una base.",
        "distractores_verdad": [
            "Es una base siempre que los vectores tengan longitudes diferentes para cubrir distintas escalas del plano.",
            "La independencia asegura que no haya pérdida de información, permitiendo que los vectores se multipliquen entre sí.",
            "Dos vectores LI forman una base porque el ángulo entre ellos es siempre de 90 grados en el espacio euclidiano."
        ],
        "distractores_falso": [
            "Falso, además de ser LI deben ser ortogonales y unitarios para ser considerados una base legal.",
            "No basta con ser LI; se requiere que los vectores pertenezcan a la base canónica definida por el sistema.",
            "Solo pueden ser base si su suma da como resultado el vector $(1, 1)$.",
            "Una base en $\mathbb{R}^2$ requiere al menos 3 vectores para asegurar la cobertura total de los 4 cuadrantes."
        ]
    },
    {
        "id": 7,
        "tema": "Subespacios en R^2",
        "enunciado": "El conjunto $V = \{(x, y) \in \mathbb{R}^2 \mid x + y = 1\}$ es un subespacio vectorial de $\mathbb{R}^2$.",
        "real_truth": "Falso",
        "just_correcta": "No contiene al vector nulo $(0,0)$ porque $0+0 \neq 1$, y tampoco cumple la cerradura bajo la suma.",
        "distractores_falso": [
            "No es subespacio porque representa una línea con pendiente negativa.",
            "Falla porque los componentes $x$ e $y$ están limitados por una constante unitaria.",
            "No es subespacio porque la suma de las coordenadas es un número impar."
        ],
        "distractores_verdad": [
            "Es un subespacio porque es un subconjunto de $\mathbb{R}^2$ definido por una relación lineal entre sus variables.",
            "Cumple la cerradura escalar ya que cualquier múltiplo de la recta sigue siendo una línea en el plano.",
            "Es un subespacio afín, lo cual es equivalente a un subespacio vectorial en la geometría moderna.",
            "La presencia del número 1 en la ecuación garantiza que el espacio tenga una identidad aditiva."
        ]
    },
    {
        "id": 8,
        "tema": "Combinación Lineal",
        "enunciado": "Si $u$ y $v$ son vectores en $\mathbb{R}^2$, entonces $3u - 2v$ es una combinación lineal de $\{u, v\}$.",
        "real_truth": "Verdadero",
        "just_correcta": "Se ajusta a la forma $c_1u + c_2v$ con escalares $c_1=3$ y $c_2=-2$, que pertenecen al campo de los reales.",
        "distractores_verdad": [
            "Es una combinación lineal porque el resultado es un vector que vive en el mismo plano que $u$ y $v$.",
            "Cualquier operación aritmética entre vectores resulta en una combinación lineal por la propiedad asociativa.",
            "Es válido porque los coeficientes 3 y -2 son números primos entre sí."
        ],
        "distractores_falso": [
            "Falso, las combinaciones lineales solo permiten coeficientes positivos.",
            "No es combinación lineal porque el signo negativo indica una resta, la cual no está definida en los axiomas de espacio vectorial.",
            "Solo es combinación lineal si $u$ y $v$ son perpendiculares.",
            "El escalar -2 invalida la operación al proyectar el vector fuera del subespacio original."
        ]
    },
    {
        "id": 9,
        "tema": "Independencia Lineal",
        "enunciado": "Un conjunto que contiene al vector nulo $\{(1, 2), (0, 0)\}$ es linealmente independiente.",
        "real_truth": "Falso",
        "just_correcta": "Cualquier conjunto que contenga al vector nulo es automáticamente dependiente, ya que podemos tener $0(1,2) + k(0,0) = (0,0)$ con $k \neq 0$.",
        "distractores_falso": [
            "Es dependiente porque el vector nulo tiene infinitas direcciones simultáneamente.",
            "La dependencia ocurre porque el determinante de cualquier conjunto con un cero es siempre 1.",
            "El vector $(1,2)$ absorbe al vector nulo bajo la operación de producto punto."
        ],
        "distractores_verdad": [
            "Es independiente porque el vector nulo no afecta la dirección del vector $(1,2)$.",
            "La independencia se mantiene ya que no existe un escalar que convierta al $(1,2)$ en $(0,0)$.",
            "En $\mathbb{R}^2$, el vector nulo actúa como un pivote que refuerza la base existente.",
            "Es independiente porque el conjunto tiene menos de 3 vectores."
        ]
    },
    {
        "id": 10,
        "tema": "Dimensión",
        "enunciado": "Si un subespacio de $\mathbb{R}^2$ es generado por un solo vector no nulo, su dimensión es 1.",
        "real_truth": "Verdadero",
        "just_correcta": "Un solo vector no nulo es linealmente independiente por sí mismo y forma una base para la recta que genera.",
        "distractores_verdad": [
            "La dimensión es 1 porque el vector representa una única coordenada en el sistema polar.",
            "Todo subespacio generado por un vector tiene dimensión 1 debido al teorema del rango-nulidad.",
            "Es dimensión 1 porque el vector puede estirarse infinitamente en una sola dirección."
        ],
        "distractores_falso": [
            "Falso, la dimensión es 2 porque el vector está compuesto por dos componentes $(x, y)$.",
            "La dimensión es 0 porque un solo vector no puede cerrar un área en el plano.",
            "Solo los pares de vectores tienen dimensión definida en el álgebra de Grassmann.",
            "La dimensión depende de la magnitud del vector; si es mayor a 1, la dimensión aumenta proporcionalmente."
        ]
    }
]

class NestedAlgebraQuiz:
    def __init__(self, data):
        self.data = data
        self.current_idx = 0
        self.main_output = widgets.Output()
        self.just_output = widgets.Output()
        self.render_question()

    def render_question(self):
        with self.main_output:
            clear_output()
            if self.current_idx >= len(self.data):
                display(HTML("<div style='border: 2px solid green; padding: 10px;'><h2>🏁 ¡Quiz Completado!</h2><p>Has repasado los conceptos clave de R².</p></div>"))
                return

            q = self.data[self.current_idx]
            display(HTML(f"<b style='color: #2c3e50;'>Tema: {q['tema']}</b>"))
            display(HTML(f"<div style='background-color: #f8f9fa; padding: 15px; border-radius: 5px; margin: 10px 0;'><h4>Pregunta {self.current_idx + 1}:</h4>{q['enunciado']}</div>"))
            
            # Paso 1: Selección de V/F
            self.vf_buttons = widgets.ToggleButtons(
                options=['Verdadero', 'Falso'],
                description='¿Es cierto?',
                button_style='', # 'success', 'info', 'warning', 'danger' or ''
                value=None,
                tooltips=['El enunciado es correcto', 'El enunciado es incorrecto']
            )
            self.vf_buttons.observe(self.on_vf_change, names='value')
            display(self.vf_buttons)
            
            # Contenedor para la segunda parte
            display(self.just_output)

    def on_vf_change(self, change):
        with self.just_output:
            clear_output()
            q = self.data[self.current_idx]
            seleccion_usuario = change['new']
            
            # Preparar opciones según la elección del usuario
            opciones = []
            if seleccion_usuario == q['real_truth']:
                # El usuario acertó la primera parte: mostramos 1 correcta y 3 distractores
                lista_distractores = q['distractores_verdad'] if q['real_truth'] == "Verdadero" else q['distractores_falso']
                opciones = [q['just_correcta']] + random.sample(lista_distractores, 3)
            else:
                # El usuario falló la primera parte: mostramos 4 distractores creíbles (ninguna es real)
                # O simplemente mostramos los distractores de la opción incorrecta que eligió
                lista_distractores = q['distractores_falso'] if seleccion_usuario == "Falso" else q['distractores_verdad']
                opciones = random.sample(lista_distractores, len(lista_distractores))

            random.shuffle(opciones)
            
            display(HTML("<p style='margin-top:20px;'><b>Selecciona la justificación técnica adecuada:</b></p>"))
            self.radio_just = widgets.RadioButtons(
                options=opciones,
                layout={'width': 'max-content'},
                value=None
            )
            display(self.radio_just)
            
            btn_validar = widgets.Button(description="Validar Respuesta", button_style='primary')
            btn_validar.on_click(self.check_final)
            display(btn_validar)

    def check_final(self, b):
        q = self.data[self.current_idx]
        user_vf = self.vf_buttons.value
        user_just = self.radio_just.value
        
        with self.just_output:
            # Primero borramos el botón de validar para no repetir
            if user_vf == q['real_truth'] and user_just == q['just_correcta']:
                display(HTML("<div style='color: #155724; background-color: #d4edda; padding: 10px; border-radius: 5px; margin-top: 10px;'><b>✅ ¡Excelente!</b> Has identificado tanto el valor de verdad como la justificación técnica correcta.</div>"))
            elif user_vf != q['real_truth']:
                display(HTML(f"<div style='color: #721c24; background-color: #f8d7da; padding: 10px; border-radius: 5px; margin-top: 10px;'><b>❌ Error de concepto:</b> El enunciado es en realidad <b>{q['real_truth']}</b>. Revisa los axiomas básicos.</div>"))
            else:
                display(HTML("<div style='color: #856404; background-color: #fff3cd; padding: 10px; border-radius: 5px; margin-top: 10px;'><b>⚠️ Casi...</b> Identificaste que es {user_vf}, pero elegiste una justificación falsa o incompleta.</div>"))
            
            btn_next = widgets.Button(description="Siguiente Pregunta ➡️", button_style='info')
            btn_next.on_click(self.next_question)
            display(btn_next)

    def next_question(self, b):
        self.current_idx += 1
        with self.just_output:
            clear_output()
        self.render_question()

    def start(self):
        display(self.main_output)

# Iniciar el cuestionario
quiz = NestedAlgebraQuiz(preguntas_db)
quiz.start()
```
