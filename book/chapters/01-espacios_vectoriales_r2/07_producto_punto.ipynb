# 🔹 Producto Punto

El **producto punto** (también llamado **producto escalar** o **producto interno**) es una operación fundamental entre dos vectores que produce un **número escalar** como resultado.

A diferencia de las operaciones que hemos visto anteriormente (suma de vectores, multiplicación por escalar), el producto punto nos permite:

- Calcular **ángulos** entre vectores
- Determinar si dos vectores son **perpendiculares**
- Calcular **proyecciones** de un vector sobre otro
- Medir la **magnitud** (norma) de un vector

En esta sección exploraremos las definiciones algebraica y geométrica del producto punto, sus propiedades, y sus aplicaciones más importantes.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import warnings
warnings.filterwarnings('ignore')
```

## 1. Definición Algebraica

:::{admonition} Definición: Producto Punto
:class: tip

Dados dos vectores $\mathbf{u} = (u_1, u_2)$ y $\mathbf{v} = (v_1, v_2)$ en $\mathbb{R}^2$, el **producto punto** se define como:

$$
\mathbf{u} \cdot \mathbf{v} = u_1 v_1 + u_2 v_2
$$

El resultado es un **escalar** (número real), no un vector.
:::

Esta operación también se conoce como:
- **Producto interno** (*inner product*)
- **Producto escalar** (*scalar product*)

### Ejemplos

**Ejemplo 1:** Sean $\mathbf{u} = (3, 4)$ y $\mathbf{v} = (2, 5)$:

$$
\mathbf{u} \cdot \mathbf{v} = (3)(2) + (4)(5) = 6 + 20 = 26
$$

**Ejemplo 2:** Sean $\mathbf{a} = (1, -2)$ y $\mathbf{b} = (4, 2)$:

$$
\mathbf{a} \cdot \mathbf{b} = (1)(4) + (-2)(2) = 4 - 4 = 0
$$

¡El resultado es cero! Esto tiene un significado geométrico especial que veremos más adelante.

```python
# Verificación con NumPy
u = np.array([3, 4])
v = np.array([2, 5])

# Método 1: usando np.dot
print(f"u · v = {np.dot(u, v)}")

# Método 2: usando el operador @
print(f"u · v = {u @ v}")

# Ejemplo 2
a = np.array([1, -2])
b = np.array([4, 2])
print(f"a · b = {np.dot(a, b)}")
```

## 2. Definición Geométrica

:::{admonition} Definición Geométrica del Producto Punto
:class: tip

El producto punto de dos vectores $\mathbf{u}$ y $\mathbf{v}$ también puede expresarse como:

$$
\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\| \|\mathbf{v}\| \cos\theta
$$

donde:
- $\|\mathbf{u}\|$ es la magnitud (norma) del vector $\mathbf{u}$
- $\|\mathbf{v}\|$ es la magnitud (norma) del vector $\mathbf{v}$
- $\theta$ es el ángulo entre los dos vectores
:::

### Demostración (bosquejo)

Para conectar ambas definiciones, consideremos dos vectores $\mathbf{u}$ y $\mathbf{v}$ con ángulos $\alpha$ y $\beta$ respecto al eje $x$:

$$
\mathbf{u} = \|\mathbf{u}\|(\cos\alpha, \sin\alpha), \quad \mathbf{v} = \|\mathbf{v}\|(\cos\beta, \sin\beta)
$$

El producto punto algebraico es:

$$
\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\|\|\mathbf{v}\|(\cos\alpha\cos\beta + \sin\alpha\sin\beta)
$$

Usando la identidad trigonométrica $\cos(\alpha - \beta) = \cos\alpha\cos\beta + \sin\alpha\sin\beta$:

$$
\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\|\|\mathbf{v}\|\cos(\alpha - \beta) = \|\mathbf{u}\|\|\mathbf{v}\|\cos\theta
$$

donde $\theta = \alpha - \beta$ es el ángulo entre los vectores.

```python
# Visualización del ángulo entre vectores
fig, ax = plt.subplots(figsize=(8, 8))

u = np.array([3, 1])
v = np.array([1, 2])

# Dibujar vectores
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.02, label=r'$\mathbf{u}$')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.02, label=r'$\mathbf{v}$')

# Calcular ángulo
theta = np.arccos(np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v)))
angle_u = np.arctan2(u[1], u[0])
angle_v = np.arctan2(v[1], v[0])

# Dibujar arco del ángulo
angles = np.linspace(min(angle_u, angle_v), max(angle_u, angle_v), 30)
r = 0.5
ax.plot(r * np.cos(angles), r * np.sin(angles), 'g-', linewidth=2)
ax.annotate(r'$\theta$', xy=(0.6, 0.3), fontsize=14, color='green')

# Configuración
ax.set_xlim(-0.5, 4)
ax.set_ylim(-0.5, 3)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.legend(fontsize=12)
ax.set_title(f'Ángulo entre vectores: θ = {np.degrees(theta):.1f}°', fontsize=14)

plt.tight_layout()
plt.show()
```

## 3. Propiedades

:::{admonition} Teorema: Propiedades del Producto Punto
:class: important

Sean $\mathbf{u}$, $\mathbf{v}$ y $\mathbf{w}$ vectores en $\mathbb{R}^2$, y $c$ un escalar. Entonces:

1. **Conmutativa:** $\mathbf{u} \cdot \mathbf{v} = \mathbf{v} \cdot \mathbf{u}$

2. **Distributiva:** $\mathbf{u} \cdot (\mathbf{v} + \mathbf{w}) = \mathbf{u} \cdot \mathbf{v} + \mathbf{u} \cdot \mathbf{w}$

3. **Asociativa con escalares:** $(c\mathbf{u}) \cdot \mathbf{v} = c(\mathbf{u} \cdot \mathbf{v}) = \mathbf{u} \cdot (c\mathbf{v})$

4. **Positiva definida:** $\mathbf{v} \cdot \mathbf{v} \geq 0$, con igualdad si y solo si $\mathbf{v} = \mathbf{0}$
:::

### Verificación numérica

```python
u = np.array([2, 3])
v = np.array([4, -1])
w = np.array([1, 5])
c = 3

# Propiedad conmutativa
print("Conmutativa:")
print(f"  u · v = {np.dot(u, v)}")
print(f"  v · u = {np.dot(v, u)}")

# Propiedad distributiva
print("\nDistributiva:")
print(f"  u · (v + w) = {np.dot(u, v + w)}")
print(f"  u · v + u · w = {np.dot(u, v) + np.dot(u, w)}")

# Asociativa con escalares
print("\nAsociativa con escalares:")
print(f"  (cu) · v = {np.dot(c*u, v)}")
print(f"  c(u · v) = {c * np.dot(u, v)}")
```

## 4. Norma (Magnitud)

El producto punto nos proporciona una forma elegante de calcular la **norma** (o magnitud) de un vector.

:::{admonition} Definición: Norma de un Vector
:class: tip

La **norma** (o magnitud) de un vector $\mathbf{v} = (v_1, v_2)$ se define como:

$$
\|\mathbf{v}\| = \sqrt{\mathbf{v} \cdot \mathbf{v}} = \sqrt{v_1^2 + v_2^2}
$$

Geométricamente, representa la **longitud** del vector.
:::

### Vectores Unitarios

Un **vector unitario** es aquel cuya norma es igual a 1:

$$
\|\hat{\mathbf{v}}\| = 1
$$

Los vectores unitarios estándar en $\mathbb{R}^2$ son:
- $\hat{\mathbf{i}} = (1, 0)$ → dirección del eje $x$
- $\hat{\mathbf{j}} = (0, 1)$ → dirección del eje $y$

### Normalización

Cualquier vector no nulo puede **normalizarse** (convertirse en unitario) dividiéndolo por su norma:

$$
\hat{\mathbf{v}} = \frac{\mathbf{v}}{\|\mathbf{v}\|}
$$

El vector normalizado $\hat{\mathbf{v}}$ tiene la **misma dirección** que $\mathbf{v}$ pero magnitud 1.

```python
v = np.array([3, 4])

# Calcular la norma
norma = np.linalg.norm(v)
print(f"v = {v}")
print(f"||v|| = √(3² + 4²) = √25 = {norma}")

# Normalizar el vector
v_normalizado = v / norma
print(f"\nVector normalizado: v̂ = {v_normalizado}")
print(f"Norma del vector normalizado: ||v̂|| = {np.linalg.norm(v_normalizado)}")
```

## 5. Ángulo entre Vectores

De la definición geométrica del producto punto, podemos despejar el ángulo entre dos vectores:

:::{admonition} Fórmula del Ángulo entre Vectores
:class: tip

$$
\cos\theta = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \|\mathbf{v}\|}
$$

Por lo tanto:

$$
\theta = \arccos\left(\frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \|\mathbf{v}\|}\right)
$$
:::

### Ejemplo Resuelto

Encontrar el ángulo entre $\mathbf{u} = (1, 3)$ y $\mathbf{v} = (2, 1)$.

**Solución:**

1. Producto punto: $\mathbf{u} \cdot \mathbf{v} = (1)(2) + (3)(1) = 5$

2. Normas: $\|\mathbf{u}\| = \sqrt{1 + 9} = \sqrt{10}$, $\|\mathbf{v}\| = \sqrt{4 + 1} = \sqrt{5}$

3. Coseno del ángulo: $\cos\theta = \frac{5}{\sqrt{10}\sqrt{5}} = \frac{5}{\sqrt{50}} = \frac{5}{5\sqrt{2}} = \frac{1}{\sqrt{2}}$

4. Ángulo: $\theta = \arccos\left(\frac{1}{\sqrt{2}}\right) = 45°$

```python
# Verificación computacional
u = np.array([1, 3])
v = np.array([2, 1])

# Producto punto
producto = np.dot(u, v)
print(f"u · v = {producto}")

# Normas
norma_u = np.linalg.norm(u)
norma_v = np.linalg.norm(v)
print(f"||u|| = {norma_u:.4f}")
print(f"||v|| = {norma_v:.4f}")

# Coseno del ángulo
cos_theta = producto / (norma_u * norma_v)
print(f"\ncos(θ) = {cos_theta:.4f}")

# Ángulo en radianes y grados
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)
print(f"θ = {theta_rad:.4f} radianes = {theta_deg:.1f}°")
```

## 6.  Exploración: Producto Punto y Ángulo

En esta exploración interactiva, fijaremos el vector $\mathbf{u} = (1, 0)$ y rotaremos el vector $\mathbf{v}$ usando un control deslizante para el ángulo.

Observa cómo cambia el producto punto en función del ángulo.

```python
def explorar_producto_punto(theta_grados=45):
    theta = np.radians(theta_grados)
    
    # Vector fijo
    u = np.array([1, 0])
    # Vector rotado
    v = np.array([np.cos(theta), np.sin(theta)])
    
    # Calcular producto punto
    producto = np.dot(u, v)
    cos_theta = producto  # Ambos vectores son unitarios
    
    # Crear figura
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Dibujar círculo unitario
    circle = plt.Circle((0, 0), 1, fill=False, color='lightgray', linestyle='--')
    ax.add_patch(circle)
    
    # Dibujar vectores
    ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1,
              color='blue', width=0.03, label=r'$\mathbf{u} = (1, 0)$')
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
              color='red', width=0.03, label=f'$\\mathbf{{v}} = ({v[0]:.2f}, {v[1]:.2f})$')
    
    # Dibujar arco del ángulo
    if theta_grados != 0:
        angles = np.linspace(0, theta, 30)
        r = 0.3
        ax.plot(r * np.cos(angles), r * np.sin(angles), 'g-', linewidth=2)
    
    # Configuración
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.legend(loc='upper right', fontsize=11)
    
    # Mostrar información
    info_text = f'θ = {theta_grados}°\n'
    info_text += f'cos(θ) = {cos_theta:.4f}\n'
    info_text += f'u · v = {producto:.4f}'
    
    # Colorear según el signo del producto punto
    if producto > 0.01:
        color = 'green'
        estado = '(POSITIVO)'
    elif producto < -0.01:
        color = 'red'
        estado = '(NEGATIVO)'
    else:
        color = 'orange'
        estado = '(CERO)'
    
    ax.text(0.02, 0.98, info_text + f'\n{estado}', transform=ax.transAxes, fontsize=12,
            verticalalignment='top', bbox=dict(boxstyle='round', facecolor=color, alpha=0.3))
    
    ax.set_title('Producto Punto: u · v = ||u|| ||v|| cos(θ)', fontsize=14)
    plt.tight_layout()
    plt.show()

# Crear widget interactivo
interact(explorar_producto_punto, 
         theta_grados=FloatSlider(min=0, max=360, step=5, value=45, 
                                   description='θ (grados):', 
                                   continuous_update=True,
                                   style={'description_width': 'initial'}));
```

## 7. Ortogonalidad

:::{admonition} Teorema: Condición de Ortogonalidad
:class: important

Dos vectores $\mathbf{u}$ y $\mathbf{v}$ son **ortogonales** (perpendiculares) si y solo si su producto punto es cero:

$$
\mathbf{u} \perp \mathbf{v} \iff \mathbf{u} \cdot \mathbf{v} = 0
$$
:::

### Interpretación Geométrica

Si $\mathbf{u} \cdot \mathbf{v} = 0$, entonces:

$$
\|\mathbf{u}\| \|\mathbf{v}\| \cos\theta = 0
$$

Si ambos vectores son no nulos ($\|\mathbf{u}\| \neq 0$ y $\|\mathbf{v}\| \neq 0$), entonces debe ser que:

$$
\cos\theta = 0 \implies \theta = 90°
$$

### Ejemplos de Vectores Ortogonales

```python
# Ejemplo 1: Vectores unitarios estándar
i = np.array([1, 0])
j = np.array([0, 1])
print(f"i · j = {np.dot(i, j)} → ortogonales")

# Ejemplo 2: Otros vectores ortogonales
u = np.array([3, 4])
v = np.array([-4, 3])  # Rotación de 90°
print(f"u = {u}, v = {v}")
print(f"u · v = {np.dot(u, v)} → ortogonales")

# Ejemplo 3: Verificar ortogonalidad
a = np.array([2, 5])
b = np.array([5, -2])
print(f"\na = {a}, b = {b}")
print(f"a · b = {np.dot(a, b)} → ortogonales")
```

```python
# Visualización de vectores ortogonales
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

ejemplos = [
    (np.array([1, 0]), np.array([0, 1]), r'$\hat{i}$ y $\hat{j}$'),
    (np.array([3, 4]), np.array([-4, 3]), r'$(3,4)$ y $(-4,3)$'),
    (np.array([2, 5]), np.array([5, -2]), r'$(2,5)$ y $(5,-2)$')
]

for ax, (u, v, titulo) in zip(axes, ejemplos):
    # Normalizar para mejor visualización
    max_val = max(np.max(np.abs(u)), np.max(np.abs(v))) + 1
    
    ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1,
              color='blue', width=0.03, label=r'$\mathbf{u}$')
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
              color='red', width=0.03, label=r'$\mathbf{v}$')
    
    ax.set_xlim(-max_val, max_val)
    ax.set_ylim(-max_val, max_val)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.legend()
    ax.set_title(f'{titulo}\nu · v = {np.dot(u, v)}', fontsize=12)

plt.suptitle('Ejemplos de Vectores Ortogonales', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

## 8. Proyección de un Vector

La **proyección** de un vector sobre otro es una de las aplicaciones más importantes del producto punto.

:::{admonition} Definición: Proyección y Componente Escalar
:class: tip

La **proyección vectorial** de $\mathbf{u}$ sobre $\mathbf{v}$ es:

$$
\text{proy}_{\mathbf{v}}\mathbf{u} = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{v}\|^2}\mathbf{v} = \frac{\mathbf{u} \cdot \mathbf{v}}{\mathbf{v} \cdot \mathbf{v}}\mathbf{v}
$$

La **componente escalar** (o proyección escalar) de $\mathbf{u}$ sobre $\mathbf{v}$ es:

$$
\text{comp}_{\mathbf{v}}\mathbf{u} = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{v}\|}
$$

La componente escalar es la "longitud con signo" de la proyección vectorial.
:::

### Interpretación Geométrica

Imagina que una luz brilla perpendicularmente a $\mathbf{v}$. La proyección de $\mathbf{u}$ sobre $\mathbf{v}$ es la "sombra" que $\mathbf{u}$ proyecta sobre la línea definida por $\mathbf{v}$.

```python
# Ejemplo: Proyección de u sobre v
u = np.array([3, 4])
v = np.array([5, 0])

# Proyección vectorial
proy_v_u = (np.dot(u, v) / np.dot(v, v)) * v
print(f"u = {u}")
print(f"v = {v}")
print(f"\nProyección de u sobre v: proy_v(u) = {proy_v_u}")

# Componente escalar
comp_v_u = np.dot(u, v) / np.linalg.norm(v)
print(f"Componente escalar: comp_v(u) = {comp_v_u}")
```

```python
# Visualización de la proyección
fig, ax = plt.subplots(figsize=(10, 8))

u = np.array([3, 4])
v = np.array([5, 1])

# Calcular proyección
proy = (np.dot(u, v) / np.dot(v, v)) * v

# Dibujar vectores
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.02, label=r'$\mathbf{u}$')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
          color='red', width=0.02, label=r'$\mathbf{v}$')
ax.quiver(0, 0, proy[0], proy[1], angles='xy', scale_units='xy', scale=1,
          color='green', width=0.02, label=r'$\text{proy}_{\mathbf{v}}\mathbf{u}$')

# Línea punteada desde u hasta la proyección
ax.plot([u[0], proy[0]], [u[1], proy[1]], 'k--', linewidth=1, alpha=0.7)

# Extensión de v (línea guía)
t = np.linspace(-0.5, 2, 100)
v_norm = v / np.linalg.norm(v)
ax.plot(t * v[0] / np.linalg.norm(v) * 6, t * v[1] / np.linalg.norm(v) * 6, 
        'r--', alpha=0.3, linewidth=1)

# Marcar ángulo recto
perp_size = 0.3
perp_dir = (u - proy) / np.linalg.norm(u - proy)
proj_dir = proy / np.linalg.norm(proy)
corner = proy + perp_size * proj_dir
ax.plot([proy[0], proy[0] + perp_size * perp_dir[0]], 
        [proy[1], proy[1] + perp_size * perp_dir[1]], 'k-', linewidth=1)
ax.plot([proy[0] + perp_size * perp_dir[0], proy[0] + perp_size * (perp_dir[0] - proj_dir[0])], 
        [proy[1] + perp_size * perp_dir[1], proy[1] + perp_size * (perp_dir[1] - proj_dir[1])], 'k-', linewidth=1)

# Configuración
ax.set_xlim(-1, 6)
ax.set_ylim(-1, 5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.legend(fontsize=12, loc='upper left')
ax.set_title('Proyección de u sobre v', fontsize=14)

plt.tight_layout()
plt.show()
```

## 9.  Exploración: Proyecciones

En esta exploración interactiva, fijaremos $\mathbf{v} = (3, 1)$ y moveremos $\mathbf{u}$ usando controles deslizantes.

Observa cómo cambia la proyección de $\mathbf{u}$ sobre $\mathbf{v}$.

```python
def explorar_proyeccion(u1=2, u2=3):
    v = np.array([3, 1])
    u = np.array([u1, u2])
    
    # Calcular proyección
    proy = (np.dot(u, v) / np.dot(v, v)) * v
    comp_escalar = np.dot(u, v) / np.linalg.norm(v)
    
    # Crear figura
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Extensión de v (línea guía)
    t = np.linspace(-2, 2, 100)
    ax.plot(t * v[0], t * v[1], 'r--', alpha=0.3, linewidth=1, label='Línea de v')
    
    # Dibujar vectores
    ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1,
              color='blue', width=0.02, label=f'u = ({u1}, {u2})')
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
              color='red', width=0.02, label='v = (3, 1)')
    ax.quiver(0, 0, proy[0], proy[1], angles='xy', scale_units='xy', scale=1,
              color='green', width=0.025, label=f'proy = ({proy[0]:.2f}, {proy[1]:.2f})')
    
    # Línea punteada desde u hasta la proyección
    ax.plot([u[0], proy[0]], [u[1], proy[1]], 'k--', linewidth=1.5, alpha=0.7)
    
    # Configuración
    ax.set_xlim(-5, 6)
    ax.set_ylim(-4, 5)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.legend(fontsize=11, loc='upper left')
    
    # Mostrar información
    info_text = f'u · v = {np.dot(u, v):.2f}\n'
    info_text += f'||v||² = {np.dot(v, v):.2f}\n'
    info_text += f'Componente escalar = {comp_escalar:.2f}\n'
    info_text += f'||proy|| = {np.linalg.norm(proy):.2f}'
    
    ax.text(0.98, 0.02, info_text, transform=ax.transAxes, fontsize=11,
            verticalalignment='bottom', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    ax.set_title('Proyección de u sobre v', fontsize=14)
    plt.tight_layout()
    plt.show()

# Crear widget interactivo
interact(explorar_proyeccion, 
         u1=FloatSlider(min=-4, max=5, step=0.5, value=2, description='u₁:'),
         u2=FloatSlider(min=-3, max=4, step=0.5, value=3, description='u₂:'));
```

Dados los vectores $\mathbf{u} = (3, 4)$ y $\mathbf{v} = (1, 2)$, calculemos:

1. El producto punto
2. El ángulo entre ellos

### Solución Analítica

**1. Producto punto:**
$$\mathbf{u} \cdot \mathbf{v} = (3)(1) + (4)(2) = 3 + 8 = 11$$

**2. Ángulo:**
$$\|\mathbf{u}\| = \sqrt{9 + 16} = 5, \quad \|\mathbf{v}\| = \sqrt{1 + 4} = \sqrt{5}$$

$$\cos\theta = \frac{11}{5\sqrt{5}} = \frac{11}{5\sqrt{5}} \approx 0.9839$$

$$\theta = \arccos(0.9839) \approx 10.3°$$

**3. Ortogonalidad:**
Como $\mathbf{u} \cdot \mathbf{v} = 11 \neq 0$, los vectores **no son ortogonales**.

**4. Proyección:**
$$\text{proy}_{\mathbf{v}}\mathbf{u} = \frac{11}{5}(1, 2) = (2.2, 4.4)$$

```python
# Verificación con NumPy
u = np.array([3, 4])
v = np.array([1, 2])

# 1. Producto punto
producto = np.dot(u, v)
print(f"1. Producto punto: u · v = {producto}")

# 2. Ángulo
norma_u = np.linalg.norm(u)
norma_v = np.linalg.norm(v)
cos_theta = producto / (norma_u * norma_v)
theta = np.degrees(np.arccos(cos_theta))
print(f"\n2. Ángulo:")
print(f"   ||u|| = {norma_u}")
print(f"   ||v|| = {norma_v:.4f}")
print(f"   cos(θ) = {cos_theta:.4f}")
print(f"   θ = {theta:.2f}°")

# 3. Ortogonalidad
print(f"\n3. ¿Son ortogonales? {'Sí' if producto == 0 else 'No'} (u · v = {producto})")

# 4. Proyección
proy = (producto / np.dot(v, v)) * v
print(f"\n4. Proyección de u sobre v: proy = {proy}")
```
