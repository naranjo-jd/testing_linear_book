Observa la siguiente casa que está dibujada en plano cartesiano con un área de 4.5 unidades cuadradas.
Busca una forma de de trasladar la casa al punto $(5,4)$ con el doble de área y que mantenga la misma forma y proporciones.

```python
import matplotlib.pyplot as plt
import numpy as np

# Calcular factor de escala k para que el Área Total sea 4
# La casa se compone de:
# 1. Un rectángulo base de ancho 2k y alto 2k (Área = 4k^2)
# 2. Un triángulo techo de base 2k y altura k (Área = k^2)
# Área Total = 5k^2 = 4  =>  k = sqrt(4/5)
k = np.sqrt(4/5)

# Definir vértices de la casa 
# Para centrarla en el origen (centro geométrico del bounding box):
# Coordenadas X van de -k a k (centro 0)
# Coordenadas Y van de -1.5k a 1.5k (centro 0)
vertices_x = [-k, k, k, 0, -k, -k]  # Cerrando el polígono
vertices_y = [-1.5*k, -1.5*k, 0.5*k, 1.5*k, 0.5*k, -1.5*k]

# Crear la figura
fig, ax = plt.subplots(figsize=(6, 6))

# Dibujar la casa
ax.plot(vertices_x, vertices_y, color='darkblue', linewidth=2, label='Casa')
ax.fill(vertices_x, vertices_y, color='skyblue', alpha=0.6)

# Configurar el gráfico
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True, linestyle='--', alpha=0.7)
ax.set_aspect('equal')
ax.set_xlim(-2, 6)
ax.set_ylim(-2, 6)
ax.set_title(f'Casa centrada en el origen (Área = 4)')
ax.legend()

plt.show()
```
