## 🟦 Ejercicio 1: Matriz Triangular — Autovalores sin Determinante

Sea:
$$A = \begin{pmatrix} 2 & 1 \\ 0 & 3 \end{pmatrix}$$

**Propiedad:** Para una matriz triangular (superior o inferior), los autovalores son
exactamente las **entradas de la diagonal principal**.

**Demostración:**
$$\det(A - \lambda I) = \det\begin{pmatrix} 2-\lambda & 1 \\ 0 & 3-\lambda \end{pmatrix}
= (2-\lambda)(3-\lambda)$$

Por lo tanto $\lambda_1 = 2$, $\lambda_2 = 3$, que son exactamente las entradas $(1,1)$ y $(2,2)$.

**Generalización:** Para una matriz triangular $n\times n$:
$$p(\lambda) = \prod_{i=1}^n (a_{ii} - \lambda)$$

Los autovalores son siempre $a_{11}, a_{22}, \ldots, a_{nn}$.

```python
import sympy as sp

lam = sp.Symbol('lambda')

print('Ejercicio 1: Matrices triangulares')
print('='*55)

matrices_triangulares = [
    ('Triangular superior 2x2', sp.Matrix([[2, 1], [0, 3]])),
    ('Triangular inferior 2x2', sp.Matrix([[5, 0], [2, 4]])),
    ('Triangular superior 3x3', sp.Matrix([[1, 2, 3], [0, 4, 5], [0, 0, 6]])),
    ('Diagonal 3x3', sp.Matrix([[7, 0, 0], [0, 3, 0], [0, 0, -1]])),
]

for nombre, A in matrices_triangulares:
    n = A.shape[0]
    char_poly = sp.expand((A - lam*sp.eye(n)).det())
    roots = sp.solve(char_poly, lam)
    diag_entries = [A[i, i] for i in range(n)]
    match = sorted([float(r) for r in roots]) == sorted([float(d) for d in diag_entries])
    print(f'\n{nombre}:')
    print(f'  Diagonal: {diag_entries}')
    print(f'  Autovalores: {roots}')
    print(f'  ¿Coinciden? {match} ✅')
```

## 🟩 Ejercicio 2: Demostración — $A\mathbf{v} = \lambda\mathbf{v}$ implica $A^2\mathbf{v} = \lambda^2\mathbf{v}$

**Enunciado:** Si $\mathbf{v}$ es autovector de $A$ con autovalor $\lambda$, entonces $\mathbf{v}$ también
es autovector de $A^k$ con autovalor $\lambda^k$, para todo entero $k \geq 0$.

**Demostración (por inducción):**

**Caso base** ($k = 1$): Por hipótesis, $A\mathbf{v} = \lambda\mathbf{v}$. ✓

**Paso inductivo:** Supongamos que $A^k\mathbf{v} = \lambda^k\mathbf{v}$. Entonces:
$$A^{k+1}\mathbf{v} = A \cdot (A^k\mathbf{v}) = A \cdot (\lambda^k\mathbf{v}) = \lambda^k (A\mathbf{v}) = \lambda^k \cdot \lambda\mathbf{v} = \lambda^{k+1}\mathbf{v}$$

Por inducción, se cumple para todo $k \geq 0$. $\blacksquare$

**Corolario:** Si $A$ es invertible ($\lambda \neq 0$), entonces $A^{-1}\mathbf{v} = \frac{1}{\lambda}\mathbf{v}$.

*Prueba del corolario:* Aplicar $A^{-1}$ a $A\mathbf{v} = \lambda\mathbf{v}$:
$$\mathbf{v} = A^{-1}(\lambda\mathbf{v}) = \lambda A^{-1}\mathbf{v} \implies A^{-1}\mathbf{v} = \frac{1}{\lambda}\mathbf{v}$$

```python
import numpy as np

print('Ejercicio 2: Verificación numérica de A^k v = λ^k v')
print('='*55)

A = np.array([[3.0, 1.0], [0.0, 2.0]])
eigenvalues, eigenvectors = np.linalg.eig(A)

for i, (lam, v) in enumerate(zip(eigenvalues, eigenvectors.T)):
    print(f'\nAutovector v_{i+1} con λ_{i+1} = {lam}:')
    for k in [1, 2, 3, 5, 10]:
        Ak = np.linalg.matrix_power(A, k)
        Akv = Ak @ v
        lamk_v = lam**k * v
        ok = '✅' if np.allclose(Akv, lamk_v, rtol=1e-8) else '❌'
        print(f'  k={k:2d}: A^k·v = λ^k·v  {ok}  (λ^k = {lam**k:.4f})')

# Corolario: A^{-1} v = (1/λ) v
print('\nCorolario: A⁻¹v = (1/λ)v')
A_inv = np.linalg.inv(A)
for i, (lam, v) in enumerate(zip(eigenvalues, eigenvectors.T)):
    Ainv_v = A_inv @ v
    inv_lam_v = (1.0/lam) * v
    ok = '✅' if np.allclose(Ainv_v, inv_lam_v) else '❌'
    print(f'  v_{i+1}: A⁻¹v = {np.round(Ainv_v,4)}, (1/λ)v = {np.round(inv_lam_v,4)} {ok}')
```

## 🟨 Ejercicio 3: ¿Es $A = \begin{pmatrix} 1 & 1 \\ 0 & 1 \end{pmatrix}$ Diagonalizable?

Esta matriz es el **bloque de Jordan** de tamaño 2, y es el contraejemplo clásico de
una matriz **no diagonalizable**.

**Análisis:**

1. **Polinomio característico:**
$$p(\lambda) = (1-\lambda)^2 \implies \lambda_1 = \lambda_2 = 1$$
Multiplicidad algebraica = 2.

2. **Subespacio propio:**
$$A - I = \begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix}$$
$$\ker(A - I) = \{\mathbf{v} : v_2 = 0\} = \text{span}\left\{\begin{pmatrix}1\\0\end{pmatrix}\right\}$$
Multiplicidad geométrica = 1.

3. **Conclusión:**
$$\underbrace{1}_{\text{mult. geom.}} < \underbrace{2}_{\text{mult. alg.}} \implies \text{NO diagonalizable}$$

> 🔑 Una matriz es diagonalizable $\Leftrightarrow$ mult. geométrica = mult. algebraica para **todos** sus autovalores.

```python
import sympy as sp
import numpy as np

print('Ejercicio 3: Bloque de Jordan — NO diagonalizable')
print('='*55)

A = sp.Matrix([[1, 1], [0, 1]])
lam = sp.Symbol('lambda')

# Polinomio característico
char_poly = sp.expand((A - lam*sp.eye(2)).det())
print(f'Polinomio característico: {char_poly}')

roots = sp.solve(char_poly, lam)
print(f'Autovalores: {roots}')
print(f'Multiplicidad algebraica de λ=1: 2')

# Subespacio propio
M = A - sp.eye(2)
null = M.nullspace()
print(f'\n(A - I) = {M}')
print(f'Kernel(A-I) = {null}')
print(f'Multiplicidad geométrica de λ=1: {len(null)}')
print()
print('Como mult. algebraica (2) > mult. geométrica (1):')
print('→ A NO es diagonalizable sobre ℝ.')

# Demostrar que no se puede diagonalizar
print('\nVerificación: si A fuera diagonalizable, debería existir P tal que')
print('P⁻¹AP = D = [[1,0],[0,1]] = I')
print('Pero eso implicaría A = PIP⁻¹ = I, lo cual es falso.')

# Forma de Jordan
print('\nForma canónica de Jordan de A:')
P_J, J = A.jordan_form()
print(f'Bloque de Jordan J = {J}')
print(f'P_Jordan = {P_J}')
print('(La diagonal no es 0 en las entradas superdiagonales → NO es diagonal)')
```

## 🟥 Ejercicio 4: Forma General del Polinomio Característico $2 \times 2$

**Enunciado:** Demostrar que para cualquier matriz $A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$,
el polinomio característico es:

$$p(\lambda) = \lambda^2 - \text{tr}(A)\cdot\lambda + \det(A)$$

**Demostración:**

$$p(\lambda) = \det(A - \lambda I) = \det\begin{pmatrix} a-\lambda & b \\ c & d-\lambda \end{pmatrix}$$

$$= (a-\lambda)(d-\lambda) - bc = ad - a\lambda - d\lambda + \lambda^2 - bc$$

$$= \lambda^2 - (a+d)\lambda + (ad-bc) = \lambda^2 - \text{tr}(A)\lambda + \det(A) \quad \blacksquare$$

**Consecuencia inmediata:** Las fórmulas de Vieta dan:
$$\lambda_1 + \lambda_2 = \text{tr}(A), \qquad \lambda_1 \cdot \lambda_2 = \det(A)$$

```python
import sympy as sp

print('Ejercicio 4: Verificación de p(λ) = λ² - tr(A)λ + det(A)')
print('='*55)

a, b, c, d, lam = sp.symbols('a b c d lambda')

A_gen = sp.Matrix([[a, b], [c, d]])
char_poly_gen = sp.expand((A_gen - lam*sp.eye(2)).det())
print(f'Forma general: p(λ) = {char_poly_gen}')

traza = a + d
det_val = a*d - b*c
formula = sp.expand(lam**2 - traza*lam + det_val)
print(f'Fórmula tr/det: λ² - tr(A)λ + det(A) = {formula}')
print(f'¿Son iguales? {sp.simplify(char_poly_gen - formula) == 0} ✅')

# Verificar con ejemplos concretos
print('\nVerificación con matrices concretas:')
examples = [
    sp.Matrix([[3, 1], [0, 2]]),
    sp.Matrix([[4, 2], [1, 3]]),
    sp.Matrix([[0, 1], [-1, 0]]),
]

lam_sym = sp.Symbol('lambda')
for Ae in examples:
    tr = Ae.trace()
    det = Ae.det()
    p_formula = sp.expand(lam_sym**2 - tr*lam_sym + det)
    p_direct = sp.expand((Ae - lam_sym*sp.eye(2)).det())
    match = sp.simplify(p_formula - p_direct) == 0
    roots = sp.solve(p_formula, lam_sym)
    print(f'  A={list(Ae)}: tr={tr}, det={det}')
    print(f'    p(λ) = {p_formula}, λ = {roots}, ¿OK? {match} ✅')
```

## 🟪 Ejercicio 5: Matrices Semejantes tienen los mismos Autovalores

**Enunciado:** Sean $A$ y $B = PAP^{-1}$ matrices semejantes (con $P$ invertible).
Demostrar que $A$ y $B$ tienen el mismo polinomio característico y, por tanto,
los mismos autovalores.

**Demostración:**

$$\det(B - \lambda I) = \det(PAP^{-1} - \lambda I)$$

$$= \det(PAP^{-1} - \lambda PP^{-1}) = \det(P(A - \lambda I)P^{-1})$$

$$= \det(P)\cdot\det(A - \lambda I)\cdot\det(P^{-1}) = \det(A-\lambda I) \cdot \underbrace{\det(P)\cdot\det(P^{-1})}_{=1}$$

$$= \det(A - \lambda I) \quad \blacksquare$$

**Interpretación geométrica:** Las matrices semejantes representan la **misma transformación lineal**
en distintas bases. Como los autovalores son propiedades de la transformación (no de la base),
deben coincidir.

```python
import numpy as np

print('Ejercicio 5: Matrices semejantes tienen los mismos autovalores')
print('='*55)

A = np.array([[3.0, 1.0], [2.0, 4.0]])

# Diferentes matrices de cambio de base
P_matrices = [
    np.array([[2.0, 1.0], [1.0, 1.0]]),
    np.array([[1.0, 3.0], [0.0, 1.0]]),
    np.array([[0.0, 1.0], [1.0, 0.0]]),
]

evals_A = np.sort(np.linalg.eigvals(A).real)
print(f'Autovalores de A: {np.round(evals_A, 6)}')
print(f'tr(A) = {np.trace(A)}, det(A) = {np.linalg.det(A):.4f}')
print()

for i, P in enumerate(P_matrices):
    B = P @ A @ np.linalg.inv(P)  # B = PAP^{-1}
    evals_B = np.sort(np.linalg.eigvals(B).real)
    tr_B = np.trace(B)
    det_B = np.linalg.det(B)
    same = np.allclose(evals_A, evals_B)
    print(f'P_{i+1}: B = PAP⁻¹')
    print(f'  Autovalores de B: {np.round(evals_B, 6)}')
    print(f'  tr(B) = {tr_B:.4f} (= tr(A)={np.trace(A)}) ✅')
    print(f'  det(B) = {det_B:.4f} (= det(A)={np.linalg.det(A):.4f}) ✅')
    print(f'  ¿Mismos autovalores que A? {same} ✅\n')
```
