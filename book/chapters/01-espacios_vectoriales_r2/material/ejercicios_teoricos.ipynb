```python

import matplotlib.pyplot as plt
import numpy as np

# Configurar figuras
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# --- Ejercicio 8 ---
# Definir una recta que pasa por puntos enteros claros
# Ejemplo: Pasa por (1, 1) y (2, 3). 
# Pendiente m = (3-1)/(2-1) = 2. Ecuación: y - 1 = 2(x - 1) => y = 2x - 1
x = np.linspace(-2, 4, 100)
y = 2 * x - 1

ax1.plot(x, y, label='Recta', color='blue', linewidth=2)
ax1.grid(True, which='both', linestyle='--', linewidth=0.7)
ax1.axhline(y=0, color='k', linewidth=1)
ax1.axvline(x=0, color='k', linewidth=1)

# Puntos enteros visuales para ayudar al estudiante
puntos_x = [0, 1, 2, 3]
puntos_y = [2*px - 1 for px in puntos_x]
ax1.scatter(puntos_x, puntos_y, color='black', s=30, zorder=5)

ax1.set_xlim(-2, 4)
ax1.set_ylim(-3, 6)
ax1.set_xticks(np.arange(-2, 5, 1))
ax1.set_yticks(np.arange(-3, 7, 1))
ax1.set_title("Ejercicio 8: Gráfica para identificar ecuación")
ax1.set_xlabel("x")
ax1.set_ylabel("y")


# --- Ejercicio 9 ---
# Punto P y vector director v
# P = (1, 2), v = (3, 1)
P = np.array([1, 2])
v = np.array([3, 1])

# Dibujar el punto P
ax2.scatter(P[0], P[1], color='green', s=100, label='Punto $P$', zorder=5)
ax2.text(P[0], P[1] + 0.3, "P", color='green', fontsize=14, fontweight='bold', ha='center')

# Dibujar el vector v partiendo de P
ax2.quiver(P[0], P[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector $\mathbf{v}$')
# Anotación del vector
ax2.text(P[0] + v[0]/2, P[1] + v[1]/2 + 0.2, r'$\mathbf{v}$', color='red', fontsize=14, fontweight='bold')

# Configuración del gráfico
ax2.grid(True, linestyle='--', linewidth=0.7)
ax2.axhline(y=0, color='k', linewidth=1)
ax2.axvline(x=0, color='k', linewidth=1)
ax2.set_xlim(0, 6)
ax2.set_ylim(0, 5)
ax2.set_xticks(np.arange(0, 7, 1))
ax2.set_yticks(np.arange(0, 6, 1))
ax2.set_title("Ejercicio 9: Punto y Vector Director")
ax2.set_aspect('equal')
ax2.legend()

plt.tight_layout()
plt.show()
```
