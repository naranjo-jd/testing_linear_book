## 🧱 Parte 1: La clase `TransformadorLineal2D`

Construiremos la clase con los siguientes métodos:

| Método | Descripción |
|--------|-------------|
| `rotar(theta_grados)` | Devuelve la matriz de rotación de ángulo $\theta$ |
| `reflejar(eje)` | Devuelve la matriz de reflexión sobre `'x'`, `'y'` o `'y=x'` |
| `escalar(sx, sy)` | Devuelve la matriz de escala por factores $s_x$ y $s_y$ |
| `cizallar(kx, ky)` | Devuelve la matriz de cizallamiento |
| `componer(*matrices)` | Compone varias matrices (en orden de aplicación) |
| `aplicar(M, puntos)` | Aplica la matriz $M$ a un array de puntos |
| `visualizar(original, transformado, titulo)` | Dibuja el polígono antes y después |

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

class TransformadorLineal2D:
    """Motor de transformaciones lineales en R^2 basado en matrices 2x2."""

    # ---- Matrices de transformacion ----

    @staticmethod
    def rotar(theta_grados):
        """Matriz de rotacion antihoraria por theta grados."""
        theta = np.radians(theta_grados)
        c, s = np.cos(theta), np.sin(theta)
        return np.array([[c, -s], [s, c]])

    @staticmethod
    def reflejar(eje='x'):
        """Matriz de reflexion. eje puede ser 'x', 'y' o 'y=x'."""
        if eje == 'x':
            return np.array([[1, 0], [0, -1]], dtype=float)
        elif eje == 'y':
            return np.array([[-1, 0], [0, 1]], dtype=float)
        elif eje == 'y=x':
            return np.array([[0, 1], [1, 0]], dtype=float)
        elif eje == 'y=-x':
            return np.array([[0, -1], [-1, 0]], dtype=float)
        else:
            raise ValueError(f"Eje desconocido: {eje}. Usa 'x', 'y', 'y=x' o 'y=-x'.")

    @staticmethod
    def escalar(sx, sy):
        """Matriz de escala: multiplica x por sx y y por sy."""
        return np.array([[sx, 0], [0, sy]], dtype=float)

    @staticmethod
    def cizallar(kx=0, ky=0):
        """Matriz de cizallamiento horizontal (kx) y vertical (ky)."""
        return np.array([[1, kx], [ky, 1]], dtype=float)

    @staticmethod
    def componer(*matrices):
        """
        Compone matrices en el orden dado (izquierda a derecha = primera a ultima aplicacion).
        componer(M1, M2, M3) devuelve M3 @ M2 @ M1 (aplica M1 primero).
        """
        resultado = np.eye(2)
        for M in matrices:
            resultado = M @ resultado
        return resultado

    @staticmethod
    def aplicar(M, puntos):
        """
        Aplica la matriz M a un conjunto de puntos.
        puntos: array de forma (2, N) donde cada columna es un punto.
        """
        return M @ puntos

    @staticmethod
    def visualizar(original, transformado, titulo='Transformacion',
                   ax=None, cerrar=True):
        """Dibuja el poligono original (azul) y transformado (rojo)."""
        if ax is None:
            fig, ax = plt.subplots(figsize=(6, 6))
        # Cerrar el poligono para graficar
        def cerrar_pol(pts):
            return np.hstack([pts, pts[:, :1]])
        orig_c = cerrar_pol(original)
        trans_c = cerrar_pol(transformado)
        ax.fill(orig_c[0], orig_c[1], alpha=0.25, color='steelblue')
        ax.plot(orig_c[0], orig_c[1], 'b--', lw=1.5, label='Original')
        ax.fill(trans_c[0], trans_c[1], alpha=0.4, color='tomato')
        ax.plot(trans_c[0], trans_c[1], 'r-', lw=2.5, label='Transformado')
        todos = np.hstack([original, transformado])
        margen = 0.5
        ax.set_xlim(todos[0].min()-margen, todos[0].max()+margen)
        ax.set_ylim(todos[1].min()-margen, todos[1].max()+margen)
        ax.set_aspect('equal')
        ax.axhline(0, color='gray', lw=0.5)
        ax.axvline(0, color='gray', lw=0.5)
        ax.legend(fontsize=10)
        ax.grid(True, alpha=0.3)
        ax.set_title(titulo, fontsize=11)

print("Clase TransformadorLineal2D definida correctamente.")

# Instanciar el motor
motor = TransformadorLineal2D()
print("Motor instanciado:", motor)
```

## 🏠 Parte 2: Definir el polígono — una casita

Usaremos una figura con forma de casita como polígono de prueba. Cada columna del array es un vértice $(x, y)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Casita: base + techo (en sentido antihorario)
# Vertices: esquina inferior izquierda, inferior derecha,
#           esquina superior derecha, punta del techo, esquina superior izquierda
casita = np.array([
    [-1.0, 1.0, 1.0, 0.0, -1.0],   # coordenadas x
    [ 0.0, 0.0, 1.0, 1.8,  1.0]    # coordenadas y
])

# Mostrar la casita original
fig, ax = plt.subplots(figsize=(5, 5))
casita_cerrada = np.hstack([casita, casita[:, :1]])
ax.fill(casita_cerrada[0], casita_cerrada[1], alpha=0.3, color='steelblue')
ax.plot(casita_cerrada[0], casita_cerrada[1], 'b-o', lw=2, markersize=6)
for i, (x, y) in enumerate(casita.T):
    ax.annotate(f'$P_{i+1}$', (x, y), textcoords="offset points",
                xytext=(5, 5), fontsize=10)
ax.set_xlim(-2.5, 2.5); ax.set_ylim(-0.5, 3.0)
ax.set_aspect('equal')
ax.axhline(0, color='gray', lw=0.5)
ax.axvline(0, color='gray', lw=0.5)
ax.grid(True, alpha=0.3)
ax.set_title('Polígono: casita original', fontsize=12)
plt.tight_layout()
plt.show()

print(f"Vertices de la casita:\n{casita.T}")
```

## 🔄 Parte 3: Aplicar transformaciones individuales

Aplicamos cada transformación básica a la casita y visualizamos los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt

motor = TransformadorLineal2D()

transformaciones = [
    ("Rotacion 45°",       motor.rotar(45)),
    ("Reflexion eje x",    motor.reflejar('x')),
    ("Escala (0.7, 1.3)",  motor.escalar(0.7, 1.3)),
    ("Cizallamiento kx=0.5", motor.cizallar(kx=0.5)),
]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for ax, (titulo, M) in zip(axes, transformaciones):
    casita_trans = motor.aplicar(M, casita)
    motor.visualizar(casita, casita_trans, titulo=titulo, ax=ax)
    # Mostrar la matriz
    mat_str = f"M = [[{M[0,0]:.2f}, {M[0,1]:.2f}], [{M[1,0]:.2f}, {M[1,1]:.2f}]]"
    ax.text(0.02, 0.02, mat_str, transform=ax.transAxes,
            fontsize=8, fontfamily='monospace', va='bottom')

plt.suptitle('Transformaciones individuales sobre la casita', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 🔗 Parte 4: Componer transformaciones

Comprobaremos que la composición de transformaciones corresponde al **producto de matrices**, y que el orden importa.

```python
import numpy as np
import matplotlib.pyplot as plt

motor = TransformadorLineal2D()

M_rot  = motor.rotar(60)
M_ref  = motor.reflejar('y')
M_esc  = motor.escalar(0.8, 1.2)

# Composicion: primero rotar, luego reflejar, luego escalar
M_comp_a = motor.componer(M_rot, M_ref, M_esc)
# Composicion en orden inverso: primero escalar, luego reflejar, luego rotar
M_comp_b = motor.componer(M_esc, M_ref, M_rot)

casita_a = motor.aplicar(M_comp_a, casita)
casita_b = motor.aplicar(M_comp_b, casita)

print("Composicion A (rotar -> reflejar -> escalar):")
print(np.round(M_comp_a, 4))
print("\nComposicion B (escalar -> reflejar -> rotar):")
print(np.round(M_comp_b, 4))
print(f"\n¿Son iguales? {np.allclose(M_comp_a, M_comp_b)}")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
motor.visualizar(casita, casita_a,
                 titulo='Composicion A: rotar(60°) → reflejar(y) → escalar', ax=axes[0])
motor.visualizar(casita, casita_b,
                 titulo='Composicion B: escalar → reflejar(y) → rotar(60°)', ax=axes[1])
plt.suptitle('El ORDEN de composicion importa (no conmutativa)', fontsize=13)
plt.tight_layout()
plt.show()
```

## 📊 Parte 5: Comparar determinantes y áreas

El determinante de la matriz de transformación indica cómo cambia el área del polígono: $\text{Área transformada} = |\det(M)| \cdot \text{Área original}$.

```python
import numpy as np

def area_poligono(pts):
    """Calcula el area de un poligono usando la formula del calzador (Shoelace)."""
    x, y = pts[0], pts[1]
    n = len(x)
    area = 0.0
    for i in range(n):
        j = (i + 1) % n
        area += x[i] * y[j] - x[j] * y[i]
    return abs(area) / 2

motor = TransformadorLineal2D()
area_orig = area_poligono(casita)
print(f"Area original de la casita: {area_orig:.4f}")

transformaciones_test = [
    ("Rotacion 30°",    motor.rotar(30)),
    ("Reflexion eje x", motor.reflejar('x')),
    ("Escala (2, 0.5)", motor.escalar(2, 0.5)),
    ("Escala (2, 3)",   motor.escalar(2, 3)),
    ("Cizallar kx=1",   motor.cizallar(kx=1)),
    ("Proyeccion eje x", np.array([[1,0],[0,0]], dtype=float)),
]

print(f"\n{'Transformacion':<22} {'det(M)':>8} {'Area trans':>12} {'|det|*Area_orig':>16} {'Coincide':>10}")
print("-" * 72)
for nombre, M in transformaciones_test:
    casita_t = motor.aplicar(M, casita)
    det_M    = np.linalg.det(M)
    area_t   = area_poligono(casita_t)
    predicha = abs(det_M) * area_orig
    ok       = np.isclose(area_t, predicha, atol=1e-6)
    print(f"{nombre:<22} {det_M:>8.4f} {area_t:>12.4f} {predicha:>16.4f} {'✅' if ok else '❌':>10}")
```

## 🎬 Parte 6: Reto final — Animación de rotación

Vamos a generar una secuencia de fotogramas que muestra la casita rotando de $0°$ a $360°$. Guardamos cada fotograma y creamos una visualización en grilla.

```python
import numpy as np
import matplotlib.pyplot as plt

motor = TransformadorLineal2D()

# Generar fotogramas de la animacion
angulos = np.linspace(0, 360, 13)[:-1]  # 12 fotogramas, excluir 360 (=0)

fig, axes = plt.subplots(3, 4, figsize=(14, 10))
axes = axes.flatten()

cmap_col = plt.cm.hsv

for i, (ax, ang) in enumerate(zip(axes, angulos)):
    M = motor.rotar(ang)
    casita_rot = motor.aplicar(M, casita)
    casita_c = np.hstack([casita_rot, casita_rot[:,:1]])
    orig_c   = np.hstack([casita, casita[:,:1]])
    col = cmap_col(i / len(angulos))
    ax.fill(orig_c[0], orig_c[1], alpha=0.1, color='steelblue')
    ax.plot(orig_c[0], orig_c[1], 'b--', lw=0.8, alpha=0.4)
    ax.fill(casita_c[0], casita_c[1], alpha=0.5, color=col)
    ax.plot(casita_c[0], casita_c[1], '-', color=col, lw=2)
    ax.set_xlim(-2.5, 2.5); ax.set_ylim(-2.5, 2.5)
    ax.set_aspect('equal')
    ax.axhline(0, color='gray', lw=0.3)
    ax.axvline(0, color='gray', lw=0.3)
    ax.set_title(f'$\\theta = {ang:.0f}°$', fontsize=10)
    ax.set_xticks([]); ax.set_yticks([])

plt.suptitle('Animacion de rotacion: casita girando de 0° a 330°',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("Fotogramas generados. Para una animacion real se puede usar")
print("matplotlib.animation.FuncAnimation o guardar cada frame como imagen.")
```

## 🏆 Reto adicional: Reflexión sobre una recta arbitraria

La reflexión sobre la recta $y = \tan(\alpha) \cdot x$ tiene la matriz:

$$R_\alpha = \begin{pmatrix}\cos(2\alpha) & \sin(2\alpha)\\ \sin(2\alpha) & -\cos(2\alpha)\end{pmatrix}$$

Agrega este método a la clase y visualiza la reflexión sobre distintas rectas.

```python
import numpy as np
import matplotlib.pyplot as plt

def reflexion_recta(alpha_grados):
    """Matriz de reflexion sobre la recta y = tan(alpha)*x."""
    a = np.radians(2 * alpha_grados)  # notar: 2*alpha
    return np.array([[np.cos(a), np.sin(a)],
                     [np.sin(a),-np.cos(a)]])

motor = TransformadorLineal2D()
angulos_recta = [0, 30, 45, 60, 90, 135]

fig, axes = plt.subplots(2, 3, figsize=(13, 8))
axes = axes.flatten()

for ax, ang in zip(axes, angulos_recta):
    M = reflexion_recta(ang)
    casita_ref = motor.aplicar(M, casita)

    # Dibujar la recta de reflexion
    t = np.linspace(-2.5, 2.5, 50)
    pendiente = np.tan(np.radians(ang))
    ax.plot(t, pendiente*t, 'k-', lw=1.5, alpha=0.6,
            label=f'$y={pendiente:.1f}x$')

    motor.visualizar(casita, casita_ref,
                     titulo=f'Reflexion sobre $y=\\tan({ang}°)x$', ax=ax)
    # Verificar R^2 = I (la reflexion es su propia inversa)
    es_involutiva = np.allclose(M @ M, np.eye(2))
    ax.text(0.02, 0.02, f'M²=I? {es_involutiva}',
            transform=ax.transAxes, fontsize=8, va='bottom')

plt.suptitle('Reflexión sobre distintas rectas que pasan por el origen',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print("Nota: toda reflexion es involutiva (su propia inversa): R^2 = I.")
```
