```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import random

# ═══════════════════════════════════════════════════════════════
#  BASE DE DATOS
# ═══════════════════════════════════════════════════════════════
preguntas_db = [
    {
        "id": 1,
        "tema": "Subespacios en ℝ²",
        "enunciado": r"El conjunto $W = \{(x, y) \in \mathbb{R}^2 \mid x - 2y = 0\}$ es un subespacio vectorial de $\mathbb{R}^2$.",
        "real_truth": "Verdadero",
        "just_correcta": r"Pasa por el origen (contiene el vector nulo), y es cerrado bajo suma y producto escalar: si $x_1=2y_1$ y $x_2=2y_2$, entonces $(x_1+x_2)=2(y_1+y_2)$ y $cx_1=2(cy_1)$.",
        "distractores_verdad": [
            r"Es subespacio porque la ecuación $x - 2y = 0$ tiene solución única, lo que garantiza que el conjunto sea cerrado.",
            r"El determinante de la matriz $[1\;{-2}]$ es no nulo, condición suficiente para que el conjunto sea subespacio.",
            r"Toda ecuación de la forma $ax + by = 0$ define un subespacio sin necesidad de verificar cerradura, pues la linealidad la garantiza automáticamente."
        ],
        "distractores_falso": [
            r"No es subespacio porque el coeficiente $-2$ de $y$ rompe la cerradura bajo producto escalar cuando el escalar es negativo.",
            r"Falla la cerradura bajo suma: al sumar $(2,1)+(4,2)$ obtenemos $(6,3)$, cuya norma es distinta, lo que viola la condición de subespacio.",
            r"No contiene al vector nulo, ya que al sustituir $(0,0)$ la ecuación $0-0=0$ no tiene sentido algebraico en el campo real.",
            r"Representa una recta afín, no un subespacio, porque en $\mathbb{R}^2$ solo los ejes coordenados son subespacios."
        ]
    },
    {
        "id": 2,
        "tema": "Independencia Lineal",
        "enunciado": r"Los vectores $v_1 = (1, 1)$ y $v_2 = (2, 2)$ son linealmente independientes en $\mathbb{R}^2$.",
        "real_truth": "Falso",
        "just_correcta": r"Son linealmente dependientes: $v_2 = 2v_1$, por lo que existe una combinación no trivial $2v_1 - v_2 = \mathbf{0}$. El determinante de la matriz formada por ellos es $1\cdot2 - 1\cdot2 = 0$.",
        "distractores_falso": [
            r"Son dependientes porque la suma de sus componentes ($1+1=2$ y $2+2=4$) guarda una relación de proporcionalidad entera, condición suficiente de dependencia.",
            r"La dependencia se confirma porque ambos vectores pertenecen al primer cuadrante y los vectores LD siempre comparten el mismo cuadrante.",
            r"Son dependientes ya que ninguno de los dos es un vector unitario, requisito indispensable para la independencia lineal en $\mathbb{R}^2$."
        ],
        "distractores_verdad": [
            r"Son independientes porque sus componentes son todas positivas y no existe un escalar real negativo que convierta a uno en el otro.",
            r"La independencia se verifica calculando $v_1 \cdot v_2 = 1\cdot2 + 1\cdot2 = 4 \neq 0$: producto punto no nulo implica independencia lineal.",
            r"Son LI porque tienen normas distintas: $\|v_1\|=\sqrt{2}$ y $\|v_2\|=2\sqrt{2}$, y vectores con normas distintas no pueden ser múltiplos entre sí.",
            r"Forman una base porque al apilarlos en una matriz $2\times2$ se obtiene una estructura triangular con diagonal positiva."
        ]
    },
    {
        "id": 3,
        "tema": "Espacio Generado",
        "enunciado": r"El conjunto $S = \{(1, 0),\,(0, 1),\,(1, 1)\}$ genera a todo $\mathbb{R}^2$.",
        "real_truth": "Verdadero",
        "just_correcta": r"El subconjunto $\{(1,0),(0,1)\} \subset S$ ya es base de $\mathbb{R}^2$, luego $\text{span}(S)=\mathbb{R}^2$. El tercer vector es redundante pero no impide la generación.",
        "distractores_verdad": [
            r"Genera $\mathbb{R}^2$ porque el conjunto tiene 3 vectores y $3 > \dim(\mathbb{R}^2)=2$; siempre que haya más vectores que la dimensión, el espacio queda cubierto.",
            r"Todo conjunto que incluya al menos un vector con primera componente no nula y otro con segunda componente no nula genera $\mathbb{R}^2$ por el teorema de extensión.",
            r"Genera el plano completo porque la suma $(1,0)+(0,1)+(1,1)=(2,2)$ tiene ambas componentes iguales, lo que garantiza isotropía en todas las direcciones."
        ],
        "distractores_falso": [
            r"No genera $\mathbb{R}^2$ porque $(1,1) = (1,0)+(0,1)$ es combinación lineal de los otros dos, lo que reduce el rango efectivo a 1 y colapsa el espacio generado a una recta.",
            r"El espacio generado es solo el primer cuadrante, ya que ninguno de los vectores tiene componentes negativas.",
            r"No puede generar $\mathbb{R}^2$ porque un conjunto generador debe ser linealmente independiente; al contener dependencia, el span se degrada.",
            r"Falla la generación porque la matriz $2\times3$ formada por los vectores no es cuadrada y por tanto no tiene inversa, condición requerida para generación."
        ]
    },
    {
        "id": 4,
        "tema": "Combinación Lineal",
        "enunciado": r"El vector $(5, 3)$ puede expresarse como combinación lineal de $(1, 0)$ y $(0, 1)$.",
        "real_truth": "Verdadero",
        "just_correcta": r"Basta tomar $c_1=5,\,c_2=3$: $\;5(1,0)+3(0,1)=(5,0)+(0,3)=(5,3)$. Como $\{(1,0),(0,1)\}$ es base de $\mathbb{R}^2$, todo vector admite representación única.",
        "distractores_verdad": [
            r"Es posible porque $(5,3)$ tiene norma $\sqrt{34}>1$, y los vectores generadores pueden escalarse a cualquier magnitud mayor que la unidad.",
            r"Se puede porque el producto punto $(5,3)\cdot(1,0)=5$ y $(5,3)\cdot(0,1)=3$ son enteros positivos, condición suficiente para la existencia de la combinación.",
            r"Todo vector en el primer cuadrante es combinación lineal de los vectores canónicos con coeficientes positivos, lo cual se cumple aquí."
        ],
        "distractores_falso": [
            r"No es posible porque los vectores $(1,0)$ y $(0,1)$ son unitarios y solo pueden generar vectores de norma menor o igual a $\sqrt{2}$.",
            r"La combinación lineal requiere que los escalares satisfagan $c_1^2+c_2^2=1$; como $25+9=34\neq1$, la representación no es válida.",
            r"No puede expresarse así porque $(5,3)$ no es ortogonal a ninguno de los dos vectores generadores, requisito previo para la descomposición.",
            r"Falla porque la relación entre componentes $5/3$ es irracional en sentido proyectivo, impidiendo una representación exacta con base entera."
        ]
    },
    {
        "id": 5,
        "tema": "Dimensión",
        "enunciado": r"La dimensión de un subespacio que consiste únicamente en el vector nulo $\{\mathbf{0}\}$ es 1.",
        "real_truth": "Falso",
        "just_correcta": r"La dimensión del subespacio trivial $\{\mathbf{0}\}$ es $0$. Su base es el conjunto vacío $\emptyset$ (cardinalidad 0), pues el vector nulo no puede integrar ninguna base por ser LD consigo mismo.",
        "distractores_falso": [
            r"La dimensión es 0 porque el vector nulo carece de dirección y por tanto no puede contar como generador de ningún subespacio con estructura geométrica.",
            r"Es de dimensión 0 porque $\|\mathbf{0}\|=0$, y la dimensión de un subespacio siempre coincide con la norma de su generador más pequeño.",
            r"Tiene dimensión $-1$ según la convención del álgebra lineal abstracta, donde el subespacio vacío precede al trivial en la jerarquía dimensional."
        ],
        "distractores_verdad": [
            r"La dimensión es 1 porque el conjunto contiene exactamente un vector, y la dimensión se define como el número de elementos del conjunto.",
            r"Es dimensión 1 ya que $\mathbf{0} = 0\cdot\mathbf{0}$ muestra que el vector se genera a sí mismo con escalar 0, constituyendo una base de un elemento.",
            r"La dimensión es 1 porque $\{\mathbf{0}\}$ está contenido en $\mathbb{R}^2$ (dimensión 2) y todo subespacio propio reduce la dimensión en exactamente 1.",
            r"Vale 1 porque el origen es el punto de intersección de los dos ejes coordenados, y dos generadores ortogonales colapsados dan dimensión $2-1=1$."
        ]
    },
    {
        "id": 6,
        "tema": "Base",
        "enunciado": r"Cualquier conjunto de dos vectores linealmente independientes en $\mathbb{R}^2$ es una base para $\mathbb{R}^2$.",
        "real_truth": "Verdadero",
        "just_correcta": r"En $\mathbb{R}^2$ (dimensión 2), todo conjunto de exactamente $n=2$ vectores LI es automáticamente generador del espacio, por el teorema que caracteriza bases en espacios de dimensión finita.",
        "distractores_verdad": [
            r"Es base porque la suma de los dos vectores LI siempre produce un tercer vector con norma máxima, garantizando la cobertura de todas las direcciones.",
            r"Dos vectores LI forman una base si y solo si el ángulo entre ellos es mayor a $45°$; como la independencia implica no paralelismo, esta condición angular siempre se cumple.",
            r"Constituyen una base porque al ser LI en $\mathbb{R}^2$, su determinante es no nulo, lo que por el teorema espectral implica generación completa del espacio."
        ],
        "distractores_falso": [
            r"No basta con ser LI: también deben ser ortonormales. Solo las bases ortonormales (como la canónica) son legítimas para $\mathbb{R}^2$.",
            r"La independencia garantiza no redundancia pero no cobertura; se requiere verificar adicionalmente que su span contenga al vector $(1,1)$.",
            r"Dos vectores LI solo forman base si al menos uno pertenece a la base canónica estándar $\{e_1, e_2\}$.",
            r"Se necesitan al menos 3 vectores para garantizar que los 4 cuadrantes del plano queden cubiertos; con 2 solo se cubre media plaza."
        ]
    },
    {
        "id": 7,
        "tema": "Subespacios en ℝ²",
        "enunciado": r"El conjunto $V = \{(x, y) \in \mathbb{R}^2 \mid x + y = 1\}$ es un subespacio vectorial de $\mathbb{R}^2$.",
        "real_truth": "Falso",
        "just_correcta": r"Falla la condición del vector nulo: $0+0=0\neq1$, luego $\mathbf{0}\notin V$. Además, $(1,0),(0,1)\in V$ pero $(1,0)+(0,1)=(1,1)$ satisface $1+1=2\neq1$, violando la cerradura bajo suma.",
        "distractores_falso": [
            r"No es subespacio porque la recta $x+y=1$ tiene pendiente negativa $(-1)$, y los subespacios de $\mathbb{R}^2$ solo pueden tener pendientes positivas o ser ejes.",
            r"Falla la cerradura escalar porque al multiplicar cualquier vector de $V$ por un escalar $c\neq1$ se obtiene $cx+cy=c\neq1$, saliendo del conjunto.",
            r"No es subespacio porque la ecuación tiene término independiente (el 1), y solo las ecuaciones con término independiente nulo pueden definir subespacios."
        ],
        "distractores_verdad": [
            r"Es subespacio porque $x+y=1$ es una ecuación lineal en $\mathbb{R}^2$, y toda ecuación lineal define un subespacio por el teorema de representación de Riesz.",
            r"Cumple la cerradura escalar: si $(x,y)\in V$ entonces $c(x,y)=(cx,cy)$ satisface $cx+cy=c(x+y)=c\cdot1=c$, que sigue siendo una constante.",
            r"Es un subespacio afín, categoría que en geometría diferencial es equivalente a subespacio vectorial para dimensiones menores que la del espacio ambiente.",
            r"El conjunto es subespacio porque contiene infinitos vectores (toda la recta), y los conjuntos infinitos en $\mathbb{R}^2$ siempre forman subespacios."
        ]
    },
    {
        "id": 8,
        "tema": "Combinación Lineal",
        "enunciado": r"Si $u$ y $v$ son vectores en $\mathbb{R}^2$, entonces $3u - 2v$ es una combinación lineal de $\{u, v\}$.",
        "real_truth": "Verdadero",
        "just_correcta": r"Por definición, $w$ es CL de $\{u,v\}$ si existen $c_1,c_2\in\mathbb{R}$ tales que $w=c_1u+c_2v$. Aquí $c_1=3$ y $c_2=-2$, ambos reales, así que $3u-2v$ cumple la definición.",
        "distractores_verdad": [
            r"Es combinación lineal porque $3$ y $-2$ son números enteros, y en $\mathbb{R}^2$ solo se permiten escalares enteros para que la combinación permanezca en el retículo del espacio.",
            r"Se verifica porque $|3|+|-2|=5>1$, condición que asegura que la combinación alcanza vectores fuera del círculo unitario y cubre todo el plano.",
            r"Es válida porque $3$ y $2$ son primos relativos, lo que garantiza por el lema de Bézout que la combinación puede representar cualquier vector entero del plano."
        ],
        "distractores_falso": [
            r"No es combinación lineal porque el coeficiente $-2$ es negativo; los escalares en una combinación lineal deben ser no negativos para preservar la orientación.",
            r"Falso: la operación $3u-2v$ es una diferencia vectorial ponderada, que no coincide con la definición formal de combinación lineal, la cual solo admite sumas.",
            r"Solo es combinación lineal si $u$ y $v$ son LI; si fueran dependientes, $3u-2v$ podría ser el vector nulo, que no se considera combinación lineal propia.",
            r"No cumple la definición porque los coeficientes deben sumar 1 (combinación convexa); $3+(-2)=1$ es una coincidencia y no garantiza validez en general."
        ]
    },
    {
        "id": 9,
        "tema": "Independencia Lineal",
        "enunciado": r"El conjunto $\{(1, 2),\,(0, 0)\}$ es linealmente independiente.",
        "real_truth": "Falso",
        "just_correcta": r"Todo conjunto que contiene al vector nulo es LD: la ecuación $0\cdot(1,2)+c\cdot(0,0)=(0,0)$ se satisface con $c$ arbitrario no nulo, contradiciendo la definición de independencia lineal.",
        "distractores_falso": [
            r"Es dependiente porque el vector $(0,0)$ tiene norma cero, y en todo conjunto LI cada vector debe tener norma estrictamente mayor que los anteriores en orden lexicográfico.",
            r"La dependencia se debe a que $(1,2)$ y $(0,0)$ comparten la misma segunda componente módulo 2, condición algebraica que fuerza la dependencia lineal.",
            r"Son dependientes porque el determinante de la matriz $\begin{pmatrix}1&0\\2&0\end{pmatrix}$ es $0$, lo que indica que la primera columna es múltiplo de la segunda."
        ],
        "distractores_verdad": [
            r"Es LI porque $(0,0)$ no puede obtenerse multiplicando $(1,2)$ por ningún escalar real distinto de cero; no existe $c\neq0$ tal que $c(1,2)=(0,0)$.",
            r"Los vectores son LI ya que son geométricamente distintos: $(1,2)$ es un punto interior y $(0,0)$ es el origen; ubicaciones diferentes implican independencia.",
            r"Son independientes porque ninguno es un múltiplo escalar positivo del otro, y la independencia lineal requiere ausencia de proporcionalidad directa con escalares positivos.",
            r"El conjunto es LI porque tiene solo 2 vectores y $2 < \dim(\mathbb{R}^2)+1 = 3$; todo conjunto con cardinalidad menor que $n+1$ es automáticamente independiente en $\mathbb{R}^n$."
        ]
    },
    {
        "id": 10,
        "tema": "Dimensión",
        "enunciado": r"Si un subespacio de $\mathbb{R}^2$ es generado por un solo vector no nulo, su dimensión es 1.",
        "real_truth": "Verdadero",
        "just_correcta": r"Un vector no nulo $v$ es LI por sí solo (la única solución de $cv=\mathbf{0}$ es $c=0$), por lo que $\{v\}$ es base del subespacio $\text{span}(v)$. Una base de 1 elemento implica dimensión 1.",
        "distractores_verdad": [
            r"La dimensión es 1 porque el vector genera una recta, y las rectas en $\mathbb{R}^2$ tienen dimensión 1 al ser objetos unidimensionales en el sentido topológico.",
            r"Es dimensión 1 porque el subespacio generado tiene exactamente dos semirrectas opuestas, lo que suma una 'dirección completa' equivalente a dimensión $1/2 + 1/2 = 1$.",
            r"Vale 1 por el teorema rango-nulidad: el rango de la aplicación $t\mapsto tv$ es 1, luego dim(imagen)=1 automáticamente para cualquier vector no nulo en el dominio."
        ],
        "distractores_falso": [
            r"La dimensión es 2 porque todo vector no nulo en $\mathbb{R}^2$ tiene dos componentes $(x,y)$, y la dimensión del subespacio generado coincide con el número de componentes no nulas.",
            r"Depende del vector: si $\|v\|>1$ la dimensión es 1, pero si $\|v\|\leq1$ el subespacio generado es tan pequeño que su dimensión es 0 por convención.",
            r"La dimensión es 2 porque al generar el subespacio se toman todos los múltiplos $tv$ con $t\in\mathbb{R}$, y el parámetro $t$ vive en $\mathbb{R}^1$, sumando una dimensión extra al vector.",
            r"No tiene dimensión definida si el vector no pertenece a la base canónica, pues la dimensión solo está bien definida en coordenadas estándar."
        ]
    }
]


# ═══════════════════════════════════════════════════════════════
#  ESTILOS
# ═══════════════════════════════════════════════════════════════
CSS = """
<style>
@import url('https://fonts.googleapis.com/css2?family=Playfair+Display:wght@700&family=Source+Serif+4:ital,opsz,wght@0,8..60,400;0,8..60,600;1,8..60,400&display=swap');
:root{
  --ink:#1a1a2e; --paper:#f5f0e8; --cream:#faf7f2;
  --gold:#c9a84c; --gold2:#e8c96a;
  --rust:#8b3a1e; --sage:#2d6040;
  --border:#d4c9b0; --sh:rgba(26,26,46,.09);
}
.qs{
  font-family:'Source Serif 4',Georgia,serif;
  background:var(--cream); border:2px solid var(--border); border-radius:4px;
  box-shadow:5px 5px 0 var(--border),0 2px 18px var(--sh);
  max-width:800px; margin:12px auto; overflow:hidden;
}
.qh{
  background:var(--ink); padding:16px 26px 12px;
  border-bottom:3px solid var(--gold);
}
.qh h1{font-family:'Playfair Display',serif;font-size:1.4rem;margin:0 0 2px;color:var(--gold2);}
.qh p{font-size:.8rem;color:#9a8fa0;font-style:italic;margin:0;}
.pw{
  background:var(--paper); padding:8px 26px;
  border-bottom:1px solid var(--border);
  display:flex; align-items:center; gap:12px;
}
.pt{flex:1;height:5px;background:var(--border);border-radius:3px;overflow:hidden;}
.pf{height:100%;background:linear-gradient(90deg,var(--gold),var(--gold2));border-radius:3px;}
.pl{font-size:.74rem;color:#7a7060;white-space:nowrap;font-style:italic;}
.qb{padding:22px 26px 18px;}
.qtag{display:inline-block;background:var(--ink);color:var(--gold);font-size:.68rem;
  letter-spacing:1.5px;text-transform:uppercase;padding:3px 8px;border-radius:2px;margin-bottom:10px;}
.qn{font-family:'Playfair Display',serif;font-size:2.4rem;color:var(--border);
  float:right;line-height:1;margin-top:-2px;user-select:none;}
.qstmt{background:white;border-left:4px solid var(--gold);padding:13px 16px;
  border-radius:0 4px 4px 0;font-size:1.04rem;color:var(--ink);
  margin-bottom:18px;line-height:1.8;box-shadow:2px 2px 6px var(--sh);}
.slbl{font-size:.71rem;letter-spacing:1.8px;text-transform:uppercase;
  color:#8a7f6a;margin-bottom:8px;padding-bottom:3px;border-bottom:1px solid var(--border);}

/* ── Opción de justificación ── */
.jopt{
  display:flex; align-items:flex-start; gap:14px;
  padding:12px 15px; border:1.5px solid var(--border); border-radius:4px;
  margin-bottom:8px; background:white;
  font-size:.93rem; line-height:1.75; color:var(--ink);
}
.jopt.sel{
  border-color:var(--gold); background:#fffdf5;
  box-shadow:0 0 0 2.5px var(--gold);
}
/* Letra A/B/C/D */
.jletter{
  min-width:28px; height:28px; border-radius:50%;
  background:var(--border); color:var(--ink);
  display:flex; align-items:center; justify-content:center;
  font-weight:700; font-size:.85rem; flex-shrink:0; margin-top:1px;
  font-family:'Playfair Display',serif;
  transition:background .15s, color .15s;
}
.jopt.sel .jletter{background:var(--gold);color:var(--ink);}

/* Botón selector (invisible sobre la opción, no: botón aparte abajo) */
.jsel-btn{
  width:100%; border:none; background:transparent;
  text-align:left; cursor:pointer; padding:0; margin:0;
  font-size:.8rem; color:#8a7f6a;
  border-bottom: 1px dashed var(--border);
  padding:3px 0 6px 0;
  margin-bottom:6px;
  display:block;
}
.jsel-btn:hover{color:var(--gold);}

/* Resultados */
.rh{background:var(--ink);padding:18px 26px 14px;border-bottom:3px solid var(--gold);text-align:center;}
.rh h2{font-family:'Playfair Display',serif;color:var(--gold2);font-size:1.5rem;margin:0 0 3px;}
.rh p{color:#9a8fa0;font-size:.82rem;font-style:italic;margin:0;}
.sa{text-align:center;padding:22px 26px 6px;}
.sn{font-family:'Playfair Display',serif;font-size:3.8rem;color:var(--gold);line-height:1;}
.sp{font-size:.83rem;color:#7a7060;margin-top:4px;font-style:italic;}
.sm{font-family:'Playfair Display',serif;font-size:1.15rem;color:var(--ink);margin-top:5px;}
.rdiv{border:none;border-top:1px solid var(--border);margin:14px 24px;}
.rc{margin:0 18px 12px;border:1.5px solid var(--border);border-radius:4px;overflow:hidden;background:white;}
.rch{display:flex;align-items:center;gap:9px;padding:9px 13px;background:var(--paper);border-bottom:1px solid var(--border);}
.rct{font-weight:600;font-size:.85rem;color:var(--ink);flex:1;}
.rcb{font-size:.66rem;padding:2px 6px;border-radius:2px;text-transform:uppercase;letter-spacing:1px;font-weight:600;}
.bok{background:#d4edda;color:#155724;}
.bvf{background:#f8d7da;color:#721c24;}
.bjf{background:#fff3cd;color:#856404;}
.rcbd{padding:10px 13px;font-size:.84rem;line-height:1.65;color:#3a3430;}
.rr{margin-bottom:4px;}
.rr b{color:var(--ink);}
.ok{color:var(--sage);font-weight:600;}
.bad{color:var(--rust);text-decoration:line-through;opacity:.8;}
.rf{padding:12px 26px 22px;text-align:center;}
</style>
"""


# ═══════════════════════════════════════════════════════════════
#  QUIZ  —  arquitectura definitiva para NotebookTeach / Jupyter
#
#  Problema raíz de los RadioButtons:
#    ipywidgets.RadioButtons renderiza texto plano; el LaTeX
#    de MathJax NO se procesa dentro del widget.
#
#  Solución adoptada:
#    • Cada opción se muestra en un widgets.Output (renderiza HTML+LaTeX)
#    • El "clic" se captura con un widgets.Button etiquetado con
#      una letra (A / B / C / D) justo debajo de cada Output.
#    • Al seleccionar, solo se re-renderiza el Output correspondiente
#      (cambia la clase CSS de .jopt a .jopt.sel).
#    • Todo comunica en Python puro — sin JS que cruce iframes.
# ═══════════════════════════════════════════════════════════════
LETRAS = ['A', 'B', 'C', 'D']

class AlgebraQuiz:

    def __init__(self, data):
        self.data       = data
        self.total      = len(data)
        self.idx        = 0
        self.answers    = []
        self._user_vf   = None
        self._just_opts = []
        self._sel_idx   = None
        self._opt_outs  = []

        self.root = widgets.Output()
        display(HTML(CSS))
        display(self.root)
        self._phase1()

    # ── helpers ──────────────────────────────────────────────────

    def _pct(self):
        return int(self.idx / self.total * 100)

    def _header_html(self, sub="Valor de Verdad &amp; Justificación Técnica"):
        return f"""
        <div class="qh">
          <h1>Álgebra Lineal — ℝ²</h1>
          <p>{sub}</p>
        </div>
        <div class="pw">
          <div class="pt"><div class="pf" style="width:{self._pct()}%"></div></div>
          <span class="pl">Pregunta {self.idx+1} de {self.total}</span>
        </div>"""

    def _stmt_html(self, q):
        return f"""
        <span class="qn">#{self.idx+1:02d}</span>
        <div class="qtag">{q['tema']}</div>
        <div class="qstmt">{q['enunciado']}</div>"""

    def _opt_card(self, i, text, selected=False):
        cls = 'jopt sel' if selected else 'jopt'
        return f"""
        <div class="{cls}">
          <div class="jletter">{LETRAS[i]}</div>
          <div style="flex:1">{text}</div>
        </div>"""

    # ── Phase 1 ───────────────────────────────────────────────────

    def _phase1(self):
        self._user_vf = None
        self._sel_idx = None
        q = self.data[self.idx]

        btn_v = widgets.Button(
            description='✓  Verdadero',
            layout=widgets.Layout(flex='1', height='46px'),
            style=dict(button_color='#f5f0e8', font_weight='bold', font_size='15px')
        )
        btn_f = widgets.Button(
            description='✗  Falso',
            layout=widgets.Layout(flex='1', height='46px'),
            style=dict(button_color='#f5f0e8', font_weight='bold', font_size='15px')
        )

        def on_v(b):
            btn_v.style.button_color = '#1a1a2e'
            btn_v.disabled = btn_f.disabled = True
            self._on_vf('Verdadero')

        def on_f(b):
            btn_f.style.button_color = '#3d1f1f'
            btn_v.disabled = btn_f.disabled = True
            self._on_vf('Falso')

        btn_v.on_click(on_v)
        btn_f.on_click(on_f)

        with self.root:
            clear_output(wait=True)
            display(HTML(f"""
            <div class="qs">
              {self._header_html()}
              <div class="qb">
                {self._stmt_html(q)}
                <div class="slbl">¿El enunciado es verdadero o falso?</div>
              </div>
            </div>"""))
            display(widgets.HBox(
                [btn_v, btn_f],
                layout=widgets.Layout(gap='10px', margin='4px 0 0 0')
            ))

    # ── construye opciones ────────────────────────────────────────

    def _on_vf(self, vf):
        self._user_vf = vf
        q = self.data[self.idx]
        if vf == q['real_truth']:
            pool = q['distractores_verdad'] if q['real_truth'] == 'Verdadero' else q['distractores_falso']
            opts = [q['just_correcta']] + random.sample(pool, min(3, len(pool)))
        else:
            pool = q['distractores_verdad'] if vf == 'Verdadero' else q['distractores_falso']
            opts = random.sample(pool, min(4, len(pool)))
        random.shuffle(opts)
        self._just_opts = opts
        self._phase2(q, vf, opts)

    # ── Phase 2  ──────────────────────────────────────────────────
    #
    #  Layout por opción:
    #    [ Output con HTML+LaTeX  ]   ← renderiza la tarjeta bonita
    #    [ Button  «Seleccionar A»]   ← captura el clic en Python
    #
    #  Al hacer clic en el Button:
    #    1. self._sel_idx = i
    #    2. Re-renderiza todos los Output (actualiza clase .sel)
    #    3. Habilita botón «Siguiente»

    def _phase2(self, q, vf, opts):
        n = len(opts)
        self._sel_idx  = None
        self._opt_outs = [widgets.Output() for _ in range(n)]

        vf_bg  = '#1a1a2e' if vf == 'Verdadero' else '#3d1f1f'
        vf_fg  = '#e8c96a' if vf == 'Verdadero' else '#f0b090'

        btn_sig = widgets.Button(
            description='Siguiente →',
            disabled=True,
            layout=widgets.Layout(width='160px', height='40px'),
            style=dict(button_color='#1a1a2e', font_weight='bold', font_size='14px')
        )

        # ── render de UN Output ───────────────────────────────────
        def render_opt(i):
            with self._opt_outs[i]:
                clear_output(wait=True)
                display(HTML(self._opt_card(i, opts[i], selected=(i == self._sel_idx))))

        def render_all():
            for i in range(n):
                render_opt(i)

        render_all()    # render inicial

        # ── botones de selección ──────────────────────────────────
        sel_btns = []
        for i in range(n):
            def make_cb(idx):
                def cb(b):
                    self._sel_idx = idx
                    render_all()
                    btn_sig.disabled = False
                return cb
            sb = widgets.Button(
                description=f'  ← Seleccionar {LETRAS[i]}',
                layout=widgets.Layout(
                    width='180px', height='26px', margin='1px 0 10px 2px'
                ),
                style=dict(button_color='#ede8de', font_size='11px')
            )
            sb.on_click(make_cb(i))
            sel_btns.append(sb)

        # ── submit ────────────────────────────────────────────────
        def on_submit(b):
            if self._sel_idx is None:
                return
            self.answers.append({
                'q':         q,
                'user_vf':   self._user_vf,
                'user_just': self._just_opts[self._sel_idx],
            })
            self.idx += 1
            if self.idx >= self.total:
                self._results()
            else:
                self._phase1()

        btn_sig.on_click(on_submit)

        # ── display ───────────────────────────────────────────────
        with self.root:
            clear_output(wait=True)
            display(HTML(f"""
            <div class="qs">
              {self._header_html()}
              <div class="qb">
                {self._stmt_html(q)}
                <span style="background:{vf_bg};color:{vf_fg};display:inline-block;
                  padding:3px 13px;border-radius:3px;font-size:.81rem;font-weight:600;
                  letter-spacing:.4px;margin-bottom:16px;">
                  Tu respuesta: {vf}
                </span>
                <div class="slbl">Selecciona la justificación técnica correcta:</div>
              </div>
            </div>"""))

            for i in range(n):
                display(self._opt_outs[i])   # tarjeta HTML+LaTeX
                display(sel_btns[i])         # botón de selección

            display(widgets.HBox(
                [btn_sig],
                layout=widgets.Layout(justify_content='flex-end', margin='10px 0 4px 0')
            ))

    # ── Results ───────────────────────────────────────────────────

    def _results(self):
        ok_vf   = sum(1 for a in self.answers if a['user_vf']   == a['q']['real_truth'])
        ok_both = sum(1 for a in self.answers
                      if a['user_vf']   == a['q']['real_truth']
                      and a['user_just'] == a['q']['just_correcta'])
        pct = int(ok_both / self.total * 100)
        medal, msg = (
            ('🥇', 'Dominio sobresaliente')  if pct >= 90 else
            ('🥈', 'Buen desempeño')         if pct >= 70 else
            ('🥉', 'Desempeño aceptable')    if pct >= 50 else
            ('📖', 'Revisa los conceptos')
        )

        cards = ''
        for i, a in enumerate(self.answers):
            q      = a['q']
            vf_ok  = a['user_vf']   == q['real_truth']
            jst_ok = a['user_just'] == q['just_correcta']
            if vf_ok and jst_ok:
                icon, bc, bt = '[OK]', 'bok', 'Correcto'
            elif not vf_ok:
                icon, bc, bt = '[X]', 'bvf', 'V/F incorrecto'
            else:
                icon, bc, bt = '⚠️', 'bjf', 'Justificación incorrecta'
            vf_h = (
                f'<span class="ok">{a["user_vf"]}</span>' if vf_ok else
                f'<span class="bad">{a["user_vf"]}</span> → <span class="ok">{q["real_truth"]}</span>'
            )
            jh = (
                f'<span class="ok">{a["user_just"]}</span>' if jst_ok else
                f'<span class="bad">{a["user_just"]}</span>'
            )
            cr = ('' if jst_ok else
                  f'<div class="rr"><b>Justificación correcta:</b> <span class="ok">{q["just_correcta"]}</span></div>')
            cards += f"""
            <div class="rc">
              <div class="rch">
                <span>{icon}</span>
                <span class="rct">P{i+1} — {q['tema']}</span>
                <span class="rcb {bc}">{bt}</span>
              </div>
              <div class="rcbd">
                <div class="rr" style="color:#666;font-style:italic;margin-bottom:7px;">{q['enunciado']}</div>
                <div class="rr"><b>Tu V/F:</b> {vf_h}</div>
                <div class="rr"><b>Tu justificación:</b> {jh}</div>
                {cr}
              </div>
            </div>"""

        btn_r = widgets.Button(
            description='↺  Reintentar Quiz',
            layout=widgets.Layout(width='200px', height='42px'),
            style=dict(button_color='#c9a84c', font_weight='bold', font_size='14px')
        )
        btn_r.on_click(lambda b: self.__init__(self.data))

        with self.root:
            clear_output(wait=True)
            display(HTML(f"""
            <div class="qs">
              <div class="rh">
                <h2>Resultados Finales</h2>
                <p>Álgebra Lineal — ℝ²</p>
              </div>
              <div class="sa">
                <div class="sn">{medal} {ok_both}/{self.total}</div>
                <div class="sp">{pct}% completamente correctas &nbsp;·&nbsp; V/F correcto: {ok_vf}/{self.total}</div>
                <div class="sm">{msg}</div>
              </div>
              <hr class="rdiv">
              {cards}
              <div class="rf"></div>
            </div>"""))
            display(widgets.HBox(
                [btn_r],
                layout=widgets.Layout(justify_content='center', margin='0 0 14px 0')
            ))


# ═══════════════════════════════════════════════════════════════
#  INICIAR
# ═══════════════════════════════════════════════════════════════
quiz = AlgebraQuiz(preguntas_db)
```
