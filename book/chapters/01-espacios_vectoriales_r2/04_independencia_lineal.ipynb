# 1.4 Independencia Lineal

La **independencia lineal** es uno de los conceptos más fundamentales del álgebra lineal. Nos permite determinar si un conjunto de vectores es "esencialmente diferente" o si algunos de ellos son redundantes.

¿Por qué es importante?

- **Bases**: Solo conjuntos linealmente independientes pueden formar una base de un espacio vectorial.
- **Sistemas de ecuaciones**: La independencia lineal determina si un sistema tiene solución única.
- **Dimensión**: El número máximo de vectores linealmente independientes en un espacio define su dimensión.
- **Aplicaciones**: En física, ingeniería y ciencias de datos, nos permite identificar información redundante.

En esta sección estudiaremos la independencia lineal en $\mathbb{R}^2$, desarrollando la intuición geométrica y las herramientas algebraicas necesarias.

**Motivación:** Cuando trabajamos con varios vectores, surge una pregunta clave: ¿todos aportan información nueva o alguno es redundante? Verificar ejemplos numéricos puntuales ayuda, pero no alcanza para decidir en general. Necesitamos un criterio matemático que distinga entre conjuntos que realmente "abren" el espacio y conjuntos donde un vector puede escribirse con los otros. Esa necesidad introduce la **independencia lineal** como herramienta central.

## 1.4.1. Definición

:::{admonition} Definición: Independencia Lineal
:class: tip

Un conjunto de vectores $\{\mathbf{v}_1, \mathbf{v}_2\}$ en $\mathbb{R}^2$ es **linealmente independiente** si la ecuación:

$$c_1\mathbf{v}_1 + c_2\mathbf{v}_2 = \mathbf{0}$$

tiene como **única solución** la **solución trivial**: $c_1 = 0$ y $c_2 = 0$.
:::

### ¿Qué significa "solo la solución trivial"?

La ecuación $c_1\mathbf{v}_1 + c_2\mathbf{v}_2 = \mathbf{0}$ siempre tiene al menos una solución: $c_1 = c_2 = 0$. Esta se llama la **solución trivial** porque es "obvia" — cualquier combinación con coeficientes cero da el vector cero.

La pregunta clave es: *¿existe alguna otra forma de combinar los vectores para obtener el vector cero?*

- **Si NO existe otra forma** → los vectores son **linealmente independientes**
- **Si SÍ existe otra forma** → los vectores son **linealmente dependientes**

:::{admonition} Definición: Dependencia Lineal
:class: tip

Un conjunto de vectores $\{\mathbf{v}_1, \mathbf{v}_2\}$ es **linealmente dependiente** si existe una combinación lineal no trivial que da el vector cero, es decir, existen escalares $c_1, c_2$ **no ambos cero** tales que:

$$c_1\mathbf{v}_1 + c_2\mathbf{v}_2 = \mathbf{0}$$

Esto significa que uno de los vectores puede escribirse como múltiplo escalar del otro.
:::

---

## 1.4.2. Interpretación Geométrica

En $\mathbb{R}^2$, la independencia lineal tiene una interpretación geométrica muy clara:

| Condición | Significado geométrico |
|-----------|------------------------|
| **Linealmente independientes** | Los vectores **NO** son paralelos (no colineales) |
| **Linealmente dependientes** | Los vectores **SÍ** son paralelos (uno es múltiplo escalar del otro) |

### ¿Por qué?

Si $\mathbf{v}_2 = k \cdot \mathbf{v}_1$ para algún escalar $k$, entonces:

$$k \cdot \mathbf{v}_1 + (-1) \cdot \mathbf{v}_2 = k \mathbf{v}_1 - \mathbf{v}_2 = k \mathbf{v}_1 - k \mathbf{v}_1 = \mathbf{0}$$

Los coeficientes $c_1 = k$ y $c_2 = -1$ dan una combinación no trivial que produce el vector cero.

### Visualización

Observemos la diferencia entre vectores independientes y dependientes:

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Vectores INDEPENDIENTES
ax1 = axes[0]
v1_ind = np.array([2, 1])
v2_ind = np.array([1, 3])

ax1.quiver(0, 0, v1_ind[0], v1_ind[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.02, label=r'$\mathbf{v}_1 = (2, 1)$')
ax1.quiver(0, 0, v2_ind[0], v2_ind[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.02, label=r'$\mathbf{v}_2 = (1, 3)$')

ax1.set_xlim(-1, 4)
ax1.set_ylim(-1, 4)
ax1.set_aspect('equal')
ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)
ax1.grid(True, alpha=0.3)
ax1.legend(loc='upper left')
ax1.set_title('INDEPENDIENTES\n(No paralelos)', fontsize=12, fontweight='bold', color='green')

ax1.text(2.5, -0.5, 'Pueden generar cualquier vector del plano', 
         fontsize=11, ha='center')

# Vectores DEPENDIENTES
ax2 = axes[1]
v1_dep = np.array([2, 1])
v2_dep = np.array([4, 2])  # v2 = 2 * v1

ax2.quiver(0, 0, v1_dep[0], v1_dep[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.02, label=r'$\mathbf{v}_1 = (2, 1)$')
ax2.quiver(0, 0, v2_dep[0], v2_dep[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.02, label=r'$\mathbf{v}_2 = (4, 2)$')

ax2.set_xlim(-1, 5)
ax2.set_ylim(-1, 4)
ax2.set_aspect('equal')
ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.axvline(x=0, color='k', linewidth=0.5)
ax2.grid(True, alpha=0.3)
ax2.legend(loc='upper left')
ax2.set_title('DEPENDIENTES\n(Paralelos: $\\mathbf{v}_2 = 2\\mathbf{v}_1$)', 
              fontsize=12, fontweight='bold', color='red')

ax2.text(2.5, -0.5, 'Solo generan una línea (no todo el plano)', 
         fontsize=11, ha='center')

plt.tight_layout()
plt.show()
```

**Observaciones:**
- A la izquierda, los vectores apuntan en direcciones diferentes → son independientes
- A la derecha, los vectores están sobre la misma recta → son dependientes

## 1.4.3. Criterio del Determinante

Existe un método algebraico eficiente para verificar la independencia lineal de dos vectores en $\mathbb{R}^2$: el **criterio del determinante**.

:::{admonition} Teorema: Criterio del Determinante
:class: important

Sean $\mathbf{v}_1 = (a, b)$ y $\mathbf{v}_2 = (c, d)$ dos vectores en $\mathbb{R}^2$.

Entonces, los $\mathbf{v}_1$ y $\mathbf{v}_2$ son linealmente independientes si, y sólo si, 

$$a \cdot d - b \cdot c \neq 0.$$
:::

### ¿Por qué funciona?

El determinante mide el "área con signo" del paralelogramo formado por los vectores:
- Si el área es **cero**, los vectores son paralelos (colineales) → dependientes
- Si el área es **distinta de cero**, los vectores no son paralelos → independientes

## 1.4.4. Exploración: Independencia y Paralelismo

Fijamos el vector $\mathbf{v}_1 = (2, 1)$ y rotamos $\mathbf{v}_2$ usando un ángulo $\theta$.

Observa cómo cambia el determinante y cuándo los vectores se vuelven dependientes.

```python
def explorar_independencia(theta):
    v1 = np.array([2, 1])
    # v2 es un vector de magnitud 2 rotado por theta
    v2 = 2 * np.array([np.cos(theta), np.sin(theta)])
    
    # Calcular determinante
    det = v1[0] * v2[1] - v1[1] * v2[0]
    
    # Crear figura
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Dibujar vectores
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
              color='blue', width=0.03, label=r'$\mathbf{v}_1 = (2, 1)$ (fijo)')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
              color='red', width=0.03, label=f'$\\mathbf{{v}}_2 = ({v2[0]:.2f}, {v2[1]:.2f})$')
    
    # Configurar ejes
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_aspect('equal')
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(loc='upper left')
    
    # Determinar estado
    if np.abs(det) < 0.1:
        estado = "DEPENDIENTES"
        color = 'red'
    else:
        estado = "INDEPENDIENTES"
        color = 'green'
    
    ax.set_title(f'Determinante = {det:.3f}\n{estado}', 
                 fontsize=14, fontweight='bold', color=color)
    
    # Mostrar ángulo
    ax.text(2.5, -2.5, f'θ = {np.degrees(theta):.1f}°', fontsize=11)
    
    plt.tight_layout()
    plt.show()

# Widget interactivo
interact(explorar_independencia, 
         theta=FloatSlider(value=1.0, min=0, max=2*np.pi, step=0.05, 
                          description='Ángulo θ:', continuous_update=True));
```

:::{admonition} Pregunta de reflexión
:class: hint

**¿En qué ángulos los vectores se vuelven dependientes?**

Pista: $\mathbf{v}_1 = (2, 1)$ forma un ángulo de $\arctan(1/2) \approx 26.57°$ con el eje $x$. Los vectores son dependientes cuando $\mathbf{v}_2$ apunta en la misma dirección o en la dirección opuesta.
:::

## 1.4.5. Máximo número de vectores independientes en $\mathbb{R}^2$

Una pregunta natural es: **¿cuántos vectores linealmente independientes pueden existir en $\mathbb{R}^2$?**

:::{admonition} Resultado importante
:class: important

En $\mathbb{R}^2$, **como máximo 2 vectores** pueden ser linealmente independientes.

Cualquier conjunto de 3 o más vectores en $\mathbb{R}^2$ es **siempre** linealmente dependiente.
:::

### ¿Por qué?

**Intuición geométrica:** En el plano solo hay "2 direcciones fundamentales". Una vez que tienes dos vectores no paralelos, cualquier tercer vector puede expresarse como combinación de los primeros dos.

**Argumento algebraico (principio del palomar):** Si tienes 3 vectores en $\mathbb{R}^2$, el sistema:

$$c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + c_3\mathbf{v}_3 = \mathbf{0}$$

representa 2 ecuaciones con 3 incógnitas. Un sistema con más incógnitas que ecuaciones siempre tiene soluciones no triviales.

### Conexión con la dimensión

Este número máximo de vectores independientes es precisamente la **dimensión** del espacio:

$$\dim(\mathbb{R}^2) = 2$$

Este concepto se generalizará más adelante: $\dim(\mathbb{R}^n) = n$.

## 1.4.6. Introducción a las Bases

El concepto de independencia lineal nos lleva directamente al concepto de **base**.

:::{admonition} Definición: Base de $\mathbb{R}^2$
:class: tip

Una **base** de $\mathbb{R}^2$ es un conjunto de **2 vectores linealmente independientes**.

Una base tiene dos propiedades fundamentales:
1. Los vectores son **linealmente independientes** (ninguno es redundante)
2. **Generan** todo $\mathbb{R}^2$ (cualquier vector del plano puede escribirse como combinación lineal de ellos)
:::

### La base estándar (canónica)

La base más utilizada en $\mathbb{R}^2$ es la **base estándar**:

$$\mathbf{e}_1 = (1, 0), \quad \mathbf{e}_2 = (0, 1)$$

Cualquier vector $(a, b) \in \mathbb{R}^2$ puede escribirse como:

$$(a, b) = a \cdot \mathbf{e}_1 + b \cdot \mathbf{e}_2 = a(1, 0) + b(0, 1)$$

Los coeficientes $a$ y $b$ son precisamente las **coordenadas** del vector en esta base.

## 1.4.7. Ejemplo Completo

**Problema:** Determinar si los vectores $\mathbf{v}_1 = (2, 4)$ y $\mathbf{v}_2 = (1, 2)$ son linealmente independientes.

### Método 1: Criterio del determinante

$$\det\begin{pmatrix} 2 & 1 \\ 4 & 2 \end{pmatrix} = 2 \cdot 2 - 4 \cdot 1 = 4 - 4 = 0$$

**Conclusión:** Como $\det(A) = 0$, los vectores son **linealmente dependientes**.

### Método 2: Verificación directa

Observemos que $\mathbf{v}_1 = 2 \cdot \mathbf{v}_2$:
$$(2, 4) = 2 \cdot (1, 2)$$

Esto significa que existe una combinación no trivial:
$$1 \cdot \mathbf{v}_1 + (-2) \cdot \mathbf{v}_2 = (2, 4) - 2(1, 2) = (2-2, 4-4) = (0, 0) = \mathbf{0}$$

Los coeficientes $c_1 = 1$ y $c_2 = -2$ son no triviales, confirmando la dependencia.
