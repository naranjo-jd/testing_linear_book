## 🟦 Ejercicio 1: Probar linealidad y encontrar la matriz

Sea $T: \mathbb{R}^3 \to \mathbb{R}^3$ definida por:
$$T(x, y, z) = (x+y,\; y+z,\; x+z)$$

**(a)** Demuestra que $T$ es lineal.

**(b)** Encuentra la matriz $A$ tal que $T(\mathbf{v}) = A\mathbf{v}$.

**(c)** Calcula $\det(A)$. ¿Es $T$ invertible?

---

### Solución

**(a) Demostración de linealidad:**

Sean $\mathbf{u} = (u_1, u_2, u_3)$, $\mathbf{v} = (v_1, v_2, v_3) \in \mathbb{R}^3$ y $c \in \mathbb{R}$.

**Aditividad:**
$$T(\mathbf{u}+\mathbf{v}) = T(u_1+v_1,\, u_2+v_2,\, u_3+v_3)$$
$$= ((u_1+v_1)+(u_2+v_2),\; (u_2+v_2)+(u_3+v_3),\; (u_1+v_1)+(u_3+v_3))$$
$$= (u_1+u_2,\,u_2+u_3,\,u_1+u_3) + (v_1+v_2,\,v_2+v_3,\,v_1+v_3) = T(\mathbf{u})+T(\mathbf{v}). \checkmark$$

**Homogeneidad:**
$$T(c\mathbf{u}) = T(cu_1, cu_2, cu_3) = (cu_1+cu_2,\; cu_2+cu_3,\; cu_1+cu_3)$$
$$= c(u_1+u_2,\; u_2+u_3,\; u_1+u_3) = cT(\mathbf{u}). \checkmark$$

**(b) Matriz de representación:**

$$T(\mathbf{e}_1) = T(1,0,0) = (1,0,1), \quad T(\mathbf{e}_2) = T(0,1,0) = (1,1,0), \quad T(\mathbf{e}_3) = T(0,0,1) = (0,1,1)$$

$$A = \begin{pmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \\ 1 & 0 & 1 \end{pmatrix}$$

**(c)** $\det(A) = 1(1\cdot1 - 1\cdot0) - 1(0\cdot1 - 1\cdot1) + 0 = 1 + 1 = 2 \neq 0$. **$T$ es invertible.** $\checkmark$

```python
import numpy as np

def T1(v):
    x, y, z = v
    return np.array([x+y, y+z, x+z])

# Construir A desde la base estandar
A = np.column_stack([T1(e) for e in np.eye(3)])
print("Matriz A:")
print(A)
print(f"det(A) = {np.linalg.det(A):.4f}")
print(f"Invertible: {np.abs(np.linalg.det(A)) > 1e-10}")

# Verificar linealidad
u = np.array([1.0, 2.0, -1.0])
v = np.array([3.0, 0.0,  2.0])
print(f"\nAditividad: {np.allclose(T1(u+v), T1(u)+T1(v))}")
print(f"Homogeneidad (c=3): {np.allclose(T1(3*u), 3*T1(u))}")

# Inversa
A_inv = np.linalg.inv(A)
print("\nMatriz inversa A^{-1}:")
print(np.round(A_inv, 4))
```

## 🟩 Ejercicio 2: Núcleo e imagen de una matriz $3 \times 3$

Sea $T: \mathbb{R}^3 \to \mathbb{R}^3$ con matriz:
$$B = \begin{pmatrix} 1 & 2 & 1 \\ 2 & 4 & 2 \\ 3 & 6 & 3 \end{pmatrix}$$

**(a)** Calcula $\ker(T)$ analíticamente.

**(b)** Calcula $\operatorname{Im}(T)$ analíticamente.

**(c)** Verifica el teorema rango-nulidad.

---

### Solución

Observamos que $B$ tiene todas las filas proporcionales: fila 2 = 2 × fila 1, fila 3 = 3 × fila 1. Entonces:
- $\operatorname{rango}(B) = 1$
- $\operatorname{nulidad}(B) = 3 - 1 = 2$

**(a) Núcleo:** El sistema $B\mathbf{x} = \mathbf{0}$ se reduce a $x_1 + 2x_2 + x_3 = 0$, con $x_2 = s$ y $x_3 = t$ libres:
$$\ker(T) = \operatorname{span}\left\{\begin{pmatrix}-2\\1\\0\end{pmatrix}, \begin{pmatrix}-1\\0\\1\end{pmatrix}\right\}$$

**(b) Imagen:** El espacio columna de $B$ es $\operatorname{span}\left\{\begin{pmatrix}1\\2\\3\end{pmatrix}\right\}$ (una recta en $\mathbb{R}^3$).

**(c)** Rango-nulidad: $1 + 2 = 3 = \dim(\mathbb{R}^3)$. ✓

```python
import sympy as sp
import numpy as np

B = sp.Matrix([[1, 2, 1],
               [2, 4, 2],
               [3, 6, 3]])

print("Matriz B:")
sp.pprint(B)

rango   = B.rank()
nucleo  = B.nullspace()
imagen  = B.columnspace()

print(f"\nRango = {rango}")
print(f"Nulidad = {len(nucleo)}")

print("\nBase del nucleo:")
for v in nucleo:
    sp.pprint(v.T)

print("\nBase de la imagen:")
for v in imagen:
    sp.pprint(v.T)

n_dominio = B.shape[1]
print(f"\nVerificacion rango-nulidad: {rango} + {len(nucleo)} = {rango+len(nucleo)} (debe ser {n_dominio})")
print(f"Teorema verificado: {rango + len(nucleo) == n_dominio}")
```

## 🟨 Ejercicio 3: Aplicar el teorema rango-nulidad

Sea $T: \mathbb{R}^5 \to \mathbb{R}^4$ una transformación lineal con $\operatorname{rango}(T) = 3$.

**(a)** ¿Cuánto vale $\dim(\ker T)$?

**(b)** ¿Puede $T$ ser inyectiva? ¿Por qué?

**(c)** ¿Puede $T$ ser suprayectiva? ¿Por qué?

**(d)** Construye explícitamente una matriz $4 \times 5$ de rango $3$ y verifica los cálculos.

---

### Solución

**(a)** Por rango-nulidad: $\dim(\ker T) = n - \operatorname{rango}(T) = 5 - 3 = \mathbf{2}$.

**(b)** **No puede ser inyectiva.** Para ser inyectiva necesitaría $\ker(T) = \{\mathbf{0}\}$, es decir, nulidad $= 0$. Pero hemos calculado nulidad $= 2 \neq 0$.

**(c)** **No puede ser suprayectiva.** Para serlo necesitaría $\operatorname{Im}(T) = \mathbb{R}^4$, es decir, rango $= 4$. Pero rango $= 3 < 4$.

```python
import numpy as np

# Construir una matriz 4x5 de rango 3
# Estrategia: tomar 3 vectores independientes y hacer las otras columnas combinacion lineal
col1 = np.array([1, 0, 0, 1])
col2 = np.array([0, 1, 0, 1])
col3 = np.array([0, 0, 1, 1])
col4 = col1 + col2          # combinacion lineal
col5 = col2 - col3          # combinacion lineal

C = np.column_stack([col1, col2, col3, col4, col5]).astype(float)
print("Matriz C (4x5):")
print(C)

rango   = np.linalg.matrix_rank(C)
n_dom   = C.shape[1]  # 5
n_cod   = C.shape[0]  # 4
nulidad = n_dom - rango

print(f"\nRango = {rango} (esperado: 3)")
print(f"Nulidad = {nulidad} (esperado: 2)")
print(f"Rango + Nulidad = {rango + nulidad} (debe ser {n_dom})")
print(f"\nInyectiva (nulidad=0): {nulidad == 0}")
print(f"Suprayectiva (rango={n_cod}): {rango == n_cod}")
```

## �� Ejercicio 4: La composición de transformaciones lineales es lineal

**Teorema:** Si $T_1: U \to V$ y $T_2: V \to W$ son transformaciones lineales, entonces $T_2 \circ T_1: U \to W$ también es lineal.

**Demostración:**

Sean $\mathbf{u}, \mathbf{v} \in U$ y $a, b \in \mathbb{R}$. Debemos verificar:
$$(T_2 \circ T_1)(a\mathbf{u} + b\mathbf{v}) = a(T_2 \circ T_1)(\mathbf{u}) + b(T_2 \circ T_1)(\mathbf{v})$$

**Cálculo:**
$$(T_2 \circ T_1)(a\mathbf{u} + b\mathbf{v}) = T_2(T_1(a\mathbf{u} + b\mathbf{v}))$$
$$= T_2(aT_1(\mathbf{u}) + bT_1(\mathbf{v})) \quad \text{(por linealidad de } T_1\text{)}$$
$$= aT_2(T_1(\mathbf{u})) + bT_2(T_1(\mathbf{v})) \quad \text{(por linealidad de } T_2\text{)}$$
$$= a(T_2 \circ T_1)(\mathbf{u}) + b(T_2 \circ T_1)(\mathbf{v}). \quad \checkmark$$

Matricialmente: la composición corresponde al **producto de matrices** $A_2 A_1$.

```python
import numpy as np

# Verificacion computacional del teorema
# T1: R^3 -> R^2,  T2: R^2 -> R^4
A1 = np.array([[1, 2, 0],
               [0, 1, 1]], dtype=float)   # 2x3

A2 = np.array([[1, 0],
               [2,-1],
               [0, 1],
               [1, 1]], dtype=float)       # 4x2

# Composicion: T2 o T1, matriz = A2 @ A1
A_comp = A2 @ A1  # 4x3
print(f"A1 shape: {A1.shape}")
print(f"A2 shape: {A2.shape}")
print(f"A_comp = A2 @ A1 shape: {A_comp.shape}")

# Probar linealidad de la composicion
np.random.seed(99)
N = 20
U = np.random.randn(3, N)
V = np.random.randn(3, N)
a, b = 2.3, -1.7

lhs = A_comp @ (a*U + b*V)
rhs = a*(A_comp @ U) + b*(A_comp @ V)

print(f"\nVerificacion (T2 o T1)(au+bv) = a(T2 o T1)(u) + b(T2 o T1)(v):")
print(f"Error maximo: {np.max(np.abs(lhs - rhs)):.2e}")
print(f"Linealidad verificada: {np.allclose(lhs, rhs)}")
```

## 🟪 Ejercicio 5: Inyectividad, suprayectividad y biyectividad

Para cada una de las siguientes transformaciones, determina si es inyectiva, suprayectiva y/o biyectiva. Justifica usando rango y nulidad.

**(a)** $T_a: \mathbb{R}^2 \to \mathbb{R}^2$ con $A = \begin{pmatrix}3 & 1\\ 6 & 2\end{pmatrix}$

**(b)** $T_b: \mathbb{R}^2 \to \mathbb{R}^3$ con $B = \begin{pmatrix}1 & 0\\ 0 & 1\\ 1 & 1\end{pmatrix}$

**(c)** $T_c: \mathbb{R}^3 \to \mathbb{R}^2$ con $C = \begin{pmatrix}1 & 0 & -1\\ 2 & 1 & 0\end{pmatrix}$

**(d)** $T_d: \mathbb{R}^3 \to \mathbb{R}^3$ con $D = \begin{pmatrix}1 & 0 & 1\\ 0 & 1 & 0\\ 1 & 0 & 2\end{pmatrix}$

---

### Solución esperada

| Caso | $n$ | $m$ | Rango | Nulidad | Inyectiva | Suprayectiva | Biyectiva |
|------|-----|-----|-------|---------|-----------|--------------|----------|
| (a) $A$ | 2 | 2 | 1 | 1 | ❌ | ❌ | ❌ |
| (b) $B$ | 2 | 3 | 2 | 0 | ✅ | ❌ | ❌ |
| (c) $C$ | 3 | 2 | 2 | 1 | ❌ | ✅ | ❌ |
| (d) $D$ | 3 | 3 | 3 | 0 | ✅ | ✅ | ✅ |

```python
import numpy as np

def analizar(nombre, M):
    m, n = M.shape
    r = np.linalg.matrix_rank(M)
    nul = n - r
    iny = (nul == 0)
    sur = (r == m)
    bij = iny and sur
    print(f"\n{nombre}: {m}x{n}, rango={r}, nulidad={nul}")
    print(f"  Inyectiva: {iny}  |  Suprayectiva: {sur}  |  Biyectiva: {bij}")

A = np.array([[3, 1], [6, 2]], dtype=float)
B = np.array([[1, 0], [0, 1], [1, 1]], dtype=float)
C = np.array([[1, 0, -1], [2, 1, 0]], dtype=float)
D = np.array([[1, 0, 1], [0, 1, 0], [1, 0, 2]], dtype=float)

analizar("(a) A", A)
analizar("(b) B", B)
analizar("(c) C", C)
analizar("(d) D", D)

print("\nPara (d), det(D) =", np.round(np.linalg.det(D), 4))
print("Al ser det != 0 y ser cuadrada, D es invertible -> T_d es biyectiva.")
```
