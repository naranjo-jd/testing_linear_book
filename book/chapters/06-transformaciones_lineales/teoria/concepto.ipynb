# 📐 Concepto de Transformación Lineal

Una **transformación lineal** es una función $T: V \to W$ entre dos espacios vectoriales que respeta las dos operaciones fundamentales: la suma de vectores y la multiplicación por escalar. Esta propiedad hace que las transformaciones lineales sean el objeto central del álgebra lineal, conectando la geometría con el álgebra mediante matrices.

> **En este notebook** aprenderemos a identificar, verificar y representar transformaciones lineales, y estudiaremos sus espacios fundamentales.

## 🎯 ¿Qué es una transformación lineal?

Sea $V$ y $W$ espacios vectoriales sobre $\mathbb{R}$. Una función $T: V \to W$ es una **transformación lineal** si cumple **ambas** propiedades:

| Propiedad | Condición |
|-----------|----------|
| **Aditividad** | $T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})$ para todo $\mathbf{u}, \mathbf{v} \in V$ |
| **Homogeneidad** | $T(c\mathbf{u}) = cT(\mathbf{u})$ para todo $c \in \mathbb{R}$, $\mathbf{u} \in V$ |

Estas dos condiciones se pueden **combinar** en una sola:

$$T(a\mathbf{u} + b\mathbf{v}) = aT(\mathbf{u}) + bT(\mathbf{v}) \quad \forall\, a,b \in \mathbb{R},\; \mathbf{u},\mathbf{v} \in V$$

Esta propiedad se llama **linealidad** y dice, informalmente, que $T$ *respeta las combinaciones lineales*.

### 💡 Consecuencias inmediatas

De la definición se deducen directamente:
- $T(\mathbf{0}) = \mathbf{0}$ (el origen siempre se mapea al origen)
- $T(-\mathbf{v}) = -T(\mathbf{v})$
- $T(a_1\mathbf{v}_1 + \cdots + a_k\mathbf{v}_k) = a_1T(\mathbf{v}_1) + \cdots + a_kT(\mathbf{v}_k)$

## 🔍 Verificación de linealidad

Para **demostrar** que $T$ es lineal se siguen estos pasos:

1. **Sean $\mathbf{u}, \mathbf{v} \in V$ y $c \in \mathbb{R}$ arbitrarios** (no particulares).
2. Calcular $T(\mathbf{u} + \mathbf{v})$ y comparar con $T(\mathbf{u}) + T(\mathbf{v})$.
3. Calcular $T(c\mathbf{u})$ y comparar con $cT(\mathbf{u})$.
4. Si ambas igualdades se cumplen algebraicamente, $T$ es lineal.

Para **refutar** que $T$ es lineal basta encontrar **un contraejemplo**: un par de vectores o escalar para los que falle alguna propiedad.

### Ejemplo rápido de verificación

Sea $T: \mathbb{R}^2 \to \mathbb{R}^2$ definida por $T(x, y) = (2x - y,\; x + 3y)$.

**Verificamos aditividad:** Sean $(x_1, y_1)$ y $(x_2, y_2)$.
$$T\bigl((x_1+x_2,\, y_1+y_2)\bigr) = (2(x_1+x_2)-(y_1+y_2),\; (x_1+x_2)+3(y_1+y_2))$$
$$= (2x_1-y_1,\, x_1+3y_1) + (2x_2-y_2,\, x_2+3y_2) = T(x_1,y_1)+T(x_2,y_2). \checkmark$$

**Verificamos homogeneidad:** $T(cx, cy) = (2cx - cy,\; cx + 3cy) = c(2x-y, x+3y) = cT(x,y)$. $\checkmark$

```python
import numpy as np

# Verificacion numerica de linealidad para T(x) = R_theta * x (rotacion)
theta = np.pi / 4  # 45 grados
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

# Vectores de prueba
u = np.array([1.0, 2.0])
v = np.array([-1.0, 3.0])
a, b = 2.0, -1.5

# Verificar aditividad: T(u+v) == T(u) + T(v)
lhs_add = R @ (u + v)
rhs_add = R @ u + R @ v
print("=== Verificacion de Aditividad ===")
print(f"T(u+v)      = {lhs_add}")
print(f"T(u) + T(v) = {rhs_add}")
print(f"Son iguales: {np.allclose(lhs_add, rhs_add)}")

print()
# Verificar homogeneidad: T(cu) == c*T(u)
c = 3.0
lhs_hom = R @ (c * u)
rhs_hom = c * (R @ u)
print("=== Verificacion de Homogeneidad ===")
print(f"T(c*u)  = {lhs_hom}")
print(f"c*T(u)  = {rhs_hom}")
print(f"Son iguales: {np.allclose(lhs_hom, rhs_hom)}")

print()
# Propiedad combinada: T(au + bv) == a*T(u) + b*T(v)
lhs_comb = R @ (a*u + b*v)
rhs_comb = a*(R @ u) + b*(R @ v)
print("=== Propiedad Combinada ===")
print(f"T(au+bv)         = {lhs_comb}")
print(f"a*T(u) + b*T(v)  = {rhs_comb}")
print(f"Son iguales: {np.allclose(lhs_comb, rhs_comb)}")
```

## 🗂️ Ejemplos canónicos de transformaciones en $\mathbb{R}^2$

Las siguientes son transformaciones lineales fundamentales. Cada una tiene una interpretación geométrica clara:

| Transformación | Matriz $A$ | Efecto geométrico |
|----------------|------------|-------------------|
| **Rotación** por ángulo $\theta$ | $\begin{pmatrix}\cos\theta & -\sin\theta \\ \sin\theta & \cos\theta\end{pmatrix}$ | Rota vectores en sentido antihorario |
| **Reflexión** sobre eje $x$ | $\begin{pmatrix}1 & 0 \\ 0 & -1\end{pmatrix}$ | Invierte la componente $y$ |
| **Reflexión** sobre eje $y$ | $\begin{pmatrix}-1 & 0 \\ 0 & 1\end{pmatrix}$ | Invierte la componente $x$ |
| **Escala** uniforme por $k$ | $\begin{pmatrix}k & 0 \\ 0 & k\end{pmatrix}$ | Estira o comprime uniformemente |
| **Cizallamiento** horizontal | $\begin{pmatrix}1 & k \\ 0 & 1\end{pmatrix}$ | Inclina horizontalmente |
| **Transformación cero** | $\begin{pmatrix}0 & 0 \\ 0 & 0\end{pmatrix}$ | Todo va al origen |
| **Identidad** | $\begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix}$ | No cambia nada |

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def transformar_cuadrado(A):
    """Aplica la transformacion A al cuadrado unitario y devuelve los vertices."""
    # Vertices del cuadrado unitario (columnas)
    cuadrado = np.array([[0, 1, 1, 0, 0],
                         [0, 0, 1, 1, 0]], dtype=float)
    return A @ cuadrado

theta = np.pi / 3  # 60 grados
transformaciones = [
    ("Rotacion 60°",   np.array([[np.cos(theta), -np.sin(theta)],
                                  [np.sin(theta),  np.cos(theta)]])),
    ("Reflexion eje x", np.array([[1, 0], [0, -1]], dtype=float)),
    ("Escala 1.5x",     np.array([[1.5, 0], [0, 1.5]])),
    ("Cizallamiento k=1", np.array([[1, 1], [0, 1]], dtype=float)),
]

cuadrado_orig = np.array([[0,1,1,0,0],[0,0,1,1,0]], dtype=float)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for ax, (titulo, A) in zip(axes, transformaciones):
    trans = transformar_cuadrado(A)
    ax.fill(cuadrado_orig[0], cuadrado_orig[1], alpha=0.2, color='steelblue', label='Original')
    ax.plot(cuadrado_orig[0], cuadrado_orig[1], 'b--', linewidth=1.5)
    ax.fill(trans[0], trans[1], alpha=0.4, color='tomato', label='Transformado')
    ax.plot(trans[0], trans[1], 'r-', linewidth=2)
    ax.set_xlim(-2.5, 2.5)
    ax.set_ylim(-2.5, 2.5)
    ax.set_aspect('equal')
    ax.axhline(0, color='k', linewidth=0.5)
    ax.axvline(0, color='k', linewidth=0.5)
    ax.set_title(titulo, fontsize=12, fontweight='bold')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

plt.suptitle('Transformaciones canónicas en $\\mathbb{R}^2$', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 🧮 Representación matricial

**Teorema fundamental:** Sea $T: \mathbb{R}^n \to \mathbb{R}^m$ una transformación lineal. Entonces existe una **única** matriz $A$ de tamaño $m \times n$ tal que

$$T(\mathbf{x}) = A\mathbf{x} \quad \text{para todo } \mathbf{x} \in \mathbb{R}^n.$$

### ¿Cómo construir $A$?

Las columnas de $A$ son las imágenes de los vectores de la **base estándar** $\{\mathbf{e}_1, \mathbf{e}_2, \ldots, \mathbf{e}_n\}$:

$$A = \begin{pmatrix} T(\mathbf{e}_1) & T(\mathbf{e}_2) & \cdots & T(\mathbf{e}_n) \end{pmatrix}$$

**¿Por qué?** Cualquier vector $\mathbf{x} = x_1\mathbf{e}_1 + \cdots + x_n\mathbf{e}_n$, entonces por linealidad:
$$T(\mathbf{x}) = x_1T(\mathbf{e}_1) + \cdots + x_nT(\mathbf{e}_n) = A\mathbf{x}.$$

### Ejemplo

Sea $T: \mathbb{R}^3 \to \mathbb{R}^2$ definida por $T(x, y, z) = (x + 2y - z,\; 3x - y + 2z)$.

Calculamos las imágenes de la base estándar:
- $T(1,0,0) = (1, 3)$
- $T(0,1,0) = (2, -1)$
- $T(0,0,1) = (-1, 2)$

Por lo tanto: $A = \begin{pmatrix} 1 & 2 & -1 \\ 3 & -1 & 2 \end{pmatrix}$

```python
import numpy as np

# Definicion de T como funcion Python
def T(v):
    """T(x,y,z) = (x+2y-z, 3x-y+2z)"""
    x, y, z = v
    return np.array([x + 2*y - z, 3*x - y + 2*z], dtype=float)

# Construir la matriz aplicando T a cada vector de la base estandar
n = 3  # dimension del dominio
base_estandar = np.eye(n)

# Las columnas de A son T(e_1), T(e_2), T(e_3)
A = np.column_stack([T(e) for e in base_estandar])
print("Matriz A de la transformacion T:")
print(A)

# Verificacion: aplicar T directamente vs multiplicar por A
x = np.array([2.0, -1.0, 3.0])
print(f"\nVector de prueba: x = {x}")
print(f"T(x) calculado con la funcion:  {T(x)}")
print(f"A @ x (multiplicacion matricial): {A @ x}")
print(f"Resultados iguales: {np.allclose(T(x), A @ x)}")
```

## 🔎 Núcleo e Imagen

Toda transformación lineal $T: V \to W$ tiene asociados dos subespacios vectoriales fundamentales:

### Núcleo (Kernel)

$$\ker(T) = \{\mathbf{v} \in V : T(\mathbf{v}) = \mathbf{0}\}$$

El núcleo vive en el **dominio** $V$. Mide "cuánta información pierde" la transformación. Si $\ker(T) = \{\mathbf{0}\}$, entonces $T$ no pierde nada (es inyectiva).

### Imagen (Range)

$$\operatorname{Im}(T) = \{T(\mathbf{v}) : \mathbf{v} \in V\} = \{\mathbf{w} \in W : \mathbf{w} = T(\mathbf{v}) \text{ para algún } \mathbf{v}\}$$

La imagen vive en el **codominio** $W$. Es el conjunto de todos los vectores que $T$ puede alcanzar.

### Interpretación matricial

Si $T(\mathbf{x}) = A\mathbf{x}$:
- $\ker(T) = $ **espacio nulo** de $A$ = soluciones de $A\mathbf{x} = \mathbf{0}$
- $\operatorname{Im}(T) = $ **espacio columna** de $A$ = combinaciones lineales de columnas de $A$

```python
import numpy as np

# Ejemplo: T: R^4 -> R^3 con matriz A
A = np.array([[1,  2,  0,  1],
              [2,  4,  1,  3],
              [1,  2, -1,  0]], dtype=float)

print("Matriz A:")
print(A)
print(f"Forma: {A.shape} (3x4, de R^4 a R^3)")

# --- Encontrar el nucleo usando SVD ---
# El espacio nulo de A son los vectores singulares derechos
# correspondientes a valores singulares cero (o casi cero)
U, s, Vt = np.linalg.svd(A)
print(f"\nValores singulares: {np.round(s, 4)}")

tolerancia = 1e-10
nucleo_indices = np.where(s < tolerancia)[0] if len(s) == A.shape[1] else []
# Vt tiene shape (4,4); las filas de Vt con valores singulares ~0 forman el nucleo
# Primero completamos con scipy si disponible, sino calculamos manualmente
rango = np.linalg.matrix_rank(A)
nulidad = A.shape[1] - rango
print(f"Rango de A: {rango}")
print(f"Nulidad (dim del nucleo): {nulidad}")

# Bases del espacio nulo: ultimas 'nulidad' filas de Vt
if nulidad > 0:
    base_nucleo = Vt[rango:, :].T  # columnas son base del nucleo
    print(f"\nBase del nucleo (columnas):")
    print(np.round(base_nucleo, 4))
    # Verificar: A @ base_nucleo ~ 0
    print(f"Verificacion A @ base_nucleo ~ 0:")
    print(np.round(A @ base_nucleo, 10))

# --- Imagen: columnas de A despues de reduccion ---
print(f"\nDimension de la imagen (rango): {rango}")
print("La imagen es el espacio generado por las columnas de A.")
```

## 📏 Teorema Rango-Nulidad

Sea $T: V \to W$ una transformación lineal con $\dim(V) = n$. Entonces:

$$\boxed{\dim(\ker T) + \dim(\operatorname{Im} T) = n}$$

Los nombres:
- $\operatorname{nulidad}(T) = \dim(\ker T)$
- $\operatorname{rango}(T) = \dim(\operatorname{Im} T)$

### Interpretación

El teorema dice que la dimensión del dominio se *reparte* entre lo que se pierde (núcleo) y lo que se alcanza (imagen). Si la dimensión del dominio es $n$, entonces un subespacio de dimensión $k$ se *colapsa* al origen, y el resto (dimensión $n-k$) se *proyecta fielmente* en la imagen.

### Consecuencias inmediatas

- Si $T: \mathbb{R}^n \to \mathbb{R}^n$ y $\operatorname{rango}(T) = n$ ⟹ $\ker(T) = \{\mathbf{0}\}$ ⟹ $T$ es invertible.
- Si $m < n$ (de $\mathbb{R}^n$ a $\mathbb{R}^m$), entonces $\ker(T)$ tiene dimensión al menos $n - m > 0$.
- Si $m > n$, entonces $\operatorname{Im}(T)$ tiene dimensión a lo más $n < m$, por lo que $T$ no puede ser suprayectiva.

```python
import sympy as sp

# Calculo simbolico: nucleo, imagen, rango y nulidad
A_sym = sp.Matrix([[1, 2, 0, 1],
                   [2, 4, 1, 3],
                   [1, 2,-1, 0]])

print("Matriz A (simbolica):")
sp.pprint(A_sym)

# Rango
rango = A_sym.rank()
print(f"\nRango = {rango}")

# Nucleo (espacio nulo)
nucleo = A_sym.nullspace()
print(f"\nNulidad = {len(nucleo)}")
print("Base del nucleo:")
for i, v in enumerate(nucleo):
    print(f"  v{i+1} =", end=" ")
    sp.pprint(v.T)

# Imagen (espacio columna)
imagen = A_sym.columnspace()
print(f"\nDimension imagen = {len(imagen)}")
print("Base de la imagen:")
for i, v in enumerate(imagen):
    print(f"  w{i+1} =", end=" ")
    sp.pprint(v.T)

# Verificar teorema rango-nulidad
n_dom = A_sym.shape[1]  # dimension del dominio
print(f"\nVerificacion: nulidad + rango = {len(nucleo)} + {rango} = {len(nucleo)+rango}")
print(f"Dimension del dominio n = {n_dom}")
print(f"Teorema rango-nulidad se cumple: {len(nucleo) + rango == n_dom}")
```

## 🔑 Transformaciones inyectivas, suprayectivas y biyectivas

### Definiciones

| Tipo | Definición | Condición matricial |
|------|-----------|---------------------|
| **Inyectiva** (1-a-1) | $T(\mathbf{u}) = T(\mathbf{v}) \Rightarrow \mathbf{u} = \mathbf{v}$ | $\ker(T) = \{\mathbf{0}\}$, rango $= n$ |
| **Suprayectiva** (sobre) | $\operatorname{Im}(T) = W$ | Rango $= m$ |
| **Biyectiva** (invertible) | Inyectiva **y** suprayectiva | $A$ es cuadrada e invertible |

### Criterio práctico con el teorema rango-nulidad

Para $T: \mathbb{R}^n \to \mathbb{R}^m$ con matriz $A$ ($m \times n$):

- **Inyectiva** $\Leftrightarrow$ $\operatorname{nulidad}(A) = 0$ $\Leftrightarrow$ $\operatorname{rango}(A) = n$
- **Suprayectiva** $\Leftrightarrow$ $\operatorname{rango}(A) = m$
- **Biyectiva** $\Leftrightarrow$ $m = n$ y $\det(A) \neq 0$

### Observación geométrica

- Una rotación en $\mathbb{R}^2$ es biyectiva (no pierde información, cubre todo el plano).
- Una proyección sobre el eje $x$ **no** es inyectiva (colapsa el eje $y$) y **no** es suprayectiva (solo alcanza el eje $x$).

```python
import numpy as np

def clasificar_transformacion(A):
    """Clasifica una transformacion lineal como inyectiva, suprayectiva o biyectiva."""
    m, n = A.shape
    rango = np.linalg.matrix_rank(A)
    nulidad = n - rango

    inyectiva = (nulidad == 0)
    suprayectiva = (rango == m)
    biyectiva = inyectiva and suprayectiva

    print(f"  Dimension dominio n={n}, codominio m={m}")
    print(f"  Rango={rango}, Nulidad={nulidad}")
    print(f"  Inyectiva:    {inyectiva}")
    print(f"  Suprayectiva: {suprayectiva}")
    print(f"  Biyectiva:    {biyectiva}")

# Ejemplo 1: Rotacion (2x2, biyectiva)
theta = np.pi / 6
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
print("Rotacion 30 grados (2x2):")
clasificar_transformacion(R)

# Ejemplo 2: Proyeccion sobre eje x (2x2, ni inyectiva ni suprayectiva como T: R^2->R^2)
P = np.array([[1, 0], [0, 0]], dtype=float)
print("\nProyeccion sobre eje x (2x2):")
clasificar_transformacion(P)

# Ejemplo 3: T: R^3 -> R^2 con rango 2 (suprayectiva pero no inyectiva)
B = np.array([[1, 0, 1],
              [0, 1, 1]], dtype=float)
print("\nT: R^3 -> R^2 con rango 2:")
clasificar_transformacion(B)

# Ejemplo 4: T: R^2 -> R^3 con rango 2 (inyectiva pero no suprayectiva)
C = np.array([[1, 0],
              [0, 1],
              [1, 1]], dtype=float)
print("\nT: R^2 -> R^3 con rango 2:")
clasificar_transformacion(C)
```
