# Operaciones Vectoriales en $\mathbb{R}^2$

Las operaciones vectoriales son los bloques fundamentales que definen la estructura de un **espacio vectorial**. En $\mathbb{R}^2$, estas operaciones no solo tienen una definición algebraica precisa, sino también una interpretación geométrica intuitiva que nos ayuda a comprender su significado.

En esta sección estudiaremos:
- La **suma de vectores** y la regla del paralelogramo
- La **multiplicación por escalar** y su efecto geométrico
- El **vector cero** y el **vector inverso**
- Los **axiomas** que definen un espacio vectorial

Estas operaciones son la base para todo el álgebra lineal que estudiaremos en el curso.

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import warnings
warnings.filterwarnings('ignore')

# Configuración para gráficos
plt.rcParams['figure.figsize'] = [8, 6]
plt.rcParams['font.size'] = 12
````

## 1. Suma de vectores

### Definición algebraica

Dados dos vectores $\mathbf{u} = (u_1, u_2)$ y $\mathbf{v} = (v_1, v_2)$ en $\mathbb{R}^2$, su **suma** se define componente a componente:

$$
\mathbf{u} + \mathbf{v} = (u_1 + v_1, u_2 + v_2)
$$

:::{admonition} Ejemplo
:class: tip
Si $\mathbf{u} = (3, 1)$ y $\mathbf{v} = (1, 2)$, entonces:

$$\mathbf{u} + \mathbf{v} = (3+1, 1+2) = (4, 3)$$
:::

### Interpretación geométrica: Regla del paralelogramo

Geométricamente, la suma de dos vectores se puede visualizar de dos formas equivalentes:

1. **Método punta-cola**: Se coloca el inicio del segundo vector en la punta del primero. El vector suma va desde el origen hasta la punta del segundo vector.

2. **Regla del paralelogramo**: Se construye un paralelogramo con los dos vectores como lados adyacentes. La diagonal que parte del origen es el vector suma.

La siguiente visualización muestra ambos vectores y su suma:

````python
# Visualización de la suma de vectores
fig, ax = plt.subplots(figsize=(9, 7))

# Definir vectores
u = np.array([3, 1])
v = np.array([1, 2])
suma = u + v

# Dibujar vectores desde el origen
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.02, label=r'$\mathbf{u} = (3, 1)$')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.02, label=r'$\mathbf{v} = (1, 2)$')
ax.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, 
          color='green', width=0.025, label=r'$\mathbf{u} + \mathbf{v} = (4, 3)$')

# Dibujar el paralelogramo (líneas punteadas)
ax.plot([u[0], suma[0]], [u[1], suma[1]], 'r--', alpha=0.5, linewidth=1.5)
ax.plot([v[0], suma[0]], [v[1], suma[1]], 'b--', alpha=0.5, linewidth=1.5)

# Configuración del gráfico
ax.set_xlim(-0.5, 5.5)
ax.set_ylim(-0.5, 4.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('Suma de vectores: Regla del paralelogramo', fontsize=14)
ax.legend(loc='upper left', fontsize=11)

# Etiquetas de puntos
ax.annotate('O', (0, 0), textcoords="offset points", xytext=(-10,-10), fontsize=11)
ax.plot(suma[0], suma[1], 'go', markersize=6)

plt.tight_layout()
plt.show()
````

## 2.  Exploración: Regla del Paralelogramo

Usa los controles deslizantes para modificar los componentes de los vectores $\mathbf{u}$ y $\mathbf{v}$. Observa cómo cambia el paralelogramo y el vector suma.

````python
def graficar_suma_interactiva(u1=2.0, u2=1.0, v1=1.0, v2=2.0):
    """Visualización interactiva de la suma de vectores."""
    fig, ax = plt.subplots(figsize=(9, 7))
    
    u = np.array([u1, u2])
    v = np.array([v1, v2])
    suma = u + v
    
    # Dibujar vectores
    ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1,
              color='blue', width=0.015, label=f'$\\mathbf{{u}} = ({u1:.1f}, {u2:.1f})$')
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
              color='red', width=0.015, label=f'$\\mathbf{{v}} = ({v1:.1f}, {v2:.1f})$')
    ax.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1,
              color='green', width=0.02, label=f'$\\mathbf{{u}} + \\mathbf{{v}} = ({suma[0]:.1f}, {suma[1]:.1f})$')
    
    # Paralelogramo completo
    parallelogram = np.array([[0, 0], [u[0], u[1]], [suma[0], suma[1]], [v[0], v[1]], [0, 0]])
    ax.fill(parallelogram[:, 0], parallelogram[:, 1], alpha=0.15, color='purple')
    ax.plot([u[0], suma[0]], [u[1], suma[1]], 'r--', alpha=0.6, linewidth=1.5)
    ax.plot([v[0], suma[0]], [v[1], suma[1]], 'b--', alpha=0.6, linewidth=1.5)
    
    # Configuración
    max_val = max(abs(suma[0]), abs(suma[1]), abs(u[0]), abs(u[1]), abs(v[0]), abs(v[1]), 1) + 1
    ax.set_xlim(-max_val, max_val)
    ax.set_ylim(-max_val, max_val)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_title('Exploración interactiva: Suma de vectores', fontsize=14)
    ax.legend(loc='best', fontsize=10)
    plt.tight_layout()
    plt.show()

interact(graficar_suma_interactiva,
         u1=FloatSlider(value=2.0, min=-3.0, max=3.0, step=0.5, description='u₁'),
         u2=FloatSlider(value=1.0, min=-3.0, max=3.0, step=0.5, description='u₂'),
         v1=FloatSlider(value=1.0, min=-3.0, max=3.0, step=0.5, description='v₁'),
         v2=FloatSlider(value=2.0, min=-3.0, max=3.0, step=0.5, description='v₂'));
````

## 3. Multiplicación por escalar

### Definición

Dado un vector $\mathbf{v} = (v_1, v_2)$ y un escalar $c \in \mathbb{R}$, el **producto escalar** se define como:

$$
c \cdot \mathbf{v} = c(v_1, v_2) = (cv_1, cv_2)
$$

### Interpretación geométrica

La multiplicación por un escalar tiene diferentes efectos geométricos según el valor de $c$:

| Valor de $c$ | Efecto sobre $\mathbf{v}$ |
|:---:|:---|
| $c > 1$ | **Estira** el vector (aumenta su longitud) |
| $0 < c < 1$ | **Contrae** el vector (disminuye su longitud) |
| $c = 1$ | El vector permanece igual |
| $c = 0$ | Produce el **vector cero** |
| $c < 0$ | **Invierte** la dirección y escala por $|c|$ |

:::{admonition} Observación importante
:class: note
La multiplicación por escalar siempre produce un vector **paralelo** (o antiparalelo) al original. Los vectores $\mathbf{v}$ y $c\mathbf{v}$ siempre están sobre la misma recta que pasa por el origen.
:::

````python
# Visualización de multiplicación por escalar
fig, axes = plt.subplots(1, 3, figsize=(14, 5))

v = np.array([2, 1])
escalares = [2, 0.5, -1]
titulos = [r'$c = 2$ (estiramiento)', r'$c = 0.5$ (contracción)', r'$c = -1$ (inversión)']
colores_cv = ['green', 'orange', 'purple']

for ax, c, titulo, color in zip(axes, escalares, titulos, colores_cv):
    cv = c * v
    
    # Vector original
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
              color='blue', width=0.03, label=r'$\mathbf{v} = (2, 1)$')
    # Vector escalado
    ax.quiver(0, 0, cv[0], cv[1], angles='xy', scale_units='xy', scale=1,
              color=color, width=0.03, alpha=0.8, 
              label=f'$c\\mathbf{{v}} = ({cv[0]:.1f}, {cv[1]:.1f})$')
    
    # Línea de dirección
    t = np.linspace(-3, 3, 100)
    ax.plot(t * v[0]/np.linalg.norm(v) * 3, t * v[1]/np.linalg.norm(v) * 3, 
            'k--', alpha=0.2, linewidth=1)
    
    ax.set_xlim(-5, 5)
    ax.set_ylim(-3, 3)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.set_title(titulo, fontsize=12)
    ax.legend(loc='upper left', fontsize=9)

plt.suptitle('Multiplicación por escalar: diferentes valores de $c$', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
````

## 4.  Exploración: Multiplicación Escalar

Usa el control deslizante para variar el escalar $c$ y observa cómo se transforma el vector $\mathbf{v} = (2, 1)$.

````python
def graficar_escalar_interactivo(c=1.0):
    """Visualización interactiva de la multiplicación por escalar."""
    fig, ax = plt.subplots(figsize=(9, 7))
    
    v = np.array([2, 1])
    cv = c * v
    
    # Línea de dirección
    t = np.linspace(-4, 4, 100)
    ax.plot(t * v[0]/np.linalg.norm(v) * 5, t * v[1]/np.linalg.norm(v) * 5, 
            'k--', alpha=0.2, linewidth=1.5, label='Línea de dirección')
    
    # Vector original (siempre visible)
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
              color='blue', width=0.02, alpha=0.5, label=r'$\mathbf{v} = (2, 1)$')
    
    # Vector escalado
    if c != 0:
        color = 'green' if c > 0 else 'red'
        ax.quiver(0, 0, cv[0], cv[1], angles='xy', scale_units='xy', scale=1,
                  color=color, width=0.025, 
                  label=f'$c\\mathbf{{v}} = {c:.1f} \\cdot (2,1) = ({cv[0]:.1f}, {cv[1]:.1f})$')
    else:
        ax.plot(0, 0, 'ko', markersize=10, label=r'$c\mathbf{v} = \mathbf{0}$')
    
    # Información del escalar
    if c > 1:
        estado = "Estiramiento"
    elif 0 < c < 1:
        estado = "Contracción"
    elif c == 1:
        estado = "Sin cambio"
    elif c == 0:
        estado = "Vector cero"
    elif -1 < c < 0:
        estado = "Inversión + Contracción"
    elif c == -1:
        estado = "Inversión pura"
    else:
        estado = "Inversión + Estiramiento"
    
    ax.set_xlim(-7, 7)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_title(f'Multiplicación por escalar: $c = {c:.1f}$ — {estado}', fontsize=14)
    ax.legend(loc='upper left', fontsize=10)
    plt.tight_layout()
    plt.show()

interact(graficar_escalar_interactivo,
         c=FloatSlider(value=1.0, min=-3.0, max=3.0, step=0.25, description='Escalar c'));
````

## 5. Vector cero y vector inverso

### El vector cero

El **vector cero** en $\mathbb{R}^2$ es:

$$
\mathbf{0} = (0, 0)
$$

Este vector es el **elemento neutro** de la suma de vectores. Para cualquier vector $\mathbf{v}$:

$$
\mathbf{v} + \mathbf{0} = \mathbf{0} + \mathbf{v} = \mathbf{v}
$$

Geométricamente, el vector cero no tiene dirección ni magnitud; es simplemente un punto en el origen.

### El vector inverso (opuesto aditivo)

Para cada vector $\mathbf{v} = (v_1, v_2)$, existe un único **vector inverso** o **opuesto aditivo**:

$$
-\mathbf{v} = (-v_1, -v_2)
$$

Este vector satisface:

$$
\mathbf{v} + (-\mathbf{v}) = \mathbf{0}
$$

:::{admonition} Observación
:class: note
El vector inverso $-\mathbf{v}$ es equivalente a multiplicar $\mathbf{v}$ por el escalar $-1$:

$$-\mathbf{v} = (-1) \cdot \mathbf{v}$$

Geométricamente, $-\mathbf{v}$ tiene la misma magnitud que $\mathbf{v}$ pero apunta en la dirección exactamente opuesta.
:::

### La resta de vectores

La **resta** de vectores se define en términos de la suma y el inverso:

$$
\mathbf{u} - \mathbf{v} = \mathbf{u} + (-\mathbf{v}) = (u_1 - v_1, u_2 - v_2)
$$

````python
# Visualización del vector inverso
fig, ax = plt.subplots(figsize=(8, 6))

v = np.array([3, 2])
neg_v = -v

# Vectores
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.02, label=r'$\mathbf{v} = (3, 2)$')
ax.quiver(0, 0, neg_v[0], neg_v[1], angles='xy', scale_units='xy', scale=1,
          color='red', width=0.02, label=r'$-\mathbf{v} = (-3, -2)$')

# Vector cero
ax.plot(0, 0, 'ko', markersize=8)
ax.annotate(r'$\mathbf{0}$', (0, 0), textcoords="offset points", 
            xytext=(10, -15), fontsize=12, color='black')

# Línea de conexión
ax.plot([v[0], neg_v[0]], [v[1], neg_v[1]], 'g--', alpha=0.4, linewidth=1)

ax.set_xlim(-5, 5)
ax.set_ylim(-4, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title(r'Vector $\mathbf{v}$ y su inverso $-\mathbf{v}$', fontsize=14)
ax.legend(loc='upper left', fontsize=11)
plt.tight_layout()
plt.show()
````

## 6. Propiedades del espacio vectorial $\mathbb{R}^2$

Las operaciones de suma y multiplicación por escalar que hemos definido satisfacen un conjunto de propiedades que hacen de $\mathbb{R}^2$ un **espacio vectorial**.

:::{admonition} Los 10 axiomas de un espacio vectorial
:class: important

Para todo $\mathbf{u}, \mathbf{v}, \mathbf{w} \in \mathbb{R}^2$ y escalares $c, d \in \mathbb{R}$:

**Axiomas de la suma:**
1. **Cerradura de la suma:** $\mathbf{u} + \mathbf{v} \in \mathbb{R}^2$
2. **Conmutatividad:** $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$
3. **Asociatividad:** $(\mathbf{u} + \mathbf{v}) + \mathbf{w} = \mathbf{u} + (\mathbf{v} + \mathbf{w})$
4. **Elemento neutro:** Existe $\mathbf{0}$ tal que $\mathbf{v} + \mathbf{0} = \mathbf{v}$
5. **Elemento inverso:** Existe $-\mathbf{v}$ tal que $\mathbf{v} + (-\mathbf{v}) = \mathbf{0}$

**Axiomas de la multiplicación escalar:**
6. **Cerradura del producto:** $c \cdot \mathbf{v} \in \mathbb{R}^2$
7. **Asociatividad escalar:** $c(d\mathbf{v}) = (cd)\mathbf{v}$
8. **Identidad multiplicativa:** $1 \cdot \mathbf{v} = \mathbf{v}$

**Axiomas distributivos:**
9. **Distributividad sobre suma vectorial:** $c(\mathbf{u} + \mathbf{v}) = c\mathbf{u} + c\mathbf{v}$
10. **Distributividad sobre suma escalar:** $(c + d)\mathbf{v} = c\mathbf{v} + d\mathbf{v}$
:::

### Significado geométrico de algunos axiomas

- **Conmutatividad de la suma:** El orden en que sumamos vectores no importa. El paralelogramo se puede construir de cualquier manera.

- **Asociatividad de la suma:** Cuando sumamos tres vectores, no importa cómo los agrupemos.

- **Distributividad:** Escalar una suma es lo mismo que sumar los vectores escalados. Esto refleja que el escalamiento "se distribuye" uniformemente.

````python
# Visualización de la conmutatividad
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

u = np.array([2, 1])
v = np.array([1, 2])
suma = u + v

# u + v
ax = axes[0]
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.02, label=r'$\mathbf{u}$')
ax.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1,
          color='red', width=0.02, label=r'$\mathbf{v}$ (desde $\mathbf{u}$)')
ax.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1,
          color='green', width=0.02, alpha=0.7, label=r'$\mathbf{u} + \mathbf{v}$')
ax.set_title(r'$\mathbf{u} + \mathbf{v}$: primero $\mathbf{u}$, luego $\mathbf{v}$', fontsize=12)
ax.set_xlim(-0.5, 4.5)
ax.set_ylim(-0.5, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper left', fontsize=9)

# v + u
ax = axes[1]
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
          color='red', width=0.02, label=r'$\mathbf{v}$')
ax.quiver(v[0], v[1], u[0], u[1], angles='xy', scale_units='xy', scale=1,
          color='blue', width=0.02, label=r'$\mathbf{u}$ (desde $\mathbf{v}$)')
ax.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1,
          color='green', width=0.02, alpha=0.7, label=r'$\mathbf{v} + \mathbf{u}$')
ax.set_title(r'$\mathbf{v} + \mathbf{u}$: primero $\mathbf{v}$, luego $\mathbf{u}$', fontsize=12)
ax.set_xlim(-0.5, 4.5)
ax.set_ylim(-0.5, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper left', fontsize=9)

plt.suptitle('Conmutatividad de la suma: ¡El resultado es el mismo!', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
````
