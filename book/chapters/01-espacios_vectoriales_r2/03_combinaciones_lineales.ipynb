# 1.3 Combinaciones Lineales

Las **combinaciones lineales** son una de las ideas más fundamentales del álgebra lineal. Permiten construir nuevos vectores a partir de vectores existentes mediante operaciones simples: multiplicación por escalares y suma.

¿Por qué son tan importantes?

- **Construcción**: Nos permiten describir qué vectores podemos "alcanzar" partiendo de un conjunto dado.
- **Sistemas de ecuaciones**: Resolver $A\mathbf{x} = \mathbf{b}$ es equivalente a preguntar si $\mathbf{b}$ es combinación lineal de las columnas de $A$.
- **Espacios vectoriales**: El concepto de *span* (generado) construye subespacios a partir de combinaciones lineales.
- **Dependencia lineal**: Determinar si un vector es "redundante" en un conjunto.

En esta sección, exploraremos estas ideas con énfasis en la interpretación geométrica en $\mathbb{R}^2$.

**Motivación:** Dado un conjunto pequeño de vectores base, ¿qué vectores del plano pueden construirse realmente con ellos? Probar combinaciones al azar da ejemplos, pero no ofrece un criterio global de "alcanzabilidad". Las herramientas vistas hasta ahora permiten operar vectores, pero no responder de manera estructurada qué conjunto total generan. Esta limitación conduce a la idea de **combinación lineal** y al concepto de **generado (span)**.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import warnings
warnings.filterwarnings('ignore')
```

## 1.3.1. Definición

:::{admonition} Definición: Combinación Lineal
:class: tip

Sean $\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k$ vectores en $\mathbb{R}^n$ y sean $c_1, c_2, \ldots, c_k$ escalares en $\mathbb{R}$. 

Una **combinación lineal** de estos vectores es cualquier vector de la forma:

$$
c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \cdots + c_k\mathbf{v}_k = \sum_{i=1}^{k} c_i\mathbf{v}_i
$$

Los escalares $c_1, c_2, \ldots, c_k$ se llaman **coeficientes** de la combinación lineal.
:::

**Observaciones importantes:**
- Los coeficientes pueden ser cualquier número real: positivos, negativos o cero.
- Si $c_i = 0$, el vector $\mathbf{v}_i$ no contribuye a la combinación.
- El vector cero $\mathbf{0}$ siempre es combinación lineal de cualquier conjunto (con todos los coeficientes iguales a cero).

### Ejemplo básico

Si $\mathbf{v}_1 = (1, 0)$ y $\mathbf{v}_2 = (0, 1)$, entonces:

$$
3\mathbf{v}_1 + 2\mathbf{v}_2 = 3(1, 0) + 2(0, 1) = (3, 0) + (0, 2) = (3, 2)
$$

## 1.3.2. Interpretación Geométrica

En $\mathbb{R}^2$, una combinación lineal de dos vectores $\mathbf{v}_1$ y $\mathbf{v}_2$ tiene una interpretación geométrica clara:

1. **Escalar cada vector**: $c_1\mathbf{v}_1$ estira o comprime $\mathbf{v}_1$ (y lo invierte si $c_1 < 0$).
2. **Sumar los resultados**: El vector resultante se obtiene usando la regla del paralelogramo.

Veamos un ejemplo donde expresamos el vector $\mathbf{w} = (5, 3)$ como combinación lineal de $\mathbf{v}_1 = (2, 1)$ y $\mathbf{v}_2 = (1, 1)$.

```python
fig, ax = plt.subplots(figsize=(10, 8))

# Vectores base
v1 = np.array([2, 1])
v2 = np.array([1, 1])

# Coeficientes: w = 2*v1 + 1*v2 = (4,2) + (1,1) = (5,3)
c1, c2 = 2, 1
w = c1 * v1 + c2 * v2

# Vector escalado c1*v1
scaled_v1 = c1 * v1

# Dibujar vectores originales
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.015, label=r'$\mathbf{v}_1 = (2, 1)$')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
          color='green', width=0.015, label=r'$\mathbf{v}_2 = (1, 1)$')

# Dibujar vectores escalados (más claros)
ax.quiver(0, 0, scaled_v1[0], scaled_v1[1], angles='xy', scale_units='xy', scale=1, 
          color='lightblue', width=0.012, alpha=0.7, label=r'$2\mathbf{v}_1 = (4, 2)$')

# Dibujar la suma (desde la punta de c1*v1, sumar c2*v2)
ax.quiver(scaled_v1[0], scaled_v1[1], c2*v2[0], c2*v2[1], angles='xy', scale_units='xy', scale=1, 
          color='lightgreen', width=0.012, alpha=0.7, linestyle='--')

# Vector resultante
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.02, label=r'$\mathbf{w} = 2\mathbf{v}_1 + 1\mathbf{v}_2 = (5, 3)$')

# Paralelogramo
parallelogram = plt.Polygon([[0, 0], [scaled_v1[0], scaled_v1[1]], 
                              [w[0], w[1]], [c2*v2[0], c2*v2[1]]], 
                             fill=True, alpha=0.1, color='purple')
ax.add_patch(parallelogram)

ax.set_xlim(-1, 7)
ax.set_ylim(-1, 5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12)
ax.set_title(r'Combinación lineal: $\mathbf{w} = c_1\mathbf{v}_1 + c_2\mathbf{v}_2$', fontsize=14)
ax.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()
```

**Observa cómo:**
- El vector azul claro $2\mathbf{v}_1$ es el doble del vector $\mathbf{v}_1$.
- Desde la punta de $2\mathbf{v}_1$, sumamos $1 \cdot \mathbf{v}_2$ (vector verde claro).
- El resultado es el vector rojo $\mathbf{w}$.
- El paralelogramo sombreado ilustra la regla de suma de vectores.

## 1.3.4. Generado (Span)

El conjunto de **todas** las combinaciones lineales posibles de un conjunto de vectores tiene un nombre especial.

:::{admonition} Definición: Generado (Span)
:class: tip

Sean $\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k$ vectores en $\mathbb{R}^n$. El **generado** (o *span*) de estos vectores es el conjunto de todas sus combinaciones lineales:

$$
\text{span}\{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k\} = \left\{ c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \cdots + c_k\mathbf{v}_k \mid c_1, c_2, \ldots, c_k \in \mathbb{R} \right\}
$$
:::

### Ejemplos geométricos

**¿Cuál es el span de un solo vector no nulo?**

Si $\mathbf{v} \neq \mathbf{0}$, entonces:

$$
\text{span}\{\mathbf{v}\} = \{c\mathbf{v} \mid c \in \mathbb{R}\}
$$

Esto es una **recta** que pasa por el origen en la dirección de $\mathbf{v}$.

**¿Cuál es el span de dos vectores no paralelos en $\mathbb{R}^2$?**

Si $\mathbf{v}_1$ y $\mathbf{v}_2$ no son paralelos (ninguno es múltiplo escalar del otro), entonces:

$$
\text{span}\{\mathbf{v}_1, \mathbf{v}_2\} = \mathbb{R}^2
$$

¡Podemos alcanzar **cualquier** punto del plano!

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Caso 1: Span de un vector (recta)
ax1 = axes[0]
v = np.array([2, 1])
t = np.linspace(-3, 3, 100)
ax1.plot(t*v[0], t*v[1], 'b-', linewidth=2, label='span{v}')
ax1.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.02)
ax1.set_xlim(-7, 7)
ax1.set_ylim(-4, 4)
ax1.set_aspect('equal')
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)
ax1.set_title(r'span$\{\mathbf{v}\}$ = recta', fontsize=12)
ax1.text(2.3, 1.3, r'$\mathbf{v}$', fontsize=12, color='red')

# Caso 2: Span de dos vectores paralelos (recta)
ax2 = axes[1]
v1 = np.array([2, 1])
v2 = np.array([4, 2])  # v2 = 2*v1, son paralelos
ax2.plot(t*v1[0], t*v1[1], 'b-', linewidth=2, label='span{v₁, v₂}')
ax2.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.02)
ax2.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
           color='orange', width=0.02)
ax2.set_xlim(-7, 7)
ax2.set_ylim(-4, 4)
ax2.set_aspect('equal')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.axvline(x=0, color='k', linewidth=0.5)
ax2.set_title(r'$\mathbf{v}_1 \parallel \mathbf{v}_2$: span = recta', fontsize=12)
ax2.text(2.3, 1.3, r'$\mathbf{v}_1$', fontsize=12, color='red')
ax2.text(4.3, 2.3, r'$\mathbf{v}_2$', fontsize=12, color='orange')

# Caso 3: Span de dos vectores NO paralelos (todo R²)
ax3 = axes[2]
v1 = np.array([2, 1])
v2 = np.array([1, 2])  # No paralelos
# Sombrear todo el plano
ax3.fill([-7, 7, 7, -7], [-4, -4, 4, 4], color='lightblue', alpha=0.3)
ax3.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.02)
ax3.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.02)
ax3.set_xlim(-7, 7)
ax3.set_ylim(-4, 4)
ax3.set_aspect('equal')
ax3.grid(True, alpha=0.3)
ax3.axhline(y=0, color='k', linewidth=0.5)
ax3.axvline(x=0, color='k', linewidth=0.5)
ax3.set_title(r'$\mathbf{v}_1 \nparallel \mathbf{v}_2$: span = $\mathbb{R}^2$', fontsize=12)
ax3.text(2.3, 1.3, r'$\mathbf{v}_1$', fontsize=12, color='red')
ax3.text(1.3, 2.3, r'$\mathbf{v}_2$', fontsize=12, color='green')

plt.tight_layout()
plt.show()
```

## 1.3.5. ¿Cuándo el generado es todo $\mathbb{R}^2$?

Esta es una pregunta fundamental: dado un conjunto de vectores, ¿cuándo podemos alcanzar cualquier punto del plano?

:::{admonition} Condición clave
:class: important

En $\mathbb{R}^2$, el span de dos vectores $\mathbf{v}_1$ y $\mathbf{v}_2$ es todo $\mathbb{R}^2$ si y solo si los vectores **no son paralelos**.

Matemáticamente, $\mathbf{v}_1$ y $\mathbf{v}_2$ no son paralelos cuando ninguno es múltiplo escalar del otro:

$$
\mathbf{v}_1 \neq c \cdot \mathbf{v}_2 \quad \text{para todo } c \in \mathbb{R}
$$
:::

### Conexión con sistemas de ecuaciones

 Preguntar si el vector **w** está en span$\{\mathbf{v}_1, \mathbf{v}_2\}$ es equivalente a preguntar si se puede construir a partir de los vectores **v₁** y **v₂**:

> *¿Podemos encontrar números $c_1$ y $c_2$ tales que $c_1 v_1 + c_2 v_2 = w$*

Si escribimos los vectores en coordenadas:
- $v_2 = (b_1, b_2)$
- $v_1 = (a_1, a_2)$
- $w = (w_1, w_2)$

entonces la condición se convierte en dos ecuaciones que deben cumplirse al mismo tiempo:

- Primera coordenada:

$$
a_1 c_1 + b_1 c_2 = w_1
$$

- Segunda coordenada:

$$
a_2 c_1 + b_2 c_2 = w_2
$$

### ¿Cuándo es posible?

Esto se puede hacer para cualquier vector **w** siempre que **v₁ y v₂ no sean paralelos** (es decir, no son simplemente múltiplos uno del otro).

- Si apuntan en direcciones diferentes, pueden combinarse para formar cualquier vector del plano.
- Si están en la misma línea (uno es solo una versión escalada del otro), solo pueden generar vectores sobre esa línea, no todos los del plano.

## 1.3.6. Ejemplo Completo

**Problema:** Expresar el vector $\mathbf{w} = (5, 4)$ como combinación lineal de $\mathbf{v}_1 = (2, 1)$ y $\mathbf{v}_2 = (1, 2)$.

### Paso 1: Plantear el sistema

Buscamos $c_1$ y $c_2$ tales que:

$$
c_1 \begin{pmatrix} 2 \\ 1 \end{pmatrix} + c_2 \begin{pmatrix} 1 \\ 2 \end{pmatrix} = \begin{pmatrix} 5 \\ 4 \end{pmatrix}
$$

Esto nos da el sistema:

$$
\begin{cases}
2c_1 + c_2 = 5 \\
c_1 + 2c_2 = 4
\end{cases}
$$

### Paso 2: Resolver el sistema

De la primera ecuación: $c_2 = 5 - 2c_1$

Sustituyendo en la segunda:

$$
c_1 + 2(5 - 2c_1) = 4
$$

$$
c_1 + 10 - 4c_1 = 4
$$

$$
-3c_1 = -6
$$

$$
c_1 = 2
$$

Por lo tanto: $c_2 = 5 - 2(2) = 1$

### Paso 3: Verificar

$$
2 \begin{pmatrix} 2 \\ 1 \end{pmatrix} + 1 \begin{pmatrix} 1 \\ 2 \end{pmatrix} = \begin{pmatrix} 4 \\ 2 \end{pmatrix} + \begin{pmatrix} 1 \\ 2 \end{pmatrix} = \begin{pmatrix} 5 \\ 4 \end{pmatrix} \checkmark
$$

Expresa el vector $\mathbf{w} = (7, 5)$ como combinación lineal de $\mathbf{v}_1 = (1, 1)$ y $\mathbf{v}_2 = (2, 1)$.

1. Plantea el sistema de ecuaciones.
2. Resuelve para encontrar $c_1$ y $c_2$.
3. Verifica tu respuesta.

:::{admonition} Solución
:class: dropdown

**Sistema:**

$$
\begin{cases}
c_1 + 2c_2 = 7 \\
c_1 + c_2 = 5
\end{cases}
$$

Restando la segunda de la primera: $c_2 = 2$

Sustituyendo: $c_1 = 5 - 2 = 3$

**Respuesta:** $\mathbf{w} = 3\mathbf{v}_1 + 2\mathbf{v}_2$

**Verificación:** $3(1,1) + 2(2,1) = (3,3) + (4,2) = (7,5)$ ✓
:::
