# 🌌 Espacios Vectoriales en $\mathbb{R}^2$

En esta lección aprenderás a representar, operar y visualizar **vectores bidimensionales** en el espacio $\mathbb{R}^2$.  
También revisaremos las operaciones básicas que permiten definir la estructura de un **espacio vectorial**, así como su interpretación geométrica.


## 🔹 1. Representación de vectores en $\mathbb{R}^2$

Un **vector en $\mathbb{R}^2$** se representa como un par ordenado de números reales:

$$
\mathbf{v} = (v_1, v_2), \quad v_i \in \mathbb{R}
$$

Cada componente indica cuánto se avanza en las tres direcciones ortogonales del espacio:

- $x$ → dirección horizontal  
- $y$ → dirección lateral  
  

Geométricamente, un vector puede verse como una **flecha que parte del origen $(0,0)$** y termina en el punto $(v_1,v_2)$.

---

### 💻 Ejemplo en código: graficar un vector en 2D


```python
import warnings
warnings.filterwarnings("ignore", category=SyntaxWarning)
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir el vector
v = np.array([2, 3])

# Crear figura 2D
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
    
# Dibujar el vector desde el origen
# Es fundamental agregar angles='xy', scale_units='xy', scale=1 para que el vector 
# se dibuje con la longitud real acorde a las unidades de los ejes.
ax.quiver(0, 0, v[0], v[1], color='red', angles='xy', scale_units='xy', scale=1, linewidth=2)

# Configurar ejes
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Vector en $\mathbb{R}^2$')
ax.grid(True, linestyle='--', alpha=0.6) # Agregar grilla para mejor visualización
plt.show()
```


## 2. Operaciones básicas entre vectores

###  2.1 Suma de vectores

Dado $\mathbf{u} = (u_1, u_2)$ y $\mathbf{v} = (v_1, v_2)$:

$$
\mathbf{u} + \mathbf{v} = (u_1 + v_1,  u_2 + v_2)
$$

Esta operación representa la **composición de desplazamientos** en el espacio.






```python
import warnings
warnings.filterwarnings("ignore", category=SyntaxWarning)
import numpy as np
import matplotlib.pyplot as plt

# Definir vectores
u = np.array([3, 1])
v = np.array([1, 2])
suma = u + v

# Crear figura
fig, ax = plt.subplots(figsize=(6, 6))

# Dibujar vector u (desde el origen)
ax.quiver(0, 0, u[0], u[1], color='blue', angles='xy', scale_units='xy', scale=1, label='u')

# Dibujar vector v comenzando desde la punta de u (Método Cabeza-Cola)
# Se usa color 'green' sólido para igualar tonalidad
ax.quiver(u[0], u[1], v[0], v[1], color='green', angles='xy', scale_units='xy', scale=1, label='v (trasladado)')

# Dibujar vector resultante u+v (desde el origen hasta la punta del v trasladado)
ax.quiver(0, 0, suma[0], suma[1], color='red', angles='xy', scale_units='xy', scale=1, label='u + v')

# Dibujar v original desde el origen con el MISMO color (sin transparencia)
ax.quiver(0, 0, v[0], v[1], color='green', angles='xy', scale_units='xy', scale=1)

# Translación del vector u (azul) al extremo de v para completar el paralelogramo
ax.quiver(v[0], v[1], u[0], u[1], color='blue', angles='xy', scale_units='xy', scale=1, label='u (trasladado)')

# Configurar gráfico
limit = 5
ax.set_xlim(-1, limit)
ax.set_ylim(-1, limit)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend()
ax.set_title("Suma de Vectores: Método del Paralelogramo")
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
import numpy as np
```


###  Multiplicación por un escalar

Si $\alpha \in \mathbb{R}$ es un número real y $\mathbf{v} = (v_1, v_2)$ es un vector, entonces la multiplicación por escalar se define como:

$$
\alpha \mathbf{v} = (\alpha v_1, \alpha v_2)
$$

Esta operación **alarga**, **contrae** o **invierte** la dirección del vector, pero no cambia su línea de acción (a menos que el escalar sea 0).



###  Ejemplo interactivo: Multiplicación por escalar
Usa el deslizador para ver cómo cambia el vector al multiplicarlo por diferentes valores de $\alpha$.



```python

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def visualizar_escalar(alpha):
    vector_base = np.array([2, 1])
    vector_escalado = alpha * vector_base
    
    fig, ax = plt.subplots(figsize=(6, 6))
    
 
    
    # Vector escalado
    color = 'green' if alpha >= 0 else 'purple'
    ax.quiver(0, 0, vector_escalado[0], vector_escalado[1], color=color, 
              angles='xy', scale_units='xy', scale=1, label=f'Escalado (x{alpha:.1f})')
       # Vector original
    ax.quiver(0, 0, vector_base[0], vector_base[1], color='blue', alpha=0.3, 
              angles='xy', scale_units='xy', scale=1, label='Original (2, 1)')
    limit = 6
    ax.set_xlim(-limit, limit)
    ax.set_ylim(-limit, limit)
    ax.grid(True, linestyle='--', alpha=0.6)
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_aspect('equal')
    ax.legend()
    ax.set_title(f"Multiplicación por Escalar: v = {alpha:.1f} * u")
    plt.show()

# Crear el slider interactivo
interact(visualizar_escalar, 
         alpha=FloatSlider(min=-3.0, max=3.0, step=0.1, value=1.0, description='Escalar (α):'));
```

### Trabajemos con vectores en Python   
Podemos usar python para sumar y realizar  la multiplicación por un escalar a los vectores con la librería numpy. Aquí un ejemplo de cómo hacerlo:



```python
u = np.array([1, 2, 3])
v = np.array([3, -1, 2])

# Suma
suma = u + v

# Multiplicación por escalar
alpha = 2
escalar_v = alpha * v

print("u + v =", suma)
print(f"{alpha} * v =", escalar_v)
```



## 🔹 6. Axiomas de un espacio vectorial

$\mathbb{R}^2$ es un **espacio vectorial real** porque cumple los siguientes **10 axiomas fundamentales**. Para todo vector $\mathbf{u}, \mathbf{v}, \mathbf{w} \in \mathbb{R}^2$ y todo escalar $c, d \in \mathbb{R}$:

| Tipo de Operación | # | Nombre del Axioma | Expresión Matemática |
| :--- | :---: | :--- | :--- |
| **Suma Vectorial** | 1 | **Cerradura bajo la suma** | Si $\mathbf{u}, \mathbf{v} \in \mathbb{R}^2$, entonces $\mathbf{u} + \mathbf{v} \in \mathbb{R}^2$ |
| | 2 | **Conmutatividad** | $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$ |
| | 3 | **Asociatividad** | $(\mathbf{u} + \mathbf{v}) + \mathbf{w} = \mathbf{u} + (\mathbf{v} + \mathbf{w})$ |
| | 4 | **Identidad Aditiva (Neutro)** | Existe $\mathbf{0} \in \mathbb{R}^2$ tal que $\mathbf{u} + \mathbf{0} = \mathbf{u}$ |
| | 5 | **Inverso Aditivo** | Para cada $\mathbf{u}$, existe $-\mathbf{u} \in \mathbb{R}^2$ tal que $\mathbf{u} + (-\mathbf{u}) = \mathbf{0}$ |
| **Multiplicación por Escalar** | 6 | **Cerradura Escalar** | Si $\mathbf{u} \in \mathbb{R}^2$ y $c \in \mathbb{R}$, entonces $c\mathbf{u} \in \mathbb{R}^2$ |
| | 7 | **Distributividad (Escalar)** | $c(\mathbf{u} + \mathbf{v}) = c\mathbf{u} + c\mathbf{v}$ |
| | 8 | **Distributividad (Vectorial)** | $(c + d)\mathbf{u} = c\mathbf{u} + d\mathbf{u}$ |
| | 9 | **Asociatividad Escalar** | $c(d\mathbf{u}) = (cd)\mathbf{u}$ |
| | 10 | **Identidad Multiplicativa** | $1\mathbf{u} = \mathbf{u}$ |

> ⚠️ **Nota sobre la cantidad de axiomas:**
> Es común encontrar en algunos textos una lista de **8 axiomas** en lugar de 10. Esto se debe a que las propiedades de **Cerradura (1 y 6)** a menudo se consideran parte de la *definición* de las operaciones ("suma" y "producto") antes de enlistar los axiomas. Sin embargo, para fines de verificación rigurosa de subespacios, es crucial comprobar explícitamente estas dos condiciones de cerradura, sumando así las 10 propiedades.

> ✅ Estas propiedades garantizan que podemos operar libremente con vectores y escalares en $\mathbb{R}^2$.
