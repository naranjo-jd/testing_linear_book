## 🟦 Actividad 1: Escalar un vector

La operación más simple: multiplicar un vector por un número. ¿Qué cambia? ¿Qué permanece igual?

Ejecuta la celda y observa cómo cambia el vector al variar el escalar $c$.

```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([1.5, 1.0])  # vector original
escalares = [-2, -1, -0.5, 0.5, 1, 1.5, 2, 3]

fig, axes = plt.subplots(2, 4, figsize=(14, 6))
axes = axes.flatten()

for ax, c in zip(axes, escalares):
    cv = c * v
    ax.annotate('', xy=v, xytext=(0,0),
                arrowprops=dict(arrowstyle='->', color='steelblue', lw=2))
    ax.annotate('', xy=cv, xytext=(0,0),
                arrowprops=dict(arrowstyle='->', color='tomato', lw=2))
    ax.text(v[0]+0.05, v[1]+0.05, 'v', fontsize=11, color='steelblue')
    ax.text(cv[0]+0.05, cv[1]+0.05, f'{c}v', fontsize=11, color='tomato')
    ax.set_xlim(-3.5, 3.5)
    ax.set_ylim(-3.5, 3.5)
    ax.set_aspect('equal')
    ax.axhline(0, color='gray', lw=0.5)
    ax.axvline(0, color='gray', lw=0.5)
    ax.set_title(f'c = {c}', fontsize=10)
    ax.grid(True, alpha=0.2)

plt.suptitle('Escalamiento de un vector $v = (1.5, 1.0)$ por distintos valores de $c$',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print("Preguntas para reflexionar:")
print("  1. Cuando c > 0, ¿en qué dirección apunta c*v respecto a v?")
print("  2. Cuando c < 0, ¿qué ocurre con la dirección?")
print("  3. ¿Qué pasa cuando c = 0?")
```

## 🟩 Actividad 2: Rotar una figura

Vamos a rotar un polígono (un triángulo) por varios ángulos y observar qué **se conserva** y qué **cambia**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Triangulo original
triangulo = np.array([[0, 1, 0.5, 0],
                       [0, 0, 1,   0]], dtype=float)

def rotar(puntos, theta):
    c, s = np.cos(theta), np.sin(theta)
    R = np.array([[c, -s], [s, c]])
    return R @ puntos

angulos = [0, 45, 90, 135, 180, 225, 270, 315]
fig, axes = plt.subplots(2, 4, figsize=(14, 6))
axes = axes.flatten()

for ax, grados in zip(axes, angulos):
    theta = np.radians(grados)
    t_rot = rotar(triangulo, theta)
    # Area del triangulo original
    v1 = triangulo[:, 1] - triangulo[:, 0]
    v2 = triangulo[:, 2] - triangulo[:, 0]
    area = 0.5 * abs(v1[0]*v2[1] - v1[1]*v2[0])

    ax.fill(triangulo[0], triangulo[1], alpha=0.2, color='steelblue')
    ax.plot(*np.hstack([triangulo, triangulo[:,:1]]), 'b--', lw=1.5)
    ax.fill(t_rot[0], t_rot[1], alpha=0.45, color='tomato')
    ax.plot(*np.hstack([t_rot, t_rot[:,:1]]), 'r-', lw=2)
    ax.set_xlim(-1.6, 1.6)
    ax.set_ylim(-1.6, 1.6)
    ax.set_aspect('equal')
    ax.axhline(0, color='gray', lw=0.4)
    ax.axvline(0, color='gray', lw=0.4)
    ax.set_title(f'$\\theta = {grados}°$', fontsize=10)
    ax.text(0.03, 0.97, f'Área={area:.2f}', transform=ax.transAxes,
            fontsize=8, va='top')
    ax.grid(True, alpha=0.2)

plt.suptitle('Rotación de un triángulo — ¿Qué se conserva?', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print("Preguntas para reflexionar:")
print("  1. ¿Se conserva el área del triángulo? ¿Y el perímetro?")
print("  2. ¿Se conserva la forma del triángulo?")
print("  3. ¿Qué pasa con el origen (0,0) durante la rotación?")
print("  4. ¿Puede la rotación mapear el triángulo exactamente sobre sí mismo? ¿Para qué ángulo?")
```

## 🟨 Actividad 3: Transformación lineal vs. no lineal

Ahora comparamos dos funciones que transforman vectores de $\mathbb{R}^2$:

| Función | Definición | ¿Lineal? |
|---------|-----------|----------|
| $T_1$ | $T_1(x, y) = (2x,\; 3y)$ | ✅ Sí |
| $T_2$ | $T_2(x, y) = (x^2,\; y + 1)$ | ❌ No |

Observa las diferencias visuales y numérica.

```python
import numpy as np
import matplotlib.pyplot as plt

def T_lineal(v):
    """T_1(x,y) = (2x, 3y) — transformacion lineal"""
    return np.array([2*v[0], 3*v[1]])

def T_nolineal(v):
    """T_2(x,y) = (x^2, y+1) — NO lineal"""
    return np.array([v[0]**2, v[1] + 1])

# Probar la propiedad T(u+v) == T(u)+T(v)
u = np.array([1.0, 2.0])
v = np.array([3.0, -1.0])

print("=== PRUEBA DE ADITIVIDAD ===\n")
print("Para T_LINEAL: T_1(x,y) = (2x, 3y)")
print(f"  T_1(u+v)       = {T_lineal(u+v)}")
print(f"  T_1(u)+T_1(v)  = {T_lineal(u) + T_lineal(v)}")
print(f"  ¿Iguales? {np.allclose(T_lineal(u+v), T_lineal(u)+T_lineal(v))}")

print()
print("Para T_NO_LINEAL: T_2(x,y) = (x^2, y+1)")
print(f"  T_2(u+v)       = {T_nolineal(u+v)}")
print(f"  T_2(u)+T_2(v)  = {T_nolineal(u) + T_nolineal(v)}")
print(f"  ¿Iguales? {np.allclose(T_nolineal(u+v), T_nolineal(u)+T_nolineal(v))}")

# Visualizar el efecto sobre una grid
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Grid de puntos
xs = np.linspace(-1, 1, 8)
ys = np.linspace(-1, 1, 8)
XX, YY = np.meshgrid(xs, ys)
puntos = np.vstack([XX.ravel(), YY.ravel()])

p_lineal   = np.array([T_lineal(puntos[:,i]) for i in range(puntos.shape[1])]).T
p_nolineal = np.array([T_nolineal(puntos[:,i]) for i in range(puntos.shape[1])]).T

titles = ['Grid original', '$T_1$ lineal: $(2x, 3y)$', '$T_2$ no lineal: $(x^2, y+1)$']
datas  = [puntos, p_lineal, p_nolineal]
colors = ['steelblue', 'tomato', 'green']

for ax, data, title, col in zip(axes, datas, titles, colors):
    ax.scatter(data[0], data[1], c=col, s=18, alpha=0.7)
    ax.set_aspect('equal')
    ax.set_title(title, fontsize=11)
    ax.axhline(0, color='gray', lw=0.5)
    ax.axvline(0, color='gray', lw=0.5)
    ax.grid(True, alpha=0.3)

plt.suptitle('Transformación lineal vs no lineal sobre una grilla de puntos', fontsize=12)
plt.tight_layout()
plt.show()
```

## 🧩 Actividad 4: El origen siempre se queda

Una propiedad **necesaria** de toda transformación lineal: $T(\mathbf{0}) = \mathbf{0}$. Esto significa que el origen siempre se mapea al origen. Si una función no cumple esto, ¡definitivamente no es lineal!

```python
import numpy as np

cero = np.array([0.0, 0.0])

funciones = [
    ("T(x,y) = (2x, 3y)",     lambda v: np.array([2*v[0], 3*v[1]])),
    ("T(x,y) = (x+1, y)",     lambda v: np.array([v[0]+1, v[1]])),   # NO lineal
    ("T(x,y) = (x*y, x+y)",   lambda v: np.array([v[0]*v[1], v[0]+v[1]])),  # NO lineal
    ("T(x,y) = (0, 0)",        lambda v: np.array([0.0, 0.0])),      # Lineal (T cero)
    ("T(x,y) = (x, y)",        lambda v: np.array([v[0], v[1]])),    # Lineal (identidad)
    ("T(x,y) = (x^2, y^2)",    lambda v: np.array([v[0]**2, v[1]**2])),  # NO lineal
]

print("Prueba necesaria: T(0) debe ser 0 para ser lineal\n")
print(f"{'Funcion':<30} {'T(0,0)':<20} {'T(0)=0?':<10}")
print("-" * 60)
for nombre, f in funciones:
    resultado = f(cero)
    es_cero = np.allclose(resultado, cero)
    marca = '✅ Posiblemente' if es_cero else '❌ Definitivamente NO'
    print(f"{nombre:<30} {str(resultado):<20} {marca}")

print("\nNota: T(0)=0 es NECESARIA pero no SUFICIENTE para garantizar linealidad.")
print("Por ejemplo, T(x,y) = (x*y, x+y) cumple T(0,0)=(0,0) pero NO es lineal.")
```
