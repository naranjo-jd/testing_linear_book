# 1.1 Vectores en $\mathbb{R}^2$

Los vectores son objetos matemáticos fundamentales que nos permiten representar cantidades que poseen tanto magnitud como dirección. En este capítulo, exploraremos los vectores en el plano bidimensional $\mathbb{R}^2$, sentando las bases para comprender espacios vectoriales más generales. Los vectores aparecen en múltiples contextos: desde la física (fuerzas, velocidades, desplazamientos) hasta la computación gráfica, el aprendizaje automático y la economía.

**Motivación:** ¿Cómo comparar rutas de movimiento en el plano cuando cada una tiene dirección y longitud distintas? Describir solo la distancia o solo el sentido no permite decidir ni combinar trayectorias de forma rigurosa. Necesitamos un objeto matemático que reúna en una sola estructura la magnitud y la dirección del desplazamiento. Esa necesidad introduce el concepto de **vector** en $\mathbb{R}^2$, punto de partida de toda la teoría que sigue.

## 1.1.1. Definición de $\mathbb{R}^2$

:::{admonition} Definición: El espacio $\mathbb{R}^2$
:class: tip

El conjunto $\mathbb{R}^2$ se define como el conjunto de todos los pares ordenados de números reales:

$$\mathbb{R}^2 = \{(x, y) \mid x, y \in \mathbb{R}\}$$

Cada elemento $(x, y) \in \mathbb{R}^2$ se denomina **vector** o **punto**, dependiendo del contexto de interpretación.
:::

El conjunto $\mathbb{R}^2$ puede visualizarse como el plano cartesiano, donde:
- La primera componente $x$ corresponde a la posición horizontal (eje de las abscisas)
- La segunda componente $y$ corresponde a la posición vertical (eje de las ordenadas)

Por ejemplo, $(3, 2)$, $(-1, 4)$ y $(0, 0)$ son elementos de $\mathbb{R}^2$.

## 1.1.2. Representación geométrica

Un par ordenado $(x, y) \in \mathbb{R}^2$ admite dos interpretaciones geométricas fundamentales:

### Interpretación como punto
El par $(x, y)$ representa una **ubicación** en el plano cartesiano. Es simplemente una posición, sin dirección ni magnitud asociada.

### Interpretación como vector
El par $(x, y)$ representa una **flecha** con:
- **Punto inicial** (origen de la flecha)
- **Punto terminal** (extremo de la flecha)
- **Magnitud** (longitud de la flecha)
- **Dirección** (hacia dónde apunta)

:::{admonition} Vector posición vs Vector libre
:class: note

- **Vector fijo o ligado**: Existen situaciones donde el punto inicial de un vector es importante. A estos vectores se les llama vectores fijos o ligados, y cuando son usados se hacen explícitos sus puntos de inicio y fin. A un vector fijo que comienza en el punto $A$ y termina en el punto $B$ se le denota comunmente con $\overrightarrow{AB}$
  
- **Vector libre**: Cuando el origen de un vector no es relevante, se le llama vector libre. Estos vectores pueden trasladarse a cualquier posición del plano sin perder sus propiedades. Dos vectores libres son equivalentes si tienen la misma magnitud y dirección, independientemente de su ubicación.
:::

La siguiente visualización muestra un vector posición con sus componentes:

````python
import numpy as np
import matplotlib.pyplot as plt

# Configuración del gráfico
fig, ax = plt.subplots(figsize=(8, 6))

# Definir el vector
v = np.array([3, 2])

# Dibujar el vector como flecha
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', width=0.02, label=r'$\mathbf{v} = (3, 2)$')

# Dibujar líneas punteadas para las componentes
ax.plot([v[0], v[0]], [0, v[1]], 'b--', alpha=0.5, linewidth=1.5)
ax.plot([0, v[0]], [v[1], v[1]], 'b--', alpha=0.5, linewidth=1.5)

# Etiquetas de componentes
ax.annotate(r'$v_1 = 3$', xy=(1.5, -0.4), fontsize=12, ha='center', color='darkblue')
ax.annotate(r'$v_2 = 2$', xy=(3.4, 1), fontsize=12, ha='left', color='darkblue')

# Marcar el punto terminal
ax.plot(v[0], v[1], 'bo', markersize=8)
ax.annotate(r'$(3, 2)$', xy=(v[0]+0.15, v[1]+0.15), fontsize=11)

# Marcar el origen
ax.plot(0, 0, 'ko', markersize=6)
ax.annotate('Origen', xy=(-0.3, -0.4), fontsize=10)

# Configuración de ejes
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12)
ax.set_title('Representación geométrica de un vector en $\mathbb{R}^2$', fontsize=14)
ax.legend(loc='upper left', fontsize=11)

plt.tight_layout()
plt.show()
````

## 1.1.3. Notación

Existen varias formas de denotar un vector en $\mathbb{R}^2$:

- **Notación con letra en negrita:** los vectores se representan con letras minúsculas en negrita, por ejemplo  
  $\mathbf{v}, \mathbf{u}, \mathbf{w}$.  
  En escritura a mano suele usarse una flecha: $\vec{v}, \vec{u}, \vec{w}$.

- **Notación por componentes (par ordenado):**   
  $\mathbf{v} = (v_1, v_2)$
  donde $v_1$ es la primera componente (horizontal) y $v_2$ la segunda (vertical).

- **Notación como vector columna:**  
  $\mathbf{v} = \begin{pmatrix} v_1 \\ v_2 \end{pmatrix}$  
  Es la más usada en álgebra lineal, especialmente con matrices y transformaciones lineales.

- **Notación como vector fila:**  
  $\mathbf{v} = \begin{pmatrix} v_1 & v_2 \end{pmatrix}$

:::{admonition} Convención importante
:class: warning

En este curso, cuando escribamos un vector como $(v_1, v_2)$, lo interpretaremos por defecto como un **vector columna**, a menos que se indique lo contrario explícitamente.
:::

## 1.1.4.  Exploración Interactiva

Utiliza los controles deslizantes para modificar las componentes del vector y observa cómo cambia su representación gráfica en tiempo real.

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def graficar_vector(x, y):
    """
    Grafica un vector desde el origen hasta el punto (x, y).
    """
    fig, ax = plt.subplots(figsize=(7, 7))
    
    # Dibujar el vector
    ax.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1,
              color='blue', width=0.02)
    
    # Dibujar líneas punteadas para las componentes
    ax.plot([x, x], [0, y], 'b--', alpha=0.4, linewidth=1)
    ax.plot([0, x], [y, y], 'b--', alpha=0.4, linewidth=1)
    
    # Marcar el punto terminal
    ax.plot(x, y, 'bo', markersize=8)
    ax.annotate(f'({x:.1f}, {y:.1f})', xy=(x + 0.2, y + 0.2), fontsize=11)
    
    # Configuración de ejes
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.8)
    ax.axvline(x=0, color='k', linewidth=0.8)
    ax.set_xlabel('$x$', fontsize=12)
    ax.set_ylabel('$y$', fontsize=12)
    ax.set_title(f'Vector $\\mathbf{{v}} = ({x:.1f}, {y:.1f})$', fontsize=14)
    
    # Mostrar información adicional
    magnitud = np.sqrt(x**2 + y**2)
    ax.text(-5.5, 5.5, f'Magnitud: {magnitud:.2f}', fontsize=11, 
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.show()

# Crear sliders interactivos
slider_x = FloatSlider(value=2.0, min=-5.0, max=5.0, step=0.1, description='x:')
slider_y = FloatSlider(value=3.0, min=-5.0, max=5.0, step=0.1, description='y:')

interact(graficar_vector, x=slider_x, y=slider_y);
````

## 1.1.5. Igualdad de vectores

:::{admonition} Definición: Igualdad de vectores
:class: tip

Dos vectores $\mathbf{u} = (u_1, u_2)$ y $\mathbf{v} = (v_1, v_2)$ en $\mathbb{R}^2$ son **iguales** si y solo si sus componentes correspondientes son iguales:

$$\mathbf{u} = \mathbf{v} \quad \Longleftrightarrow \quad u_1 = v_1 \text{ y } u_2 = v_2$$
:::

Esta definición implica que el orden de las componentes importa: $(3, 5) \neq (5, 3)$.

**Ejemplo:**

Determinar si los siguientes pares de vectores son iguales:

1. $\mathbf{u} = (2, 3)$ y $\mathbf{v} = (2, 3)$
   
   **Solución:** $u_1 = 2 = v_1$ y $u_2 = 3 = v_2$, por lo tanto $\mathbf{u} = \mathbf{v}$. ✓

2. $\mathbf{u} = (1, 4)$ y $\mathbf{v} = (4, 1)$
   
   **Solución:** $u_1 = 1 \neq 4 = v_1$, por lo tanto $\mathbf{u} \neq \mathbf{v}$. ✗

3. $\mathbf{u} = (\sqrt{4}, 6/2)$ y $\mathbf{v} = (2, 3)$
   
   **Solución:** $u_1 = \sqrt{4} = 2 = v_1$ y $u_2 = 6/2 = 3 = v_2$, por lo tanto $\mathbf{u} = \mathbf{v}$. ✓

## 1.1.6.  Ejercicio Rápido

:::{admonition} Verifica tu comprensión
:class: note

**Pregunta:** Si $\mathbf{u} = (x+1, 3)$ y $\mathbf{v} = (4, y-2)$ son vectores iguales, ¿cuáles son los valores de $x$ e $y$?

```{toggle}
**Solución:**

Por la definición de igualdad de vectores:
- $x + 1 = 4 \Rightarrow x = 3$
- $3 = y - 2 \Rightarrow y = 5$

Por lo tanto, $x = 3$ e $y = 5$.
```
:::
