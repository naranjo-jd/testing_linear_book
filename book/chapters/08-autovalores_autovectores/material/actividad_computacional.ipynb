# 🚀 Actividad Computacional — Mini-Proyecto
## Visualizador de Autovectores y Transformaciones Lineales

### 🎯 Objetivo

En esta actividad construirás una **herramienta de visualización completa** que, dada
una matriz $2\times 2$:

1. Calcula autovalores y autovectores.
2. Dibuja la imagen del círculo unitario (elipse de transformación).
3. Superpone los autovectores como flechas coloreadas.
4. Muestra cómo cada autovector se escala por su autovalor.
5. Presenta una tabla comparativa de propiedades.

---

### 📋 Matrices a analizar

| Matriz | Tipo | Autovalores esperados |
|:---:|:---|:---|
| $A_1 = \begin{pmatrix}3&0\\0&1\end{pmatrix}$ | Estiramiento puro | Reales: 3, 1 |
| $A_2 = \begin{pmatrix}2&1\\1&2\end{pmatrix}$ | Simétrica | Reales: 3, 1 |
| $A_3 = \begin{pmatrix}1&-1\\1&1\end{pmatrix}$ | Rotación + escala | Complejos: $1\pm i$ |

## 🛠️ Paso 1: Función Visualizadora Principal

Primero construimos la función central que acepta cualquier matriz $2\times 2$ y produce
la visualización completa.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec

def visualizar_transformacion(A, titulo='', ax_principal=None, mostrar_tabla=True):
    """
    Visualiza una transformación lineal 2D:
    - Círculo unitario y su imagen
    - Autovectores (cuando son reales)
    - Vectores de muestra y sus imágenes
    Retorna: autovalores, autovectores
    """
    eigenvalues, eigenvectors = np.linalg.eig(A)

    if ax_principal is None:
        fig, ax_principal = plt.subplots(figsize=(7, 6))

    ax = ax_principal

    # ── Círculo unitario y su imagen ────────────────
    t = np.linspace(0, 2*np.pi, 300)
    circle = np.vstack([np.cos(t), np.sin(t)])
    ellipse = A @ circle

    ax.plot(circle[0], circle[1], 'b--', lw=1.5, alpha=0.5, label='Círculo unitario')
    ax.plot(ellipse[0], ellipse[1], 'g-', lw=2.5, alpha=0.8, label='Imagen bajo A')

    # ── Autovectores (solo reales) ──────────────────
    colors_ev = ['#e74c3c', '#9b59b6']
    legend_patches = [
        mpatches.Patch(color='blue', alpha=0.5, label='Círculo unitario'),
        mpatches.Patch(color='green', alpha=0.8, label='Imagen bajo A'),
    ]

    for i, (lam, ev) in enumerate(zip(eigenvalues, eigenvectors.T)):
        if abs(lam.imag) < 1e-10:  # autovalor real
            ev_r = ev.real
            ev_n = ev_r / np.linalg.norm(ev_r)
            lam_r = lam.real

            # Autovector unitario (dirección original)
            ax.annotate('', xy=ev_n, xytext=(0, 0),
                        arrowprops=dict(arrowstyle='->', color=colors_ev[i], lw=3))
            # Vector transformado = λ*v
            if abs(lam_r) > 0.05:
                ax.annotate('', xy=lam_r*ev_n, xytext=(0, 0),
                            arrowprops=dict(arrowstyle='->', color=colors_ev[i],
                                            lw=2, linestyle='dashed', alpha=0.7))
            ax.text(ev_n[0]*1.15, ev_n[1]*1.15,
                    f'$\\mathbf{{v}}_{i+1}$\n$\\lambda={lam_r:.2f}$',
                    color=colors_ev[i], fontsize=10, ha='center', fontweight='bold')
            legend_patches.append(
                mpatches.Patch(color=colors_ev[i],
                               label=f'$v_{i+1}$ (λ={lam_r:.2f})'))
        else:
            ax.text(0, -1.4,
                    f'$\\lambda_{i+1} = {lam.real:.2f}{lam.imag:+.2f}i$ (complejo)',
                    color=colors_ev[i], fontsize=9, ha='center', style='italic')

    # ── Algunos vectores de muestra ─────────────────
    sample_angles = np.linspace(0, 2*np.pi, 8, endpoint=False)
    for a in sample_angles:
        v = np.array([np.cos(a), np.sin(a)])
        Av = A @ v
        ax.annotate('', xy=v*0.85, xytext=(0,0),
                    arrowprops=dict(arrowstyle='->', color='lightblue', lw=1, alpha=0.6))
        ax.annotate('', xy=Av*0.85, xytext=(0,0),
                    arrowprops=dict(arrowstyle='->', color='lightgreen', lw=1, alpha=0.6))

    ax.axhline(0, color='k', lw=0.8)
    ax.axvline(0, color='k', lw=0.8)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.legend(handles=legend_patches, fontsize=9, loc='upper right')
    ax.set_title(titulo, fontsize=11)
    ax.set_xlabel('x'); ax.set_ylabel('y')

    return eigenvalues, eigenvectors

print('✅ Función visualizar_transformacion definida')
print('Parámetros: A (matriz 2x2), titulo (str), ax_principal (Axes)')
```

## 📊 Paso 2: Análisis de la Matriz 1 — Estiramiento Puro

$$A_1 = \begin{pmatrix} 3 & 0 \\ 0 & 1 \end{pmatrix}$$

Esta es una matriz diagonal. Las predicciones son:
- $\lambda_1 = 3$, $\mathbf{v}_1 = (1, 0)$ — el eje $x$ se estira 3 veces.
- $\lambda_2 = 1$, $\mathbf{v}_2 = (0, 1)$ — el eje $y$ no cambia.

```python
import numpy as np
import matplotlib.pyplot as plt

A1 = np.array([[3.0, 0.0], [0.0, 1.0]])

fig, ax = plt.subplots(figsize=(8, 7))
evals, evecs = visualizar_transformacion(
    A1,
    titulo='$A_1 = \\mathrm{diag}(3, 1)$ — Estiramiento puro\n'
           'El círculo se convierte en elipse con semiejes 3 y 1',
    ax_principal=ax
)

ax.set_xlim(-3.5, 3.5); ax.set_ylim(-2, 2)
plt.tight_layout()
plt.savefig('act_comp_A1.png', dpi=100, bbox_inches='tight')
plt.show()

print('\nAnálisis de A1:')
print(f'  Autovalores: {np.round(evals.real, 4)}')
print(f'  Autovectores:\n{np.round(evecs.real, 4)}')
print(f'  tr(A1) = {np.trace(A1)} = {evals[0].real:.1f} + {evals[1].real:.1f} ✅')
print(f'  det(A1) = {np.linalg.det(A1):.1f} = {evals[0].real:.1f} × {evals[1].real:.1f} ✅')
print()
print('  Interpretación: el círculo unitario se transforma en una ELIPSE')
print('  con semieje mayor = λ₁ = 3 (dirección x) y semieje menor = λ₂ = 1 (dirección y)')
```

## 📊 Paso 3: Análisis de la Matriz 2 — Simétrica con Autovectores Oblicuos

$$A_2 = \begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}$$

Como es **simétrica**, sus autovectores son **ortogonales** entre sí.
Los autovalores son $\lambda = 3$ y $\lambda = 1$.

```python
import numpy as np
import matplotlib.pyplot as plt

A2 = np.array([[2.0, 1.0], [1.0, 2.0]])

fig, axes = plt.subplots(1, 2, figsize=(13, 6))

evals, evecs = visualizar_transformacion(
    A2,
    titulo='$A_2 = \\begin{pmatrix}2&1\\\\1&2\\end{pmatrix}$ — Simétrica\n'
           'Autovectores ORTOGONALES a 45° y 135°',
    ax_principal=axes[0]
)
axes[0].set_xlim(-3.5, 3.5); axes[0].set_ylim(-3.5, 3.5)

# Panel derecho: verificar ortogonalidad
ax2 = axes[1]
v1 = evecs[:, 0].real; v1 /= np.linalg.norm(v1)
v2 = evecs[:, 1].real; v2 /= np.linalg.norm(v2)

# Mostrar los autovectores y el ángulo entre ellos
ax2.annotate('', xy=v1, xytext=(0,0),
             arrowprops=dict(arrowstyle='->', color='#e74c3c', lw=2.5))
ax2.annotate('', xy=v2, xytext=(0,0),
             arrowprops=dict(arrowstyle='->', color='#9b59b6', lw=2.5))
ax2.text(v1[0]+0.05, v1[1]+0.05, f'$\\mathbf{{v}}_1 = ({v1[0]:.2f}, {v1[1]:.2f})$',
         color='#e74c3c', fontsize=11)
ax2.text(v2[0]+0.05, v2[1]-0.1, f'$\\mathbf{{v}}_2 = ({v2[0]:.2f}, {v2[1]:.2f})$',
         color='#9b59b6', fontsize=11)

# Ángulo recto
dot = np.dot(v1, v2)
ax2.text(0, -1.2, f'$\\mathbf{{v}}_1 \\cdot \\mathbf{{v}}_2 = {dot:.4f} \\approx 0$ ✅',
         ha='center', fontsize=12)
ax2.set_xlim(-1.5, 1.5); ax2.set_ylim(-1.5, 1.5)
ax2.axhline(0, color='k', lw=0.5); ax2.axvline(0, color='k', lw=0.5)
ax2.set_aspect('equal'); ax2.grid(True, alpha=0.3)
ax2.set_title('Ortogonalidad de autovectores\n(Teorema Espectral para matrices simétricas)', fontsize=10)

plt.suptitle('Análisis completo de $A_2$', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('act_comp_A2.png', dpi=100, bbox_inches='tight')
plt.show()

print(f'Autovalores reales: {np.round(evals.real, 4)} ✅')
print(f'Producto punto v1·v2 = {np.dot(v1,v2):.2e} ≈ 0 → ortogonales ✅')
```

## 📊 Paso 4: Análisis de la Matriz 3 — Rotación + Escala (Autovalores Complejos)

$$A_3 = \begin{pmatrix} 1 & -1 \\ 1 & 1 \end{pmatrix}$$

Esta matriz combina **rotación de 45°** con **escala** $\sqrt{2}$.

Sus autovalores son complejos: $\lambda = 1 \pm i$, con módulo $|\lambda| = \sqrt{2}$.

> 🔑 Cuando los autovalores son complejos, **no hay** autovectores reales.
> La transformación rota TODOS los vectores — ninguna dirección se preserva.

```python
import numpy as np
import matplotlib.pyplot as plt

A3 = np.array([[1.0, -1.0], [1.0, 1.0]])

fig, axes = plt.subplots(1, 2, figsize=(13, 6))

evals, evecs = visualizar_transformacion(
    A3,
    titulo='$A_3 = \\begin{pmatrix}1&-1\\\\1&1\\end{pmatrix}$ — Rotación + escala\n'
           'NO hay autovectores reales',
    ax_principal=axes[0]
)
axes[0].set_xlim(-2, 2); axes[0].set_ylim(-2, 2)

# Panel derecho: mostrar la acción sobre algunos vectores
ax2 = axes[1]
angles = np.linspace(0, 2*np.pi, 12, endpoint=False)
colors = plt.cm.hsv(np.linspace(0, 1, 12))

for a, color in zip(angles, colors):
    v = np.array([np.cos(a), np.sin(a)])
    Av = A3 @ v
    ax2.annotate('', xy=v, xytext=(0,0),
                 arrowprops=dict(arrowstyle='->', color=color, lw=2, alpha=0.7))
    ax2.annotate('', xy=Av*0.5, xytext=(0,0),
                 arrowprops=dict(arrowstyle='->', color=color, lw=1.5, linestyle='--', alpha=0.5))

ax2.axhline(0, color='k', lw=0.5); ax2.axvline(0, color='k', lw=0.5)
ax2.set_xlim(-1.5, 1.5); ax2.set_ylim(-1.5, 1.5)
ax2.set_aspect('equal'); ax2.grid(True, alpha=0.3)
ax2.set_title('TODOS los vectores rotan\n(ninguna dirección es invariante)', fontsize=10)

plt.suptitle('Análisis completo de $A_3$ — Autovalores complejos', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('act_comp_A3.png', dpi=100, bbox_inches='tight')
plt.show()

print('Autovalores de A3 (complejos):')
for i, lam in enumerate(evals):
    print(f'  λ_{i+1} = {lam.real:.4f} + {lam.imag:.4f}i  (módulo = {abs(lam):.4f} = √2 ✅)')
print()
print('La transformación: rota cada vector 45° y escala su longitud por √2.')
print('Como rota, no hay ninguna dirección que se preserve → no hay autovectores reales.')
```

## 📊 Paso 5: Tabla Comparativa de las Tres Matrices

Comparemos las tres matrices de forma sistemática.

```python
import numpy as np
import matplotlib.pyplot as plt

matrices_info = [
    (np.array([[3.0, 0.0], [0.0, 1.0]]),  'A1 = diag(3,1)',      'Estiramiento puro'),
    (np.array([[2.0, 1.0], [1.0, 2.0]]),  'A2 = [[2,1],[1,2]]',  'Simétrica'),
    (np.array([[1.0,-1.0], [1.0, 1.0]]),  'A3 = [[1,-1],[1,1]]', 'Rotación+escala'),
]

print('TABLA COMPARATIVA')
print('='*80)
header = f'{"Matriz":<22} {"tr(A)":<7} {"det(A)":<8} {"Autovalores":<30} {"Tipo":<15}'
print(header)
print('-'*80)

resultados = []
for A, nombre, tipo in matrices_info:
    evals = np.linalg.eigvals(A)
    tr = np.trace(A)
    det = np.linalg.det(A)
    complejos = any(abs(e.imag) > 1e-10 for e in evals)
    evals_str = ', '.join([f'{e.real:.3f}+{e.imag:.3f}i' if abs(e.imag)>1e-10
                           else f'{e.real:.3f}' for e in evals])
    tipo_ev = 'COMPLEJOS' if complejos else 'reales'
    print(f'{nombre:<22} {tr:<7.2f} {det:<8.2f} {evals_str:<30} {tipo_ev}')
    resultados.append((nombre, tipo, evals, tr, det))

print('='*80)
print()

# Verificar tr = Σλ y det = Πλ
print('Verificación de propiedades:')
for nombre, tipo, evals, tr, det in resultados:
    sum_lam = np.sum(evals).real
    prod_lam = np.prod(evals).real
    ok_tr = '✅' if np.isclose(tr, sum_lam) else '❌'
    ok_det = '✅' if np.isclose(det, prod_lam) else '❌'
    print(f'  {nombre}: tr={tr:.2f}=Σλ={sum_lam:.2f}{ok_tr}, det={det:.2f}=Πλ={prod_lam:.2f}{ok_det}')
```

## 🔬 Paso 6: Análisis de Diagonalizabilidad

¿Cuáles de las tres matrices son diagonalizables?

```python
import numpy as np

def analizar_diagonalizabilidad(A, nombre):
    """Analiza si una matriz 2x2 es diagonalizable y construye la factorización."""
    print(f'\n{nombre}:')
    evals, evecs = np.linalg.eig(A)

    # Verificar si los autovalores son reales
    reales = all(abs(e.imag) < 1e-10 for e in evals)
    print(f'  Autovalores reales: {reales}')

    if not reales:
        print('  → NO diagonalizable sobre ℝ (autovalores complejos)')
        return False

    # Verificar independencia lineal de autovectores
    P = evecs
    det_P = np.linalg.det(P)
    li = not np.isclose(det_P, 0)
    print(f'  det(P) = {det_P:.6f} → autovectores LI: {li}')

    if li:
        # Construir factorización
        D = np.diag(evals.real)
        P_inv = np.linalg.inv(P)
        A_check = (P @ D @ P_inv).real
        ok = np.allclose(A, A_check)
        print(f'  Diagonalizable: ✅')
        print(f'  Verificación A = PDP⁻¹: {ok} ✅')
        return True
    else:
        print('  → NO diagonalizable (autovectores LI insuficientes)')
        return False

A1 = np.array([[3.0, 0.0], [0.0, 1.0]])
A2 = np.array([[2.0, 1.0], [1.0, 2.0]])
A3 = np.array([[1.0,-1.0], [1.0, 1.0]])

print('ANÁLISIS DE DIAGONALIZABILIDAD')
print('='*55)
d1 = analizar_diagonalizabilidad(A1, 'A1 = diag(3,1)')
d2 = analizar_diagonalizabilidad(A2, 'A2 = [[2,1],[1,2]]')
d3 = analizar_diagonalizabilidad(A3, 'A3 = [[1,-1],[1,1]]')

print(f'\nResumen: A1={d1}, A2={d2}, A3={d3}')
```

### 💡 Lecciones aprendidas

1. **La visualización del círculo** → elipse revela inmediatamente la estructura de la transformación.
2. **Los autovectores reales** se identifican como las direcciones que la elipse comparte con el círculo (semi-ejes alineados).
3. **Autovalores complejos** ↔ la transformación incluye rotación → no hay direcciones invariantes reales.
4. **Matrices simétricas** siempre tienen autovectores ortogonales (Teorema Espectral).
