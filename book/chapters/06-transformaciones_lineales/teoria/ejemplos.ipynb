# 🔍 Ejemplos de Transformaciones Lineales

En este notebook trabajamos **cuatro ejemplos detallados** de transformaciones lineales en $\mathbb{R}^2$ y $\mathbb{R}^3$. Para cada uno derivamos la matriz de representación, calculamos núcleo e imagen, y visualizamos el efecto geométrico.

> **Prerrequisito:** haber estudiado `concepto.ipynb`.

## 🟦 Ejemplo 1: Rotación en $\mathbb{R}^2$

La **rotación** por un ángulo $\theta$ en sentido antihorario es la transformación lineal $R_\theta: \mathbb{R}^2 \to \mathbb{R}^2$ cuya matriz es:

$$R_\theta = \begin{pmatrix} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{pmatrix}$$

**Propiedades:**
- $\det(R_\theta) = \cos^2\theta + \sin^2\theta = 1$ → preserva áreas y orientación
- $R_\theta^{-1} = R_{-\theta}$ (rotar en sentido contrario)
- $\ker(R_\theta) = \{\mathbf{0}\}$ → es inyectiva
- $\operatorname{Im}(R_\theta) = \mathbb{R}^2$ → es suprayectiva
- En consecuencia: $R_\theta$ es **biyectiva** para todo $\theta$.

**Interpretación:** La rotación preserva distancias (es una isometría) y transforma líneas en líneas, círculos en círculos.

```python
import numpy as np
import matplotlib.pyplot as plt

def rotacion(theta):
    """Devuelve la matriz de rotacion de angulo theta (en radianes)."""
    c, s = np.cos(theta), np.sin(theta)
    return np.array([[c, -s], [s, c]])

theta = np.pi / 3  # 60 grados
R = rotacion(theta)
print(f"Matriz R_{{60°}}:\n{np.round(R, 4)}")
print(f"det(R) = {np.round(np.linalg.det(R), 6)} (debe ser 1)")

# Vectores a rotar
vectores = np.array([[1, 0], [0, 1], [1, 1], [-1, 0.5]]).T  # columnas
vectores_rot = R @ vectores

fig, ax = plt.subplots(figsize=(7, 7))
colors = ['steelblue', 'tomato', 'green', 'purple']
for i in range(vectores.shape[1]):
    v = vectores[:, i]
    vr = vectores_rot[:, i]
    ax.annotate('', xy=v, xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color=colors[i], lw=2))
    ax.annotate('', xy=vr, xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color=colors[i], lw=2, linestyle='dashed'))
    ax.text(v[0]*1.1, v[1]*1.1, f'$v_{i+1}$', fontsize=11, color=colors[i])
    ax.text(vr[0]*1.1, vr[1]*1.1, f'$R(v_{i+1})$', fontsize=11, color=colors[i])

# Dibujar arco de angulo
angulo_arco = np.linspace(0, theta, 50)
r_arco = 0.3
ax.plot(r_arco*np.cos(angulo_arco), r_arco*np.sin(angulo_arco), 'k-', lw=1)
ax.text(0.35, 0.08, f'$\\theta={60}°$', fontsize=10)

ax.set_xlim(-1.8, 1.8)
ax.set_ylim(-1.8, 1.8)
ax.set_aspect('equal')
ax.axhline(0, color='k', lw=0.7)
ax.axvline(0, color='k', lw=0.7)
ax.grid(True, alpha=0.3)
ax.set_title(f'Rotación de $\\theta = 60°$ en $\\mathbb{{R}}^2$', fontsize=13)
ax.text(0.05, 0.95, 'Continuo: original  |  Discontinuo: rotado',
        transform=ax.transAxes, fontsize=9, va='top')
plt.tight_layout()
plt.show()
```

## 🟩 Ejemplo 2: Proyección ortogonal sobre una recta $y = mx$

La **proyección ortogonal** de $\mathbb{R}^2$ sobre la recta $\ell: y = mx$ es la transformación que lleva cada vector al punto más cercano en $\ell$.

El vector unitario en dirección de $\ell$ es $\hat{u} = \frac{1}{\sqrt{1+m^2}}(1, m)$. La proyección es:

$$P_\ell(\mathbf{x}) = (\hat{u}\cdot\mathbf{x})\hat{u} = \hat{u}\hat{u}^\top \mathbf{x}$$

La matriz de proyección es:

$$P = \frac{1}{1+m^2}\begin{pmatrix} 1 & m \\ m & m^2 \end{pmatrix}$$

**Propiedades importantes:**
- $P^2 = P$ (proyectar dos veces es lo mismo que proyectar una)
- $\ker(P) = \ell^\perp$ (la recta perpendicular)
- $\operatorname{Im}(P) = \ell$
- $\operatorname{nulidad} = 1$, $\operatorname{rango} = 1$, y $1 + 1 = 2 = \dim(\mathbb{R}^2)$ ✓

```python
import numpy as np
import matplotlib.pyplot as plt

def matriz_proyeccion(m):
    """Matriz de proyeccion ortogonal sobre y = m*x."""
    return (1/(1 + m**2)) * np.array([[1, m], [m, m**2]])

m = 1.5  # pendiente de la recta
P = matriz_proyeccion(m)
print(f"Matriz de proyeccion sobre y = {m}x:")
print(np.round(P, 4))
print(f"\nVerificacion P^2 = P: {np.allclose(P @ P, P)}")
print(f"Rango de P: {np.linalg.matrix_rank(P)}")

# Vectores a proyectar
vectores = np.array([[2, 0.5], [-1, 2], [0, 2], [1.5, -1]]).T
proyecciones = P @ vectores

fig, ax = plt.subplots(figsize=(7, 7))

# Dibujar la recta y = m*x
t = np.linspace(-2.5, 2.5, 100)
ax.plot(t, m*t, 'k-', lw=2, label=f'$\\ell: y={m}x$', zorder=1)

colors = ['steelblue', 'tomato', 'green', 'purple']
for i in range(vectores.shape[1]):
    v = vectores[:, i]
    pv = proyecciones[:, i]
    # Vector original
    ax.annotate('', xy=v, xytext=(0,0),
                arrowprops=dict(arrowstyle='->', color=colors[i], lw=1.8))
    # Proyeccion
    ax.annotate('', xy=pv, xytext=(0,0),
                arrowprops=dict(arrowstyle='->', color=colors[i], lw=2.5))
    # Linea perpendicular
    ax.plot([v[0], pv[0]], [v[1], pv[1]], '--', color=colors[i], lw=1, alpha=0.7)
    ax.text(v[0]+0.05, v[1]+0.05, f'$v_{i+1}$', fontsize=10, color=colors[i])

ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_aspect('equal')
ax.axhline(0, color='gray', lw=0.5)
ax.axvline(0, color='gray', lw=0.5)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)
ax.set_title(f'Proyección ortogonal sobre $y = {m}x$', fontsize=13)
plt.tight_layout()
plt.show()
```

## 🟨 Ejemplo 3: Transformación de cizallamiento (*shear*)

El **cizallamiento horizontal** con parámetro $k$ es la transformación $S_k: \mathbb{R}^2 \to \mathbb{R}^2$ con matriz:

$$S_k = \begin{pmatrix} 1 & k \\ 0 & 1 \end{pmatrix}$$

Su efecto: $(x, y) \mapsto (x + ky, y)$. La componente horizontal se desplaza proporcionalmente a la altura $y$.

**Propiedades:**
- $\det(S_k) = 1$ → preserva áreas
- Es biyectiva para todo $k$ (su inversa es $S_{-k}$)
- Transforma cuadrados en paralelogramos
- Preserva líneas horizontales ($y = $ const)

Existe también el **cizallamiento vertical**: $\begin{pmatrix}1 & 0 \\ k & 1\end{pmatrix}$, que desplaza la componente vertical.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(14, 5))

# Cuadrado unitario con puntos interiores
cuadrado = np.array([[0,1,1,0,0],[0,0,1,1,0]], dtype=float)
# Grid interior
xs = np.linspace(0, 1, 6)
grid_h = [(np.array([0,1]), np.array([y,y])) for y in xs]  # lineas horizontales
grid_v = [(np.array([x,x]), np.array([0,1])) for x in xs]  # lineas verticales

valores_k = [0.5, 1.0, 2.0]

for ax, k in zip(axes, valores_k):
    S = np.array([[1, k], [0, 1]])
    cuad_t = S @ cuadrado

    ax.fill(cuadrado[0], cuadrado[1], alpha=0.2, color='steelblue')
    ax.plot(cuadrado[0], cuadrado[1], 'b--', lw=1.5, label='Original')
    ax.fill(cuad_t[0], cuad_t[1], alpha=0.35, color='tomato')
    ax.plot(cuad_t[0], cuad_t[1], 'r-', lw=2, label='Cizallado')

    # Grid transformado
    for gx, gy in grid_h:
        pts = S @ np.array([gx, gy])
        ax.plot(pts[0], pts[1], 'tomato', lw=0.5, alpha=0.5)
    for gx, gy in grid_v:
        pts = S @ np.array([gx, gy])
        ax.plot(pts[0], pts[1], 'steelblue', lw=0.5, alpha=0.5)

    ax.set_xlim(-0.3, 3.3)
    ax.set_ylim(-0.3, 1.5)
    ax.set_aspect('equal')
    ax.axhline(0, color='k', lw=0.5)
    ax.axvline(0, color='k', lw=0.5)
    ax.set_title(f'Cizallamiento $k={k}$', fontsize=12)
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.2)
    ax.text(0.05, 0.95, f'det(S)={np.linalg.det(S):.1f}',
            transform=ax.transAxes, fontsize=9, va='top')

plt.suptitle('Cizallamiento horizontal $S_k$: preserva áreas, deforma ángulos', fontsize=13)
plt.tight_layout()
plt.show()
```

## 🟥 Ejemplo 4: Transformación de $\mathbb{R}^3$ a $\mathbb{R}^2$ (reducción de dimensión)

Consideremos la transformación $T: \mathbb{R}^3 \to \mathbb{R}^2$ con matriz:

$$A = \begin{pmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{pmatrix}$$

es decir, $T(x, y, z) = (x + y,\; y + z)$.

**Análisis:**

- **Dominio:** $\mathbb{R}^3$ (dimensión 3)
- **Codominio:** $\mathbb{R}^2$ (dimensión 2)
- **Rango** de $A$: las filas $(1,1,0)$ y $(0,1,1)$ son linealmente independientes → rango $= 2$
- **Nulidad:** $3 - 2 = 1$ (por rango-nulidad)
- **Núcleo:** resolvemos $A\mathbf{x} = \mathbf{0}$:
  - $x + y = 0$ y $y + z = 0$ ⟹ $x = -y$, $z = -y$
  - $\ker(T) = \operatorname{span}\{(-1, 1, -1)^\top\}$
- **Imagen:** $\operatorname{Im}(T) = \mathbb{R}^2$ (rango $= 2 = $ dim del codominio) → $T$ es **suprayectiva**
- **¿Inyectiva?** No, porque $\ker(T) \neq \{\mathbf{0}\}$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definicion de la transformacion
A = np.array([[1, 1, 0],
              [0, 1, 1]], dtype=float)

print("Transformacion T: R^3 -> R^2")
print(f"T(x,y,z) = (x+y, y+z)")
print(f"Matriz A = \n{A}")
print(f"Rango = {np.linalg.matrix_rank(A)}")

# Calculo simbolico del nucleo
A_sym = sp.Matrix([[1,1,0],[0,1,1]])
nucleo = A_sym.nullspace()
print(f"\nBase del nucleo:")
for v in nucleo:
    sp.pprint(v.T)
    print(f"Verificacion A*v = {np.array(A @ np.array(v.T.tolist()[0], dtype=float))}")

# Visualizacion: transformar puntos de la esfera unitaria en R^3
# y mostrar su imagen en R^2
np.random.seed(42)
n_puntos = 200
# Puntos en la esfera unitaria
phi = np.random.uniform(0, 2*np.pi, n_puntos)
costheta = np.random.uniform(-1, 1, n_puntos)
theta_sph = np.arccos(costheta)
X3 = np.vstack([np.sin(theta_sph)*np.cos(phi),
                np.sin(theta_sph)*np.sin(phi),
                np.cos(theta_sph)])

# Imagen en R^2
X2 = A @ X3

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# R^3 (mostrar proyeccion 2D de la esfera)
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X3[0], X3[1], X3[2], c=phi, cmap='hsv', s=15, alpha=0.6)
ax1.set_title('Dominio: esfera en $\\mathbb{R}^3$', fontsize=11)
ax1.set_xlabel('x'); ax1.set_ylabel('y'); ax1.set_zlabel('z')

ax2 = fig.add_subplot(122)
sc = ax2.scatter(X2[0], X2[1], c=phi, cmap='hsv', s=15, alpha=0.6)
ax2.set_title('Imagen en $\\mathbb{R}^2$', fontsize=11)
ax2.set_xlabel('$x+y$'); ax2.set_ylabel('$y+z$')
ax2.set_aspect('equal')
ax2.grid(True, alpha=0.3)
plt.colorbar(sc, ax=ax2, label='angulo $\\phi$')

plt.suptitle('$T: \\mathbb{R}^3 \\to \\mathbb{R}^2$ — La esfera se mapea a una elipse', fontsize=12)
plt.tight_layout()
plt.show()
```

## 📊 Resumen comparativo de los cuatro ejemplos

| Transformación | $T: \mathbb{R}^n \to \mathbb{R}^m$ | Rango | Nulidad | Inyectiva | Suprayectiva | Invertible |
|----------------|--------------------------------------|-------|---------|-----------|--------------|------------|
| Rotación $R_\theta$ | $2 \to 2$ | 2 | 0 | ✅ | ✅ | ✅ |
| Proyección $P_\ell$ | $2 \to 2$ | 1 | 1 | ❌ | ❌ | ❌ |
| Cizallamiento $S_k$ | $2 \to 2$ | 2 | 0 | ✅ | ✅ | ✅ |
| Reducción $T$ | $3 \to 2$ | 2 | 1 | ❌ | ✅ | ❌ |

**Observaciones:**
- Las transformaciones cuadradas con rango máximo son invertibles.
- Una proyección nunca es inyectiva (siempre colapsa una dimensión).
- Una transformación de $\mathbb{R}^3$ a $\mathbb{R}^2$ nunca puede ser inyectiva (forzosamente tiene nulidad $\geq 1$).

## 🎓 Conclusión

Hemos visto cómo cuatro transformaciones lineales concretas ilustran todos los conceptos del capítulo:

- **Rotación:** biyectiva, preserva distancias y áreas.
- **Proyección:** idempotente ($P^2 = P$), no inyectiva, imagen = la recta.
- **Cizallamiento:** biyectiva, preserva áreas pero deforma ángulos.
- **Reducción $\mathbb{R}^3 \to \mathbb{R}^2$:** suprayectiva, no inyectiva, núcleo unidimensional.
